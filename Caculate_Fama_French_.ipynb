```python
from sklearn.linear_model import LinearRegression

# Lấy danh sách các cổ phiếu và năm duy nhất
unique_stocks = df['Stock'].unique()
years = df['Year'].unique()

# Tạo danh sách để lưu kết quả
list_stock = []
list_year = []
list_beta_uct = []

# Chạy hồi quy cho từng cổ phiếu và từng năm
for stock in unique_stocks:
    for year in years:
        try:
            # Lọc dữ liệu theo cổ phiếu và năm
            mask = (df['Stock'] == stock) & (df['Year'] == year)
            subset_data = df[mask].copy()

            # Kiểm tra xem có đủ dữ liệu để chạy hồi quy không (ít nhất 2 điểm)
            if len(subset_data) < 2:
                continue

            # Biến phụ thuộc (y) và biến độc lập (x)
            y = subset_data['Return']
            x = subset_data[['UCT_Value', 'VNINDEX_Return']]

            # Chạy mô hình hồi quy tuyến tính
            model = LinearRegression()
            model.fit(x, y)

            # Lưu kết quả
            list_stock.append(stock)
            list_year.append(year)
            list_beta_uct.append(model.coef_[0])  # Hệ số beta của UCT_Value

        except Exception as e:
            print(f"Error for {stock}, {year}: {e}")
            continue

# Tạo DataFrame kết quả
result_df = pd.DataFrame({
    'Stock': list_stock,
    'Year': list_year,
    'Beta_UCT': list_beta_uct
})

# Hiển thị kết quả
print(result_df.head)
```
