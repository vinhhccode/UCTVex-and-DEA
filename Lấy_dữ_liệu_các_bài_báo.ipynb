```python
# Import các thư viện cần thiết cho quá trình scraping, xử lý dữ liệu, và xuất file
import os  # Thư viện để xử lý hệ thống tệp và thư mục
import re  # Thư viện để xử lý biểu thức chính quy (regular expressions) cho việc làm sạch dữ liệu
import time  # Thư viện để thêm độ trễ (delay) trong các yêu cầu HTTP, tránh bị chặn bởi server
import datetime  # Thư viện để xử lý thời gian và ngày tháng, hỗ trợ định dạng và so sánh ngày
import random  # Thư viện để tạo số ngẫu nhiên, sử dụng trong việc quản lý độ trễ ngẫu nhiên nếu cần
import json  # Thư viện để đọc/ghi dữ liệu dạng JSON, hỗ trợ lưu trữ cấu trúc dữ liệu phức tạp
import requests  # Thư viện để thực hiện các yêu cầu HTTP (GET, POST) đến các trang web
import nltk  # Thư viện Natural Language Toolkit để xử lý ngôn ngữ tự nhiên, như tóm tắt bài báo
import pandas as pd  # Thư viện để xử lý và phân tích dữ liệu dạng bảng (DataFrame)
from bs4 import BeautifulSoup  # Thư viện để parse HTML/XML, trích xuất dữ liệu từ trang web
from newspaper import Article  # Thư viện để trích xuất nội dung bài báo (tiêu đề, nội dung, tóm tắt)
from selenium import webdriver  # Thư viện để tự động hóa trình duyệt, hỗ trợ scraping động
from selenium.webdriver.chrome.options import Options  # Cấu hình tùy chọn cho trình duyệt Chrome
from selenium.webdriver.common.by import By  # Hỗ trợ tìm phần tử trong DOM bằng cách chỉ định (ID, class, v.v.)
from selenium.common.exceptions import NoSuchElementException, TimeoutException  # Xử lý lỗi khi scraping
from concurrent.futures import ThreadPoolExecutor, as_completed  # Thư viện để xử lý song song (multithreading)
from requests.adapters import HTTPAdapter  # Tăng cường khả năng kết nối HTTP với retry logic
from datetime import date, timedelta  # Hỗ trợ xử lý ngày tháng cụ thể, như tính toán khoảng thời gian
import sys

if 'google.colab' in sys.modules:
    from google.colab import drive
    drive.mount('/content/drive')
else:
    print("Đang chạy ngoài Google Colab – bỏ qua gắn Google Drive.")

# Bắt đầu phần chú thích cho từng phần của code, nhằm đảm bảo tính minh bạch và tái tạo được trong nghiên cứu khoa học

#############################################
# PHẦN 1: XỬ LÝ BÀI BÁO (Scraping và kiểm tra từ khóa)
# Mục đích: Thu thập và phân tích nội dung bài báo từ trang web, kiểm tra từ khóa liên quan đến nghiên cứu
#############################################

def configure_driver():
    """
    Cấu hình trình duyệt Chrome để chạy ở chế độ không hiển thị giao diện (headless) nhằm tối ưu hóa tài nguyên
    và phù hợp với môi trường server hoặc nghiên cứu tự động hóa quy mô lớn.

    Returns:
        webdriver.Chrome: Đối tượng trình duyệt Chrome đã được cấu hình.
    """
    # Khởi tạo tùy chọn cho trình duyệt Chrome
    chrome_options = Options()
    # Thêm tham số để chạy ở chế độ không giao diện (headless), giảm sử dụng tài nguyên
    chrome_options.add_argument("--no-sandbox")  # Cho phép chạy trên môi trường không có sandbox (như server Linux)
    chrome_options.add_argument("--headless")  # Chạy không hiển thị giao diện, phù hợp cho scraping tự động
    chrome_options.add_argument("--disable-dev-shm-usage")  # Tối ưu hóa sử dụng bộ nhớ trên hệ thống Linux
    # Trả về trình duyệt Chrome đã cấu hình
    return webdriver.Chrome(options=chrome_options)

def save_to_text_file(title, link, date, text, summary, folder_name="articles"):
    """
    Lưu nội dung bài báo vào file .txt để lưu trữ dữ liệu thô, hỗ trợ tái tạo và kiểm tra trong nghiên cứu.

    Args:
        title (str): Tiêu đề bài báo, được sử dụng làm tên file.
        link (str): Đường dẫn URL của bài báo, dùng để tham chiếu nguồn gốc.
        date (str or datetime.date): Ngày xuất bản bài báo, định dạng theo chuẩn ISO hoặc chuỗi.
        text (str): Nội dung đầy đủ của bài báo.
        summary (str): Tóm tắt nội dung bài báo, tạo bởi thư viện Newspaper.
        folder_name (str, optional): Tên thư mục lưu trữ file, mặc định là "articles". Defaults to "articles".
    """
    # Tạo tên file từ tiêu đề, loại bỏ các ký tự không hợp lệ trong tên file (như /, :, *, ?, ", <, >, |)
    filename = os.path.join(folder_name, re.sub(r'[\/:*?"<>|]', '', title) + '.txt')
    # Tạo thư mục nếu chưa tồn tại, đảm bảo không ghi đè dữ liệu
    os.makedirs(folder_name, exist_ok=True)
    # Mở file ở chế độ ghi (write), sử dụng mã hóa UTF-8 để hỗ trợ tiếng Việt
    with open(filename, 'w', encoding="utf-8") as file:
        # Ghi các thông tin bài báo theo định dạng rõ ràng, dễ đọc
        file.write(f"Title: {title}\n\n")  # Ghi tiêu đề
        file.write(f"Article Link: {link}\n\n")  # Ghi URL
        file.write(f"Publish Date: {date}\n\n")  # Ghi ngày xuất bản
        file.write(f"Article:\n{text}\n\n")  # Ghi nội dung bài báo
        file.write(f"Summary: {summary}\n\n")  # Ghi tóm tắt
    # In thông báo xác nhận đã lưu file, hỗ trợ theo dõi quá trình thực thi
    print(f"Đã lưu bài báo: {filename}")

def is_within_date_range(date_obj):
    """
    Kiểm tra xem ngày của bài báo có nằm trong khoảng thời gian từ 01/01/2014 trở đi hay không,
    nhằm đảm bảo dữ liệu chỉ thu thập trong phạm vi nghiên cứu đã định.

    Args:
        date_obj (datetime.date or str or None): Đối tượng ngày hoặc chuỗi đại diện cho ngày xuất bản.

    Returns:
        bool: True nếu ngày nằm trong khoảng từ 01/01/2014, False nếu không hoặc nếu ngày không hợp lệ.
    """
    # Kiểm tra nếu ngày là "Unknown" hoặc None, trả về False để loại bỏ bài báo không có ngày hợp lệ
    if date_obj == "Unknown" or date_obj is None:
        return False
    # Định nghĩa ngày bắt đầu của nghiên cứu (01/01/2014)
    start_date = datetime.date(2014, 1, 1)
    # So sánh ngày bài báo với ngày bắt đầu, trả về True nếu bài báo đủ điều kiện
    return start_date <= date_obj

def get_matching_keywords(text):
    """
    Xác định các từ khóa trong văn bản dựa trên 4 danh mục (Hoa Kỳ, Trung Quốc, chủ đề, và căng thẳng),
    đồng thời loại bỏ bài viết nếu chứa từ khóa không liên quan.

    Args:
        text (str): Văn bản của bài báo (tiêu đề hoặc nội dung) để kiểm tra từ khóa.

    Returns:
        list: Danh sách các từ khóa khớp nếu có ít nhất một từ từ mỗi danh mục và không chứa từ không liên quan,
              ngược lại trả về danh sách rỗng.
    """
    # Danh sách từ khóa không liên quan
    unrelated_words = [
        # Phim/Điện ảnh
        "phim", "diễn viên", "đạo diễn","showbiz",
        "bóng đá", "cầu thủ", "trận đấu","tình yêu", "mối quan hệ", "hẹn hò",
        "người yêu", "chia tay", "hôn nhân",
        "game thủ",  "âm nhạc", "ca sĩ", "bài hát","ẩm thực", "món ăn",
        "nghỉ dưỡng","trẻ em", "cha mẹ","lễ cưới","đám tang"
    ]

    # Danh sách từ khóa chính
    us_words = ["hoa kỳ", "mỹ", "washington", "nhà trắng"]
    china_words = ["trung quốc", "bắc kinh"]
    topic_words = [
        "tăng thuế", "áp thuế", "tiền tệ", "wto", "tổ chức thương mại thế giới",
        "ngân khố", "nhà máy", "thặng dư", "nợ", "thuế quan",
        "rào cản", "khoản vay", "thâm hụt", "nhân dân tệ",
        "trợ cấp", "đô la", "xuất khẩu", "sao chép bất hợp pháp", "hack", "Huawei", "mạng", "đại sứ quán", "Hồng Kông",
        "trộm cắp", "nhập khẩu", "viễn thông", "vi phạm bản quyền", "lãnh sự quán", "nhập cư", "visa", "thẻ xanh",
        "khí hậu", "năng lượng", "bán phá giá", "bản quyền", "bằng sáng chế", "giam giữ", "biểu tình", "tự trị", "bắt giữ", "bầu cử", "đàn áp", "độc tài", "biểu tình", "phong trào",
        "quyền sở hữu trí tuệ", "tấn công mạng", "bảo hộ thương mại", "tỷ giá hối đoái", "môi trường", "tôn giáo", "Tây Tạng", "Đạt Lai", "quân sự", "an ninh", "quốc phòng", "khủng bố",
        "Lầu Năm Góc", "rocket", "tàu sân bay", "tàu chiến", "đánh chìm", "Triều Tiên", "tuần tra", "máy bay phản lực", "khẩn cấp", "kho vũ khí", "tình báo", "tàu khu trục", "khiêu khích",
        "hạt nhân", "trinh sát", "vũ khí", "tàu", "giám sát", "hạm đội", "sự cố", "liên minh", "đối thủ", "Liên Hợp Quốc", "bảo mật", "gián điệp", "radar", "tàu ngầm",
        "chuyến bay", "NATO", "phi công", "hiệp ước", "Thái Bình Dương", "lãnh thổ", "vệ tinh", "hàng hải", "quân đội", "Đài Loan", "bom",
        "Đài Bắc", "bạo lực", "Biển Đông", "đảo", "an ninh thông tin", "ô nhiễm", "tài chính", "giao dịch", "chuỗi cung ứng",
        "bảo hộ thương mại", "chủ nghĩa bảo hộ", "máy bay", "đạn đạo", "xây dựng lực lượng", "lưu vong", "đầu đạn", "tên lửa", "hải quân", "triển khai",
        "dược phẩm", "nông nghiệp", "công nghệ", "đơn phương", "đại dịch", "covid", "coronavirus", "viêm phổi", "song phương", "bán dẫn",
        "nhân quyền", "thương chiến", "kiểm soát xuất khẩu", "địa chính trị", "hàng rào thuế quan", "hạn chế xuất khẩu", "cuộc chiến thương mại"
    ]
    tension_words = [
        "đe dọa", "chiến tranh", "thù địch", "gia tăng căng thẳng", "leo thang căng thẳng", "giảm căng thẳng", "vi phạm", "kiện tụng", "trừng phạt",
        "khủng bố", "phá hủy", "tổn hại", "cấm vận", "lệnh cấm", "lệnh hạn chế", "giả mạo", "cáo buộc", "thao túng", "leo thang",
        "âm mưu", "đàn áp", "lạm dụng", "kìm hãm", "phản đối", "va chạm", "ngăn chặn", "khiêu khích", "mâu thuẫn", "chia rẽ",
        "đối thủ cạnh tranh", "đình trệ", "xung đột", "cảnh báo", "trộm cắp", "làm suy yếu", "áp đặt", "tranh cãi", "đáp trả", "nổ ra",
        "hung hăng", "kẻ thù", "phòng thủ", "chiến đấu", "xấu đi", "tranh chấp", "leo thang", "trả đũa", "phân cực", "chống lại", "đối trọng", "danh sách đen",
        "tồi tệ hơn", "đối kháng", "đối đầu", "giám sát", "cản trở", "chỉ trích", "nhượng bộ", "chi phối", "xung đột", "thống trị"
    ]

    text_lower = text.lower()

    # Kiểm tra từ khóa không liên quan trước
    for word in unrelated_words:
        if word in text_lower:
            return []  # Nếu tìm thấy từ không liên quan, trả về rỗng để loại bỏ bài viết

    # Kiểm tra từ khóa chính
    found_us = [word for word in us_words if word in text_lower]
    found_china = [word for word in china_words if word in text_lower]
    found_topic = [word for word in topic_words if word in text_lower]
    found_tension = [word for word in tension_words if word in text_lower]

    # Chỉ trả về kết quả nếu có ít nhất một từ từ mỗi danh mục
    if found_us and found_china and found_topic and found_tension:
        return list(set(found_us + found_china + found_topic + found_tension))
    return []

#############################################
# PHẦN 2: XỬ LÝ SITEMAP VÀ TRUY CẬP BÀI BÁO
# Mục đích: Thu thập danh sách URL bài báo từ sitemap và xử lý từng bài báo để trích xuất dữ liệu
#############################################

def get_sitemap_urls(main_sitemap_url="https://baodautu.vn/sitemap.xml"):
    """
    Lấy danh sách URL của các sitemap từ trang web Báo Đầu Tư,
    chỉ cho hai tháng cuối năm 2023.
    """
    sitemap_urls = []
    year = 2023
    # Chỉ lấy hai tháng cuối năm: tháng 12 và 11
    for month in [12, 11]:
        url = f"https://baodautu.vn/sitemaps/news-{year}-{month}.xml"
        sitemap_urls.append(url)
    return sitemap_urls


def get_article_urls_from_sitemap(sitemap_url):
    """
    Lấy danh sách URL bài báo từ một sitemap cụ thể, bao gồm cả ngày chỉnh sửa cuối (lastmod),
    để đảm bảo thu thập dữ liệu bài báo chính xác và có thời gian cập nhật.

    Args:
        sitemap_url (str): URL của file sitemap cần parse.

    Returns:
        list: Danh sách các dictionary chứa thông tin URL và ngày chỉnh sửa cuối của bài báo.
    """
    # Thực hiện yêu cầu HTTP GET để lấy nội dung sitemap
    r = requests.get(sitemap_url)
    # Sử dụng BeautifulSoup để parse nội dung XML của sitemap
    soup = BeautifulSoup(r.content, "xml")  # Sử dụng parser XML để đảm bảo tính chính xác
    # Tìm tất cả các thẻ <url> trong sitemap
    url_tags = soup.find_all("url")
    # Khởi tạo danh sách để lưu thông tin bài báo
    articles = []
    # Lặp qua từng thẻ URL để trích xuất thông tin
    for tag in url_tags:
        # Lấy URL từ thẻ <loc>, loại bỏ khoảng trắng đầu và cuối
        loc = tag.find("loc").text.strip()
        # Lấy ngày chỉnh sửa cuối (lastmod) nếu có, cũng loại bỏ khoảng trắng
        lastmod_tag = tag.find("lastmod")
        lastmod = lastmod_tag.text.strip() if lastmod_tag and lastmod_tag.text else None
        # Thêm thông tin bài báo vào danh sách
        articles.append({"url": loc, "lastmod": lastmod})
    # Trả về danh sách các bài báo từ sitemap
    return articles

def crawl_articles_from_sitemaps():
    """
    Quét qua tất cả các sitemap đã liệt kê để thu thập thông tin bài báo,
    lọc chỉ lấy các bài báo có ngày từ 01/01/2014 trở đi, phù hợp với phạm vi nghiên cứu.

    Returns:
        list: Danh sách các dictionary chứa thông tin URL và ngày của bài báo thỏa mãn điều kiện.
    """
    # Lấy danh sách URL sitemap từ hàm get_sitemap_urls
    sitemap_urls = get_sitemap_urls()
    # In số lượng sitemap tìm thấy để theo dõi tiến trình
    print(f"Đã tìm thấy {len(sitemap_urls)} sitemap.")
    # Khởi tạo danh sách để lưu thông tin bài báo
    all_article_infos = []
    # Lặp qua từng sitemap để xử lý
    for sitemap_url in sitemap_urls:
        print(f"Đang xử lý sitemap: {sitemap_url}")
        # Lấy danh sách bài báo từ sitemap hiện tại
        articles = get_article_urls_from_sitemap(sitemap_url)
        # Xử lý từng bài báo trong sitemap
        for art in articles:
            # Chuyển đổi ngày lastmod thành đối tượng date nếu có
            if art["lastmod"]:
                try:
                    # Chuyển đổi chuỗi ISO 8601 thành đối tượng date, xử lý trường hợp kết thúc bằng 'Z'
                    art_date = datetime.datetime.fromisoformat(art["lastmod"].replace("Z", "+00:00")).date()
                except Exception as e:
                    # Nếu có lỗi trong quá trình chuyển đổi, gán date là None
                    art_date = None
            else:
                # Nếu không có lastmod, gán date là None
                art_date = None
            # Thêm ngày vào thông tin bài báo
            art["date"] = art_date
            # Kiểm tra và thêm bài báo vào danh sách nếu ngày >= 01/01/2014
            if art_date and art_date >= datetime.date(2014, 1, 1):
                all_article_infos.append(art)
    # In tổng số bài báo thu thập được để theo dõi
    print(f"Tổng số bài báo (theo sitemap) từ 01/01/2014: {len(all_article_infos)}")
    # Trả về danh sách bài báo thỏa mãn điều kiện
    return all_article_infos

def process_article(url, sitemap_date, processed_urls, max_retries=3):
    """
    Xử lý một bài báo cụ thể: trích xuất nội dung, kiểm tra từ khóa, và lưu nếu đáp ứng tiêu chí nghiên cứu.

    Args:
        url (str): URL của bài báo cần xử lý.
        sitemap_date (datetime.date or None): Ngày chỉnh sửa cuối từ sitemap (lastmod).
        processed_urls (set): Tập hợp URL đã xử lý để tránh lặp lại.
        max_retries (int, optional): Số lần thử lại khi gặp lỗi kết nối, mặc định là 3. Defaults to 3.

    Returns:
        dict or None: Thông tin bài báo nếu đáp ứng tiêu chí từ khóa, hoặc None nếu không thành công.
    """
    # Kiểm tra xem URL đã được xử lý chưa để tránh trùng lặp dữ liệu
    if url in processed_urls:
        return None
    # Thêm URL vào tập hợp đã xử lý
    processed_urls.add(url)

    # Thử tải nội dung bài báo với cơ chế retry để xử lý lỗi kết nối
    for attempt in range(max_retries):
        try:
            # Thêm độ trễ 0.5 giây để tránh bị chặn bởi server
            time.sleep(0.5)
            # Thực hiện yêu cầu GET với timeout 30 giây để tránh treo chương trình
            response = requests.get(url, timeout=30)
            # Lấy nội dung HTML của bài báo
            html = response.text
            break  # Thoát vòng lặp nếu thành công
        except requests.exceptions.ReadTimeout as e:
            # Xử lý lỗi timeout, in thông báo và thử lại
            print(f"Timeout khi truy cập {url}. Thử lại ({attempt+1}/{max_retries})...")
            if attempt == max_retries - 1:
                # Nếu hết số lần thử, trả về None để bỏ qua bài báo
                return None
        except Exception as e:
            # Xử lý các lỗi khác (ví dụ: không kết nối được), trả về None
            return None

    try:
        # Không thu thập thông tin tác giả nữa để đơn giản hóa và tập trung vào nội dung nghiên cứu
        # Sử dụng thư viện Newspaper để trích xuất nội dung bài báo từ HTML
        article = Article(url, language="vi")  # Thiết lập ngôn ngữ tiếng Việt
        article.set_html(html)  # Gán nội dung HTML đã tải
        article.parse()  # Phân tích và trích xuất nội dung, tiêu đề
        try:
            # Tạo tóm tắt bài báo bằng NLP
            article.nlp()
        except LookupError as e:
            # Nếu chưa tải gói punkt của NLTK, tải xuống để xử lý tóm tắt
            nltk.download("punkt")
            article.nlp()

        # Sử dụng ngày từ sitemap (lastmod) làm ngày xuất bản, đảm bảo tính nhất quán với dữ liệu nguồn
        date_obj = sitemap_date if sitemap_date else "Unknown"
        # Kiểm tra xem ngày có nằm trong phạm vi nghiên cứu không
        if not is_within_date_range(date_obj):
            return None

        # Lấy nội dung và tóm tắt từ bài báo
        text = article.text  # Nội dung đầy đủ
        summary = article.summary  # Tóm tắt nội dung
        title = article.title if article.title else ""  # Tiêu đề, mặc định rỗng nếu không có

        # Kiểm tra từ khóa trong tiêu đề hoặc nội dung để xác định bài báo có liên quan đến nghiên cứu không
        matches = get_matching_keywords(title) or get_matching_keywords(text)
        if matches:
            # Nếu bài báo chứa đủ từ khóa, lưu vào file .txt để lưu trữ dữ liệu thô
            save_to_text_file(title, url, date_obj, text, summary)
            # In thông báo để theo dõi tiến trình, bao gồm từ khóa khớp
            print(f"Đã duyệt bài: \"{title}\" (Ngày: {date_obj}) | Từ khóa: {matches}")
            # Trả về thông tin bài báo để tổng hợp sau này
            return {
                "title": title,  # Tiêu đề bài báo
                "url": url,  # Đường dẫn bài báo
                "date": str(date_obj),  # Ngày xuất bản dưới dạng chuỗi
                "text": text,  # Nội dung đầy đủ
                "summary": summary,  # Tóm tắt
                "matching": matches  # Danh sách từ khóa khớp
            }
        else:
            # Nếu không có từ khóa phù hợp, trả về None để bỏ qua bài báo
            return None

    except Exception as e:
        # Xử lý lỗi trong quá trình trích xuất nội dung, trả về None
        print(f"Lỗi khi xử lý bài báo {url}: {e}")
        return None

def process_article_wrapper(article_info, processed_urls, idx):
    """
    Hàm bao bọc để xử lý bài báo, in tiến trình và xử lý các trường hợp ngoại lệ,
    đảm bảo tính theo dõi trong nghiên cứu quy mô lớn.

    Args:
        article_info (dict): Thông tin bài báo từ sitemap (URL và ngày).
        processed_urls (set): Tập hợp URL đã xử lý.
        idx (int): Chỉ số thứ tự của bài báo để theo dõi tiến trình.

    Returns:
        dict or None: Kết quả xử lý bài báo hoặc thông tin URL và ngày nếu không đạt tiêu chí từ khóa.
    """
    # In thông báo để theo dõi tiến trình xử lý từng bài báo
    print(f"Đang xử lý bài báo {idx}: {article_info['url']}")
    # Gọi hàm xử lý bài báo với thông tin ngày từ sitemap
    result = process_article(article_info["url"], article_info["date"], processed_urls)
    # Nếu bài báo được duyệt, in thông tin ngày và từ khóa
    if result and "date" in result:
        print(f"Bài báo {idx} có ngày {result['date']} và chứa từ khóa: {result.get('matching', [])}")
    # Nếu bài báo không đạt tiêu chí nhưng có ngày, trả về thông tin URL và ngày
    if result is None and article_info["date"]:
        result = {"url": article_info["url"], "date": str(article_info["date"])}
    # Trả về kết quả xử lý
    return result

def crawl_and_process_articles():
    """
    Thực hiện quét và xử lý tất cả bài báo từ các sitemap của Báo Đầu Tư,
    sử dụng xử lý song song để tăng hiệu suất và thu thập dữ liệu nghiên cứu.

    Returns:
        tuple: Danh sách bài báo được duyệt (approved_articles) và danh sách tất cả bài báo từ sitemap.
    """
    # Lấy thông tin bài báo từ các sitemap
    article_infos = crawl_articles_from_sitemaps()
    # Khởi tạo tập hợp để lưu URL đã xử lý, tránh trùng lặp
    processed_urls = set()
    # Khởi tạo danh sách để lưu bài báo được duyệt
    approved_articles = []

    # Sử dụng ThreadPoolExecutor để xử lý song song, tối ưu hóa thời gian thực thi
    with ThreadPoolExecutor(max_workers=20) as executor:  # Sử dụng 20 worker để cân bằng hiệu suất và tải server
        # Tạo danh sách các tác vụ (futures) cho từng bài báo
        futures = {executor.submit(process_article_wrapper, info, processed_urls, idx): info
                   for idx, info in enumerate(article_infos, start=1)}  # Số thứ tự bắt đầu từ 1
        # Theo dõi và thu thập kết quả khi các tác vụ hoàn thành
        for future in as_completed(futures):
            # Lấy kết quả từ mỗi tác vụ
            result = future.result()
            # Nếu bài báo được duyệt (có tiêu đề), thêm vào danh sách approved_articles
            if result and "title" in result:
                approved_articles.append(result)
    # In số lượng bài báo được duyệt để theo dõi kết quả nghiên cứu
    print(f"Đã duyệt {len(approved_articles)} bài báo (đáp ứng tiêu chí từ khóa) từ sitemap.")
    # Trả về danh sách bài báo được duyệt và danh sách tất cả bài báo từ sitemap
    return approved_articles, article_infos

#############################################
# PHẦN 3: TẠO BẢNG DATAFRAME THEO THÁNG
# Mục đích: Tổng hợp dữ liệu bài báo và từ khóa theo từng tháng để phân tích thống kê
#############################################

def get_month(date_obj):
    """
    Chuyển đổi đối tượng ngày thành chuỗi định dạng "Năm-MM" (ví dụ: "2021-03"),
    phù hợp với yêu cầu phân tích dữ liệu theo tháng trong nghiên cứu.

    Args:
        date_obj (datetime.date): Đối tượng ngày cần chuyển đổi.

    Returns:
        str: Chuỗi định dạng "YYYY-MM" đại diện cho tháng.
    """
    # Sử dụng định dạng "YYYY-MM" với tháng có 2 chữ số (padding 0)
    return f"{date_obj.year}-{date_obj.month:02d}"

def create_month_summary(approved_articles, all_article_infos):
    """
    Tạo bảng tổng hợp dữ liệu bài báo theo tháng, bao gồm tổng số bài báo, số bài báo được duyệt,
    và tỷ lệ bài báo được duyệt so với tổng số, phục vụ phân tích xu hướng theo thời gian.

    Args:
        approved_articles (list): Danh sách các bài báo được duyệt theo tiêu chí từ khóa.
        all_article_infos (list): Danh sách tất cả bài báo thu thập từ sitemap.

    Returns:
        pd.DataFrame: Bảng dữ liệu tổng hợp theo tháng với các cột: 'month', 'total_count', 'approved_count', 'approved_ratio'.
    """
    # Chuẩn bị dữ liệu cho tất cả bài báo từ sitemap
    all_data = []
    for info in all_article_infos:
        if info["date"]:  # Chỉ lấy bài báo có ngày hợp lệ
            all_data.append({"date": info["date"]})
    # Tạo DataFrame từ danh sách dữ liệu
    all_df = pd.DataFrame(all_data)
    if all_df.empty:  # Kiểm tra nếu không có dữ liệu
        return pd.DataFrame()
    # Chuyển đổi cột 'date' thành định dạng datetime để xử lý
    all_df['date'] = pd.to_datetime(all_df['date'])
    # Thêm cột 'month' theo định dạng "YYYY-MM"
    all_df['month'] = all_df['date'].dt.to_period("M").astype(str)

    # Chuẩn bị dữ liệu cho bài báo được duyệt
    approved_data = []
    for art in approved_articles:
        try:
            # Chuyển đổi ngày từ chuỗi "%Y-%m-%d" thành đối tượng datetime
            d = datetime.datetime.strptime(art["date"], "%Y-%m-%d")
            approved_data.append({"date": d})
        except Exception as e:
            # Bỏ qua bài báo nếu ngày không hợp lệ, ghi log lỗi nếu cần
            print(f"Lỗi chuyển đổi ngày cho bài báo {art.get('url', 'Unknown')}: {e}")
            continue
    # Tạo DataFrame từ danh sách bài báo được duyệt
    approved_df = pd.DataFrame(approved_data)
    if not approved_df.empty:
        # Thêm cột 'month' cho bài báo được duyệt
        approved_df['month'] = approved_df['date'].dt.to_period("M").astype(str)

    # Tính tổng số bài báo theo tháng
    total_summary = all_df.groupby("month").size().reset_index(name="total_count")
    # Tính số bài báo được duyệt theo tháng, nếu không có thì tạo DataFrame rỗng với cột phù hợp
    approved_summary = approved_df.groupby("month").size().reset_index(name="approved_count") if not approved_df.empty else pd.DataFrame(columns=["month", "approved_count"])
    # Gộp dữ liệu tổng và dữ liệu duyệt, điền giá trị thiếu bằng 0
    summary = pd.merge(total_summary, approved_summary, on="month", how="left")
    # Đảm bảo cột approved_count là số nguyên, điền giá trị 0 nếu thiếu
    summary["approved_count"] = summary["approved_count"].fillna(0).astype(int)
    # Tính tỷ lệ bài báo được duyệt so với tổng số bài báo
    summary["approved_ratio"] = summary["approved_count"] / summary["total_count"]
    # Trả về bảng tổng hợp theo tháng
    return summary

def create_keyword_frequency_df(approved_articles, keyword_list):
    """
    Tạo bảng tần suất xuất hiện của các từ khóa theo tháng, tính toán tỷ lệ xuất hiện dựa trên số bài báo trong tháng,
    phục vụ phân tích tần suất và xu hướng từ khóa trong nghiên cứu.

    Args:
        approved_articles (list): Danh sách các bài báo được duyệt theo tiêu chí từ khóa.
        keyword_list (list): Danh sách các từ khóa cần phân tích tần suất.

    Returns:
        pd.DataFrame: Bảng dữ liệu tần suất từ khóa theo tháng với cột 'month' và các cột từ khóa.
    """
    # Chuẩn bị dữ liệu từ các bài báo được duyệt
    data = []
    for art in approved_articles:
        try:
            # Chuyển đổi ngày từ chuỗi "%Y-%m-%d" thành đối tượng datetime
            d = datetime.datetime.strptime(art["date"], "%Y-%m-%d")
            # Thêm thông tin ngày và danh sách từ khóa khớp vào danh sách dữ liệu
            data.append({"date": d, "matching": art.get("matching", [])})
        except Exception as e:
            # Bỏ qua bài báo nếu ngày không hợp lệ, ghi log lỗi nếu cần
            print(f"Lỗi chuyển đổi ngày cho bài báo {art.get('url', 'Unknown')}: {e}")
            continue
    # Tạo DataFrame từ danh sách dữ liệu
    df = pd.DataFrame(data)
    if df.empty:  # Kiểm tra nếu không có dữ liệu
        return pd.DataFrame()
    # Thêm cột 'month' theo định dạng "YYYY-MM"
    df['month'] = df['date'].dt.to_period("M").astype(str)
    # Tạo cột nhị phân (0 hoặc 1) cho từng từ khóa, 1 nếu từ khóa xuất hiện trong danh sách khớp
    for kw in keyword_list:
        df[kw] = df['matching'].apply(lambda lst: 1 if kw in lst else 0)
    # Tính tổng số lần xuất hiện của từng từ khóa theo tháng
    freq = df.groupby('month').agg({kw: 'sum' for kw in keyword_list})
    # Tính tổng số bài báo theo tháng
    total = df.groupby('month').size()
    # Tính tỷ lệ xuất hiện của từng từ khóa (số lần xuất hiện / tổng số bài báo trong tháng)
    freq = freq.divide(total, axis=0)
    # Đặt lại chỉ số và chuyển thành DataFrame với cột 'month'
    freq.reset_index(inplace=True)
    # Trả về bảng tần suất từ khóa theo tháng
    return freq

#############################################
# PHẦN CHÍNH (MAIN): THỰC THI CHƯƠNG TRÌNH
# Mục đích: Kết hợp các bước scraping, xử lý, và tổng hợp dữ liệu để tạo ra kết quả nghiên cứu
#############################################
if __name__ == "__main__":
    # Bước 1: Thực hiện quét và xử lý bài báo từ các sitemap của Báo Đầu Tư
    # Mục tiêu: Thu thập dữ liệu bài báo liên quan đến nghiên cứu về căng thẳng thương mại
    approved_articles, all_article_infos = crawl_and_process_articles()

    # Bước 2: Lưu danh sách các bài báo được duyệt vào file JSON để lưu trữ dữ liệu thô,
    # hỗ trợ tái tạo và kiểm tra trong nghiên cứu
    with open("approved_articles_baodautu.json", "w", encoding="utf-8") as f:
        # Sử dụng JSON để lưu trữ dữ liệu với định dạng dễ đọc, hỗ trợ tiếng Việt
        json.dump(approved_articles, f, ensure_ascii=False, indent=4)
    # In thông báo xác nhận đã lưu file
    print("\nĐã lưu kết quả vào file 'approved_articles_baodautu.json'.")

    # Bước 3: Tạo bảng tổng hợp số bài báo theo tháng (tổng số và số được duyệt)
    # Mục tiêu: Phân tích xu hướng bài báo theo thời gian theo tháng
    month_summary_df = create_month_summary(approved_articles, all_article_infos)
    # In bảng tổng hợp để theo dõi kết quả nghiên cứu
    print("\nBảng tổng hợp theo tháng (Approved / Total):")
    print(month_summary_df)

    # Bước 4: Tạo bảng tần suất từ khóa theo tháng để phân tích xu hướng từ khóa
    # Danh sách từ khóa được định nghĩa dựa trên các chủ đề nghiên cứu về thương mại và căng thẳng
    keyword_list = [ "tăng thuế", "áp thuế", "tiền tệ", "wto", "tổ chức thương mại thế giới",
        "ngân khố", "nhà máy", "thặng dư", "nợ", "thuế quan",
        "rào cản", "khoản vay", "thâm hụt", "nhân dân tệ",
        "trợ cấp", "đô la", "xuất khẩu", "sao chép bất hợp pháp", "hack", "Huawei", "mạng", "đại sứ quán", "Hồng Kông",
        "trộm cắp", "nhập khẩu", "viễn thông", "vi phạm bản quyền", "lãnh sự quán", "nhập cư", "visa", "thẻ xanh",
        "khí hậu", "năng lượng", "bán phá giá", "bản quyền", "bằng sáng chế", "giam giữ", "biểu tình", "tự trị", "bắt giữ", "bầu cử", "đàn áp", "độc tài", "biểu tình", "phong trào",
        "quyền sở hữu trí tuệ", "tấn công mạng", "bảo hộ thương mại", "tỷ giá hối đoái", "môi trường", "tôn giáo", "Tây Tạng", "Đạt Lai", "quân sự", "an ninh", "quốc phòng", "khủng bố",
        "Lầu Năm Góc", "rocket", "tàu sân bay", "tàu chiến", "đánh chìm", "Triều Tiên", "tuần tra", "máy bay phản lực", "khẩn cấp", "kho vũ khí", "tình báo", "tàu khu trục", "khiêu khích",
        "hạt nhân", "trinh sát", "vũ khí", "tàu", "giám sát", "hạm đội", "sự cố", "liên minh", "đối thủ", "Liên Hợp Quốc", "bảo mật", "gián điệp", "radar", "tàu ngầm",
        "chuyến bay", "NATO", "phi công", "hiệp ước", "Thái Bình Dương", "lãnh thổ", "vệ tinh", "hàng hải", "quân đội", "Đài Loan", "bom",
        "Đài Bắc", "bạo lực", "Biển Đông", "đảo", "an ninh thông tin", "ô nhiễm", "tài chính", "giao dịch", "chuỗi cung ứng",
        "bảo hộ thương mại", "chủ nghĩa bảo hộ", "máy bay", "đạn đạo", "xây dựng lực lượng", "lưu vong", "đầu đạn", "tên lửa", "hải quân", "triển khai",
        "dược phẩm", "nông nghiệp", "công nghệ", "đơn phương", "đại dịch", "covid", "coronavirus", "viêm phổi", "song phương", "bán dẫn",
        "nhân quyền", "thương chiến", "kiểm soát xuất khẩu", "địa chính trị", "hàng rào thuế quan", "hạn chế xuất khẩu", "cuộc chiến thương mại",
        "đe dọa", "chiến tranh", "thù địch", "gia tăng căng thẳng", "leo thang căng thẳng", "giảm căng thẳng", "vi phạm", "kiện tụng", "trừng phạt",
        "khủng bố", "phá hủy", "tổn hại", "cấm vận", "lệnh cấm", "lệnh hạn chế", "giả mạo", "cáo buộc", "thao túng", "leo thang",
        "âm mưu", "đàn áp", "lạm dụng", "kìm hãm", "phản đối", "va chạm", "ngăn chặn", "khiêu khích", "mâu thuẫn", "chia rẽ",
        "đối thủ cạnh tranh", "đình trệ", "xung đột", "cảnh báo", "trộm cắp", "làm suy yếu", "áp đặt", "tranh cãi", "đáp trả", "nổ ra",
        "hung hăng", "kẻ thù", "phòng thủ", "chiến đấu", "xấu đi", "tranh chấp", "leo thang", "trả đũa", "phân cực", "chống lại", "đối trọng", "danh sách đen",
        "tồi tệ hơn", "đối kháng", "đối đầu", "giám sát", "cản trở", "chỉ trích", "nhượng bộ", "chi phối", "xung đột", "thống trị"
    ]
    # Tạo bảng tần suất từ khóa theo tháng
    keyword_freq_df = create_keyword_frequency_df(approved_articles, keyword_list)
    # In bảng tần suất để theo dõi kết quả nghiên cứu
    print("\nBảng tần suất từ khóa theo tháng:")
    print(keyword_freq_df)

    #############################################
# Kết nối Google Drive
drive.mount('/content/drive')

# Định nghĩa đường dẫn thư mục trong Google Drive
drive_folder = "/content/drive/My Drive/BAIBAO_Reports/"

# Kiểm tra và tạo thư mục nếu chưa tồn tại
if not os.path.exists(drive_folder):
    os.makedirs(drive_folder)

# Lưu File 1: approved_articles
approved_articles_df = pd.DataFrame(approved_articles)
file_path1 = os.path.join(drive_folder, "approved_articles_baodautu.xlsx")
approved_articles_df.to_excel(file_path1, index=False)
print(f"\nĐã lưu thông tin các bài báo được duyệt vào: {file_path1}")

# Lưu File 2: month_summary
file_path2 = os.path.join(drive_folder, "month_summary_baodautu.xlsx")
month_summary_df.to_excel(file_path2, index=False)
print(f"Đã lưu bảng tổng hợp theo tháng vào: {file_path2}")

# Lưu File 3: keyword_frequency
file_path3 = os.path.join(drive_folder, "keyword_frequency_baodautu.xlsx")
keyword_freq_df.to_excel(file_path3, index=False)
print(f"Đã lưu bảng tần suất từ khóa theo tháng vào: {file_path3}")

```

```python
# Import các thư viện cần thiết cho quá trình scraping, xử lý dữ liệu, và xuất file
import os  # Thư viện để xử lý hệ thống tệp và thư mục, hỗ trợ tạo và quản lý thư mục lưu bài báo
import re  # Thư viện để xử lý biểu thức chính quy (regular expressions), dùng để làm sạch dữ liệu như tên file
import time  # Thư viện để thêm độ trễ (delay) trong các yêu cầu HTTP, tránh bị chặn bởi server
import datetime  # Thư viện để xử lý thời gian và ngày tháng, hỗ trợ định dạng và so sánh ngày trong nghiên cứu
import random  # Thư viện để tạo số ngẫu nhiên, có thể sử dụng trong việc quản lý độ trễ ngẫu nhiên nếu cần
import json  # Thư viện để đọc/ghi dữ liệu dạng JSON, hỗ trợ lưu trữ cấu trúc dữ liệu phức tạp cho bài báo
import requests  # Thư viện để thực hiện các yêu cầu HTTP (GET, POST) đến các trang web của VnEconomy
import nltk  # Thư viện Natural Language Toolkit để xử lý ngôn ngữ tự nhiên, như tóm tắt bài báo
import pandas as pd  # Thư viện để xử lý và phân tích dữ liệu dạng bảng (DataFrame) cho phân tích thống kê
from bs4 import BeautifulSoup  # Thư viện để parse HTML/XML, trích xuất dữ liệu từ trang web
from newspaper import Article  # Thư viện để trích xuất nội dung bài báo (tiêu đề, nội dung, tóm tắt)
from selenium import webdriver  # Thư viện để tự động hóa trình duyệt, hỗ trợ scraping động cho các trang phức tạp
from selenium.webdriver.chrome.options import Options  # Cấu hình tùy chọn cho trình duyệt Chrome
from selenium.webdriver.common.by import By  # Hỗ trợ tìm phần tử trong DOM bằng cách chỉ định (ID, class, v.v.)
from selenium.common.exceptions import NoSuchElementException, TimeoutException  # Xử lý lỗi khi scraping
from concurrent.futures import ThreadPoolExecutor, as_completed  # Thư viện để xử lý song song (multithreading)
from requests.adapters import HTTPAdapter  # Tăng cường khả năng kết nối HTTP với retry logic
from datetime import date, timedelta  # Hỗ trợ xử lý ngày tháng cụ thể, như tính toán khoảng thời gian
import sys


if 'google.colab' in sys.modules:
    from google.colab import drive
    drive.mount('/content/drive')
else:
    print("Đang chạy ngoài Google Colab – bỏ qua gắn Google Drive.")

# Bắt đầu phần chú thích chi tiết cho từng phần của code, nhằm đảm bảo tính minh bạch và tái tạo được trong nghiên cứu khoa học

#############################################
# PHẦN 1: XỬ LÝ BÀI BÁO (Scraping và kiểm tra từ khóa)
# Mục đích: Thu thập và phân tích nội dung bài báo từ trang web VnEconomy, kiểm tra từ khóa liên quan đến nghiên cứu
#############################################

def configure_driver():
    """
    Cấu hình trình duyệt Chrome để chạy ở chế độ không hiển thị giao diện (headless), nhằm tối ưu hóa tài nguyên
    và phù hợp với môi trường server hoặc nghiên cứu tự động hóa quy mô lớn trên VnEconomy.

    Returns:
        webdriver.Chrome: Đối tượng trình duyệt Chrome đã được cấu hình, sử dụng cho Selenium.
    """
    # Khởi tạo tùy chọn cho trình duyệt Chrome
    chrome_options = Options()
    # Thêm tham số để chạy ở chế độ không giao diện (headless), giảm sử dụng tài nguyên
    chrome_options.add_argument("--no-sandbox")  # Cho phép chạy trên môi trường không có sandbox (như server Linux)
    chrome_options.add_argument("--headless")  # Chạy không hiển thị giao diện, phù hợp cho scraping tự động
    chrome_options.add_argument("--disable-dev-shm-usage")  # Tối ưu hóa sử dụng bộ nhớ trên hệ thống Linux
    # Trả về trình duyệt Chrome đã cấu hình để sử dụng trong scraping
    return webdriver.Chrome(options=chrome_options)

def save_to_text_file(title, link, date, text, summary, folder_name="articles"):
    """
    Lưu nội dung bài báo vào file .txt để lưu trữ dữ liệu thô, hỗ trợ tái tạo và kiểm tra trong nghiên cứu khoa học.

    Args:
        title (str): Tiêu đề bài báo, được sử dụng làm tên file và đại diện cho nội dung nghiên cứu.
        link (str): Đường dẫn URL của bài báo, dùng để tham chiếu nguồn gốc từ VnEconomy.
        date (str or datetime.date): Ngày xuất bản bài báo, định dạng theo chuẩn ISO hoặc chuỗi.
        text (str): Nội dung đầy đủ của bài báo, dùng để phân tích từ khóa.
        summary (str): Tóm tắt nội dung bài báo, tạo bởi thư viện Newspaper, hỗ trợ nghiên cứu nhanh.
        folder_name (str, optional): Tên thư mục lưu trữ file, mặc định là "articles". Defaults to "articles".
    """
    # Tạo tên file từ tiêu đề, loại bỏ các ký tự không hợp lệ trong tên file (như /, :, *, ?, ", <, >, |)
    filename = os.path.join(folder_name, re.sub(r'[\/:*?"<>|]', '', title) + '.txt')
    # Tạo thư mục nếu chưa tồn tại, đảm bảo không ghi đè dữ liệu
    os.makedirs(folder_name, exist_ok=True)
    # Mở file ở chế độ ghi (write), sử dụng mã hóa UTF-8 để hỗ trợ tiếng Việt
    with open(filename, 'w', encoding="utf-8") as file:
        # Ghi các thông tin bài báo theo định dạng rõ ràng, dễ đọc, phù hợp với tài liệu nghiên cứu
        file.write(f"Title: {title}\n\n")  # Ghi tiêu đề bài báo
        file.write(f"Article Link: {link}\n\n")  # Ghi URL bài báo từ VnEconomy
        file.write(f"Publish Date: {date}\n\n")  # Ghi ngày xuất bản
        file.write(f"Article:\n{text}\n\n")  # Ghi nội dung bài báo
        file.write(f"Summary: {summary}\n\n")  # Ghi tóm tắt bài báo
    # In thông báo xác nhận đã lưu file, hỗ trợ theo dõi tiến trình nghiên cứu
    print(f"Đã lưu bài báo: {filename}")

def is_within_date_range(date_obj):
    """
    Kiểm tra xem ngày của bài báo có nằm trong khoảng thời gian từ 01/01/2014 trở đi hay không,
    nhằm đảm bảo dữ liệu chỉ thu thập trong phạm vi nghiên cứu đã định từ VnEconomy.

    Args:
        date_obj (datetime.date or str or None): Đối tượng ngày hoặc chuỗi đại diện cho ngày xuất bản từ bài báo.

    Returns:
        bool: True nếu ngày nằm trong khoảng từ 01/01/2014, False nếu không hoặc nếu ngày không hợp lệ.
    """
    # Kiểm tra nếu ngày là "Unknown" hoặc None, trả về False để loại bỏ bài báo không có ngày hợp lệ
    if date_obj == "Unknown" or date_obj is None:
        return False
    # Định nghĩa ngày bắt đầu của nghiên cứu (01/01/2014) để phù hợp với phạm vi nghiên cứu
    start_date = datetime.date(2014, 1, 1)
    # So sánh ngày bài báo với ngày bắt đầu, trả về True nếu bài báo đủ điều kiện
    return start_date <= date_obj

def get_matching_keywords(text):
    """
    Xác định các từ khóa trong văn bản dựa trên 4 danh mục (Hoa Kỳ, Trung Quốc, chủ đề, và căng thẳng),
    đồng thời loại bỏ bài viết nếu chứa từ khóa không liên quan.

    Args:
        text (str): Văn bản của bài báo (tiêu đề hoặc nội dung) để kiểm tra từ khóa.

    Returns:
        list: Danh sách các từ khóa khớp nếu có ít nhất một từ từ mỗi danh mục và không chứa từ không liên quan,
              ngược lại trả về danh sách rỗng.
    """
    # Danh sách từ khóa không liên quan
    unrelated_words = [
        # Phim/Điện ảnh
        "phim", "diễn viên", "đạo diễn","showbiz",
        "bóng đá", "cầu thủ", "trận đấu","tình yêu", "mối quan hệ", "hẹn hò",
        "người yêu", "chia tay", "hôn nhân",
        "game thủ",  "âm nhạc", "ca sĩ", "bài hát","ẩm thực", "món ăn",
        "nghỉ dưỡng","trẻ em", "cha mẹ","lễ cưới","đám tang"
    ]

    # Danh sách từ khóa chính
    us_words = ["hoa kỳ", "mỹ", "washington", "nhà trắng"]
    china_words = ["trung quốc", "bắc kinh"]
    topic_words = [
        "tăng thuế", "áp thuế", "tiền tệ", "wto", "tổ chức thương mại thế giới",
        "ngân khố", "nhà máy", "thặng dư", "nợ", "thuế quan",
        "rào cản", "khoản vay", "thâm hụt", "nhân dân tệ",
        "trợ cấp", "đô la", "xuất khẩu", "sao chép bất hợp pháp", "hack", "Huawei", "mạng", "đại sứ quán", "Hồng Kông",
        "trộm cắp", "nhập khẩu", "viễn thông", "vi phạm bản quyền", "lãnh sự quán", "nhập cư", "visa", "thẻ xanh",
        "khí hậu", "năng lượng", "bán phá giá", "bản quyền", "bằng sáng chế", "giam giữ", "biểu tình", "tự trị", "bắt giữ", "bầu cử", "đàn áp", "độc tài", "biểu tình", "phong trào",
        "quyền sở hữu trí tuệ", "tấn công mạng", "bảo hộ thương mại", "tỷ giá hối đoái", "môi trường", "tôn giáo", "Tây Tạng", "Đạt Lai", "quân sự", "an ninh", "quốc phòng", "khủng bố",
        "Lầu Năm Góc", "rocket", "tàu sân bay", "tàu chiến", "đánh chìm", "Triều Tiên", "tuần tra", "máy bay phản lực", "khẩn cấp", "kho vũ khí", "tình báo", "tàu khu trục", "khiêu khích",
        "hạt nhân", "trinh sát", "vũ khí", "tàu", "giám sát", "hạm đội", "sự cố", "liên minh", "đối thủ", "Liên Hợp Quốc", "bảo mật", "gián điệp", "radar", "tàu ngầm",
        "chuyến bay", "NATO", "phi công", "hiệp ước", "Thái Bình Dương", "lãnh thổ", "vệ tinh", "hàng hải", "quân đội", "Đài Loan", "bom",
        "Đài Bắc", "bạo lực", "Biển Đông", "đảo", "an ninh thông tin", "ô nhiễm", "tài chính", "giao dịch", "chuỗi cung ứng",
        "bảo hộ thương mại", "chủ nghĩa bảo hộ", "máy bay", "đạn đạo", "xây dựng lực lượng", "lưu vong", "đầu đạn", "tên lửa", "hải quân", "triển khai",
        "dược phẩm", "nông nghiệp", "công nghệ", "đơn phương", "đại dịch", "covid", "coronavirus", "viêm phổi", "song phương", "bán dẫn",
        "nhân quyền", "thương chiến", "kiểm soát xuất khẩu", "địa chính trị", "hàng rào thuế quan", "hạn chế xuất khẩu", "cuộc chiến thương mại"
    ]
    tension_words = [
        "đe dọa", "chiến tranh", "thù địch", "gia tăng căng thẳng", "leo thang căng thẳng", "giảm căng thẳng", "vi phạm", "kiện tụng", "trừng phạt",
        "khủng bố", "phá hủy", "tổn hại", "cấm vận", "lệnh cấm", "lệnh hạn chế", "giả mạo", "cáo buộc", "thao túng", "leo thang",
        "âm mưu", "đàn áp", "lạm dụng", "kìm hãm", "phản đối", "va chạm", "ngăn chặn", "khiêu khích", "mâu thuẫn", "chia rẽ",
        "đối thủ cạnh tranh", "đình trệ", "xung đột", "cảnh báo", "trộm cắp", "làm suy yếu", "áp đặt", "tranh cãi", "đáp trả", "nổ ra",
        "hung hăng", "kẻ thù", "phòng thủ", "chiến đấu", "xấu đi", "tranh chấp", "leo thang", "trả đũa", "phân cực", "chống lại", "đối trọng", "danh sách đen",
        "tồi tệ hơn", "đối kháng", "đối đầu", "giám sát", "cản trở", "chỉ trích", "nhượng bộ", "chi phối", "xung đột", "thống trị"
    ]

    text_lower = text.lower()

    # Kiểm tra từ khóa không liên quan trước
    for word in unrelated_words:
        if word in text_lower:
            return []  # Nếu tìm thấy từ không liên quan, trả về rỗng để loại bỏ bài viết

    # Kiểm tra từ khóa chính
    found_us = [word for word in us_words if word in text_lower]
    found_china = [word for word in china_words if word in text_lower]
    found_topic = [word for word in topic_words if word in text_lower]
    found_tension = [word for word in tension_words if word in text_lower]

    # Chỉ trả về kết quả nếu có ít nhất một từ từ mỗi danh mục
    if found_us and found_china and found_topic and found_tension:
        return list(set(found_us + found_china + found_topic + found_tension))
    return []

#############################################
# PHẦN 2: XỬ LÝ SITEMAP VÀ TRUY CẬP BÀI BÁO
# Mục đích: Thu thập danh sách URL bài báo từ sitemap của VnEconomy và xử lý từng bài báo để trích xuất dữ liệu
#############################################

def get_sitemap_urls(main_sitemap_url="https://vneconomy.vn/sitemap.xml"):
    """
    Lấy danh sách URL của các sitemap từ trang web VnEconomy, lọc chỉ lấy các sitemap
    từ https://vneconomy.vn/sitemap/news-2025-2.xml đến https://vneconomy.vn/sitemap/news-2014-1.xml,
    phù hợp với phạm vi thời gian nghiên cứu.

    Args:
        main_sitemap_url (str, optional): URL của sitemap chính của VnEconomy, mặc định là "https://vneconomy.vn/sitemap.xml". Defaults to "https://vneconomy.vn/sitemap.xml".

    Returns:
        list: Danh sách các URL sitemap được lọc theo năm và tháng trong phạm vi nghiên cứu.
    """
    # Thực hiện yêu cầu HTTP GET để lấy nội dung sitemap chính
    r = requests.get(main_sitemap_url)
    # Sử dụng BeautifulSoup để parse nội dung XML của sitemap
    soup = BeautifulSoup(r.content, "xml")  # Sử dụng parser XML để đảm bảo tính chính xác
    # Tìm tất cả các thẻ <sitemap> trong sitemap chính
    sitemap_tags = soup.find_all("sitemap")
    # Trích xuất URL từ các thẻ <loc> trong sitemap
    sitemap_urls = [tag.find("loc").text for tag in sitemap_tags]

    # Định nghĩa URL bắt đầu và kết thúc theo phạm vi nghiên cứu
    start_url = "https://vneconomy.vn/sitemap/news-2024-12.xml"  # Sitemap mới nhất (tháng 2/2025)
    end_url = "https://vneconomy.vn/sitemap/news-2024-11.xml"  # Sitemap đầu tiên (tháng 1/2014)
    # Kiểm tra xem các URL bắt đầu và kết thúc có tồn tại trong danh sách không
    if start_url in sitemap_urls and end_url in sitemap_urls:
        # Tìm chỉ số của URL bắt đầu và kết thúc trong danh sách
        start_index = sitemap_urls.index(start_url)
        end_index = sitemap_urls.index(end_url)
        # Giả sử danh sách được sắp xếp từ mới nhất đến cũ nhất (2025 đến 2014)
        # Nếu start_index < end_index, lấy slice từ start_index đến end_index + 1, ngược lại lấy slice đảo
        if start_index <= end_index:
            filtered = sitemap_urls[start_index:end_index + 1]  # Lấy từ 2025-2 đến 2014-1
        else:
            filtered = sitemap_urls[end_index:start_index + 1]  # Lấy theo thứ tự ngược lại nếu cần
        return filtered  # Trả về danh sách sitemap đã lọc
    else:
        # Nếu không tìm thấy URL bắt đầu hoặc kết thúc, trả về danh sách rỗng
        print(f"Không tìm thấy {start_url} hoặc {end_url} trong sitemap chính.")
        return []

def get_article_urls_from_sitemap(sitemap_url):
    """
    Lấy danh sách URL bài báo từ một sitemap cụ thể của VnEconomy, bao gồm cả ngày chỉnh sửa cuối (lastmod),
    để đảm bảo thu thập dữ liệu bài báo chính xác và có thời gian cập nhật.

    Args:
        sitemap_url (str): URL của file sitemap cần parse từ VnEconomy.

    Returns:
        list: Danh sách các dictionary chứa thông tin URL và ngày chỉnh sửa cuối của bài báo.
    """
    # Thực hiện yêu cầu HTTP GET để lấy nội dung sitemap
    r = requests.get(sitemap_url)
    # Sử dụng BeautifulSoup để parse nội dung XML của sitemap
    soup = BeautifulSoup(r.content, "xml")  # Sử dụng parser XML để đảm bảo tính chính xác
    # Tìm tất cả các thẻ <url> trong sitemap
    url_tags = soup.find_all("url")
    # Khởi tạo danh sách để lưu thông tin bài báo
    articles = []
    # Lặp qua từng thẻ URL để trích xuất thông tin
    for tag in url_tags:
        # Lấy URL từ thẻ <loc>
        loc = tag.find("loc").text
        # Lấy ngày chỉnh sửa cuối (lastmod) nếu có
        lastmod = tag.find("lastmod").text if tag.find("lastmod") else None
        # Thêm thông tin bài báo vào danh sách
        articles.append({"url": loc, "lastmod": lastmod})
    # Trả về danh sách các bài báo từ sitemap
    return articles

def crawl_articles_from_sitemaps():
    """
    Quét qua tất cả các sitemap đã lọc từ VnEconomy để thu thập thông tin bài báo,
    lọc chỉ lấy các bài báo có ngày từ 01/01/2014 trở đi, phù hợp với phạm vi nghiên cứu.

    Returns:
        list: Danh sách các dictionary chứa thông tin URL và ngày của bài báo thỏa mãn điều kiện.
    """
    # Lấy danh sách URL sitemap từ hàm get_sitemap_urls
    sitemap_urls = get_sitemap_urls()
    # In số lượng sitemap tìm thấy để theo dõi tiến trình nghiên cứu
    print(f"Đã tìm thấy {len(sitemap_urls)} sitemap từ VnEconomy.")
    # Khởi tạo danh sách để lưu thông tin bài báo
    all_article_infos = []
    # Lặp qua từng sitemap để xử lý
    for sitemap_url in sitemap_urls:
        print(f"Đang xử lý sitemap: {sitemap_url}")
        # Lấy danh sách bài báo từ sitemap hiện tại
        articles = get_article_urls_from_sitemap(sitemap_url)
        # Xử lý từng bài báo trong sitemap
        for art in articles:
            # Chuyển đổi ngày lastmod thành đối tượng date nếu có
            if art["lastmod"]:
                try:
                    # Chuyển đổi chuỗi ISO 8601 thành đối tượng date, xử lý trường hợp kết thúc bằng 'Z'
                    art_date = datetime.datetime.fromisoformat(art["lastmod"].replace("Z", "+00:00")).date()
                except Exception as e:
                    # Nếu có lỗi trong quá trình chuyển đổi, gán date là None
                    art_date = None
            else:
                # Nếu không có lastmod, gán date là None
                art_date = None
            # Thêm ngày vào thông tin bài báo
            art["date"] = art_date
            # Kiểm tra và thêm bài báo vào danh sách nếu ngày >= 01/01/2014
            if art_date and art_date >= datetime.date(2014, 1, 1):
                all_article_infos.append(art)
    # In tổng số bài báo thu thập được để theo dõi
    print(f"Tổng số bài báo (theo sitemap) từ 01/01/2014: {len(all_article_infos)}")
    # Trả về danh sách bài báo thỏa mãn điều kiện nghiên cứu
    return all_article_infos

def process_article(url, sitemap_date, processed_urls, max_retries=3):
    """
    Xử lý một bài báo cụ thể từ VnEconomy: trích xuất nội dung, kiểm tra từ khóa, và lưu nếu đáp ứng tiêu chí nghiên cứu.

    Args:
        url (str): URL của bài báo cần xử lý từ VnEconomy.
        sitemap_date (datetime.date or None): Ngày chỉnh sửa cuối từ sitemap (lastmod).
        processed_urls (set): Tập hợp URL đã xử lý để tránh lặp lại trong nghiên cứu.
        max_retries (int, optional): Số lần thử lại khi gặp lỗi kết nối, mặc định là 3. Defaults to 3.

    Returns:
        dict or None: Thông tin bài báo nếu đáp ứng tiêu chí từ khóa, hoặc None nếu không thành công.
    """
    # Kiểm tra xem URL đã được xử lý chưa để tránh trùng lặp dữ liệu trong nghiên cứu
    if url in processed_urls:
        return None
    # Thêm URL vào tập hợp đã xử lý
    processed_urls.add(url)

    # Thử tải nội dung bài báo với cơ chế retry để xử lý lỗi kết nối, đảm bảo tính ổn định trong scraping
    for attempt in range(max_retries):
        try:
            # Thêm độ trễ 0.5 giây để tránh bị chặn bởi server của VnEconomy
            time.sleep(0.5)
            # Thực hiện yêu cầu GET với timeout 30 giây để tránh treo chương trình
            response = requests.get(url, timeout=30)
            # Lấy nội dung HTML của bài báo
            html = response.text
            break  # Thoát vòng lặp nếu thành công
        except requests.exceptions.ReadTimeout as e:
            # Xử lý lỗi timeout, in thông báo và thử lại
            print(f"Timeout khi truy cập {url}. Thử lại ({attempt+1}/{max_retries})...")
            if attempt == max_retries - 1:
                # Nếu hết số lần thử, trả về None để bỏ qua bài báo
                return None
        except Exception as e:
            # Xử lý các lỗi khác (ví dụ: không kết nối được), trả về None
            return None

    try:
        # Không thu thập thông tin tác giả nữa để đơn giản hóa và tập trung vào nội dung nghiên cứu
        # Sử dụng thư viện Newspaper để trích xuất nội dung bài báo từ HTML
        article = Article(url, language="vi")  # Thiết lập ngôn ngữ tiếng Việt
        article.set_html(html)  # Gán nội dung HTML đã tải
        article.parse()  # Phân tích và trích xuất nội dung, tiêu đề
        try:
            # Tạo tóm tắt bài báo bằng NLP để hỗ trợ phân tích nhanh
            article.nlp()
        except LookupError as e:
            # Nếu chưa tải gói punkt của NLTK, tải xuống để xử lý tóm tắt
            nltk.download("punkt")
            article.nlp()

        # Sử dụng ngày từ sitemap (lastmod) làm ngày xuất bản, đảm bảo tính nhất quán với dữ liệu nguồn từ VnEconomy
        date_obj = sitemap_date if sitemap_date else "Unknown"
        # Kiểm tra xem ngày có nằm trong phạm vi nghiên cứu không
        if not is_within_date_range(date_obj):
            return None

        # Lấy nội dung và tóm tắt từ bài báo
        text = article.text  # Nội dung đầy đủ để phân tích từ khóa
        summary = article.summary  # Tóm tắt nội dung để nghiên cứu nhanh
        title = article.title if article.title else ""  # Tiêu đề, mặc định rỗng nếu không có

        # Kiểm tra từ khóa trong tiêu đề hoặc nội dung để xác định bài báo có liên quan đến nghiên cứu không
        matches = get_matching_keywords(title) or get_matching_keywords(text)
        if matches:
            # Nếu bài báo chứa đủ từ khóa, lưu vào file .txt để lưu trữ dữ liệu thô
            save_to_text_file(title, url, date_obj, text, summary)
            # In thông báo để theo dõi tiến trình, bao gồm từ khóa khớp
            print(f"Đã duyệt bài: \"{title}\" (Ngày: {date_obj}) | Từ khóa: {matches}")
            # Trả về thông tin bài báo để tổng hợp sau này
            return {
                "title": title,  # Tiêu đề bài báo
                "url": url,  # Đường dẫn bài báo từ VnEconomy
                "date": str(date_obj),  # Ngày xuất bản dưới dạng chuỗi
                "text": text,  # Nội dung đầy đủ
                "summary": summary,  # Tóm tắt
                "matching": matches  # Danh sách từ khóa khớp
            }
        else:
            # Nếu không có từ khóa phù hợp, trả về None để bỏ qua bài báo
            return None

    except Exception as e:
        # Xử lý lỗi trong quá trình trích xuất nội dung, trả về None và in lỗi để theo dõi
        print(f"Lỗi khi xử lý bài báo {url}: {e}")
        return None

def process_article_wrapper(article_info, processed_urls, idx):
    """
    Hàm bao bọc để xử lý bài báo, in tiến trình và xử lý các trường hợp ngoại lệ,
    đảm bảo tính theo dõi trong nghiên cứu quy mô lớn từ VnEconomy.

    Args:
        article_info (dict): Thông tin bài báo từ sitemap (URL và ngày).
        processed_urls (set): Tập hợp URL đã xử lý.
        idx (int): Chỉ số thứ tự của bài báo để theo dõi tiến trình nghiên cứu.

    Returns:
        dict or None: Kết quả xử lý bài báo hoặc thông tin URL và ngày nếu không đạt tiêu chí từ khóa.
    """
    # In thông báo để theo dõi tiến trình xử lý từng bài báo từ VnEconomy
    print(f"Đang xử lý bài báo {idx}: {article_info['url']}")
    # Gọi hàm xử lý bài báo với thông tin ngày từ sitemap
    result = process_article(article_info["url"], article_info["date"], processed_urls)
    # Nếu bài báo được duyệt, in thông tin ngày và từ khóa
    if result and "date" in result:
        print(f"Bài báo {idx} có ngày {result['date']} và chứa từ khóa: {result.get('matching', [])}")
    # Nếu bài báo không đạt tiêu chí nhưng có ngày, trả về thông tin URL và ngày
    if result is None and article_info["date"]:
        result = {"url": article_info["url"], "date": str(article_info["date"])}
    # Trả về kết quả xử lý
    return result

def crawl_and_process_articles():
    """
    Thực hiện quét và xử lý tất cả bài báo từ các sitemap của VnEconomy,
    sử dụng xử lý song song để tăng hiệu suất và thu thập dữ liệu nghiên cứu.

    Returns:
        tuple: Danh sách bài báo được duyệt (approved_articles) và danh sách tất cả bài báo từ sitemap.
    """
    # Lấy thông tin bài báo từ các sitemap của VnEconomy
    article_infos = crawl_articles_from_sitemaps()
    # Khởi tạo tập hợp để lưu URL đã xử lý, tránh trùng lặp
    processed_urls = set()
    # Khởi tạo danh sách để lưu bài báo được duyệt
    approved_articles = []

    # Sử dụng ThreadPoolExecutor để xử lý song song, tối ưu hóa thời gian thực thi trên VnEconomy
    with ThreadPoolExecutor(max_workers=20) as executor:  # Sử dụng 20 worker để cân bằng hiệu suất và tải server
        # Tạo danh sách các tác vụ (futures) cho từng bài báo
        futures = {executor.submit(process_article_wrapper, info, processed_urls, idx): info
                   for idx, info in enumerate(article_infos, start=1)}  # Số thứ tự bắt đầu từ 1
        # Theo dõi và thu thập kết quả khi các tác vụ hoàn thành
        for future in as_completed(futures):
            # Lấy kết quả từ mỗi tác vụ
            result = future.result()
            # Nếu bài báo được duyệt (có tiêu đề), thêm vào danh sách approved_articles
            if result and "title" in result:
                approved_articles.append(result)
    # In số lượng bài báo được duyệt để theo dõi kết quả nghiên cứu
    print(f"Đã duyệt {len(approved_articles)} bài báo (đáp ứng tiêu chí từ khóa) từ sitemap của VnEconomy.")
    # Trả về danh sách bài báo được duyệt và danh sách tất cả bài báo từ sitemap
    return approved_articles, article_infos

#############################################
# PHẦN 3: TẠO BẢNG DATAFRAME THEO THÁNG
# Mục đích: Tổng hợp dữ liệu bài báo và từ khóa từ VnEconomy theo từng tháng để phân tích thống kê
#############################################

def get_month(date_obj):
    """
    Chuyển đổi đối tượng ngày thành chuỗi định dạng "Năm-MM" (ví dụ: "2021-03"),
    phù hợp với yêu cầu phân tích dữ liệu theo tháng trong nghiên cứu từ VnEconomy.

    Args:
        date_obj (datetime.date): Đối tượng ngày cần chuyển đổi.

    Returns:
        str: Chuỗi định dạng "YYYY-MM" đại diện cho tháng, dùng để nhóm dữ liệu.
    """
    # Sử dụng định dạng "YYYY-MM" với tháng có 2 chữ số (padding 0) để đảm bảo thống nhất
    return f"{date_obj.year}-{date_obj.month:02d}"

def create_month_summary(approved_articles, all_article_infos):
    """
    Tạo bảng tổng hợp dữ liệu bài báo từ VnEconomy theo tháng, bao gồm tổng số bài báo, số bài báo được duyệt,
    và tỷ lệ bài báo được duyệt so với tổng số, phục vụ phân tích xu hướng theo thời gian.

    Args:
        approved_articles (list): Danh sách các bài báo được duyệt theo tiêu chí từ khóa từ VnEconomy.
        all_article_infos (list): Danh sách tất cả bài báo thu thập từ sitemap của VnEconomy.

    Returns:
        pd.DataFrame: Bảng dữ liệu tổng hợp theo tháng với các cột: 'month', 'total_count', 'approved_count', 'approved_ratio'.
    """
    # Chuẩn bị dữ liệu cho tất cả bài báo từ sitemap
    all_data = []
    for info in all_article_infos:
        if info["date"]:  # Chỉ lấy bài báo có ngày hợp lệ
            all_data.append({"date": info["date"]})
    # Tạo DataFrame từ danh sách dữ liệu
    all_df = pd.DataFrame(all_data)
    if all_df.empty:  # Kiểm tra nếu không có dữ liệu
        return pd.DataFrame()
    # Chuyển đổi cột 'date' thành định dạng datetime để xử lý
    all_df['date'] = pd.to_datetime(all_df['date'])
    # Thêm cột 'month' theo định dạng "YYYY-MM"
    all_df['month'] = all_df['date'].dt.to_period("M").astype(str)

    # Chuẩn bị dữ liệu cho bài báo được duyệt
    approved_data = []
    for art in approved_articles:
        try:
            # Chuyển đổi ngày từ chuỗi "%Y-%m-%d" thành đối tượng datetime
            d = datetime.datetime.strptime(art["date"], "%Y-%m-%d")
            approved_data.append({"date": d})
        except Exception as e:
            # Bỏ qua bài báo nếu ngày không hợp lệ, ghi log lỗi nếu cần cho nghiên cứu
            print(f"Lỗi chuyển đổi ngày cho bài báo {art.get('url', 'Unknown')} từ VnEconomy: {e}")
            continue
    # Tạo DataFrame từ danh sách bài báo được duyệt
    approved_df = pd.DataFrame(approved_data)
    if not approved_df.empty:
        # Thêm cột 'month' cho bài báo được duyệt
        approved_df['month'] = approved_df['date'].dt.to_period("M").astype(str)

    # Tính tổng số bài báo theo tháng
    total_summary = all_df.groupby("month").size().reset_index(name="total_count")
    # Tính số bài báo được duyệt theo tháng, nếu không có thì tạo DataFrame rỗng với cột phù hợp
    approved_summary = approved_df.groupby("month").size().reset_index(name="approved_count") if not approved_df.empty else pd.DataFrame(columns=["month", "approved_count"])
    # Gộp dữ liệu tổng và dữ liệu duyệt, điền giá trị thiếu bằng 0
    summary = pd.merge(total_summary, approved_summary, on="month", how="left")
    # Đảm bảo cột approved_count là số nguyên, điền giá trị 0 nếu thiếu
    summary["approved_count"] = summary["approved_count"].fillna(0).astype(int)
    # Tính tỷ lệ bài báo được duyệt so với tổng số bài báo
    summary["approved_ratio"] = summary["approved_count"] / summary["total_count"]
    # Trả về bảng tổng hợp theo tháng
    return summary

def create_keyword_frequency_df(approved_articles, keyword_list):
    """
    Tạo bảng tần suất xuất hiện của các từ khóa từ VnEconomy theo tháng, tính toán tỷ lệ xuất hiện dựa trên số bài báo trong tháng,
    phục vụ phân tích tần suất và xu hướng từ khóa trong nghiên cứu.

    Args:
        approved_articles (list): Danh sách các bài báo được duyệt theo tiêu chí từ khóa từ VnEconomy.
        keyword_list (list): Danh sách các từ khóa cần phân tích tần suất, dựa trên các chủ đề nghiên cứu.

    Returns:
        pd.DataFrame: Bảng dữ liệu tần suất từ khóa theo tháng với cột 'month' và các cột từ khóa.
    """
    # Chuẩn bị dữ liệu từ các bài báo được duyệt
    data = []
    for art in approved_articles:
        try:
            # Chuyển đổi ngày từ chuỗi "%Y-%m-%d" thành đối tượng datetime
            d = datetime.datetime.strptime(art["date"], "%Y-%m-%d")
            # Thêm thông tin ngày và danh sách từ khóa khớp vào danh sách dữ liệu
            data.append({"date": d, "matching": art.get("matching", [])})
        except Exception as e:
            # Bỏ qua bài báo nếu ngày không hợp lệ, ghi log lỗi nếu cần
            print(f"Lỗi chuyển đổi ngày cho bài báo {art.get('url', 'Unknown')} từ VnEconomy: {e}")
            continue
    # Tạo DataFrame từ danh sách dữ liệu
    df = pd.DataFrame(data)
    if df.empty:  # Kiểm tra nếu không có dữ liệu
        return pd.DataFrame()
    # Thêm cột 'month' theo định dạng "YYYY-MM"
    df['month'] = df['date'].dt.to_period("M").astype(str)
    # Tạo cột nhị phân (0 hoặc 1) cho từng từ khóa, 1 nếu từ khóa xuất hiện trong danh sách khớp
    for kw in keyword_list:
        df[kw] = df['matching'].apply(lambda lst: 1 if kw in lst else 0)
    # Tính tổng số lần xuất hiện của từng từ khóa theo tháng
    freq = df.groupby('month').agg({kw: 'sum' for kw in keyword_list})
    # Tính tổng số bài báo theo tháng
    total = df.groupby('month').size()
    # Tính tỷ lệ xuất hiện của từng từ khóa (số lần xuất hiện / tổng số bài báo trong tháng)
    freq = freq.divide(total, axis=0)
    # Đặt lại chỉ số và chuyển thành DataFrame với cột 'month'
    freq.reset_index(inplace=True)
    # Trả về bảng tần suất từ khóa theo tháng
    return freq

#############################################
# PHẦN CHÍNH (MAIN): THỰC THI CHƯƠNG TRÌNH
# Mục đích: Kết hợp các bước scraping, xử lý, và tổng hợp dữ liệu từ VnEconomy để tạo ra kết quả nghiên cứu
#############################################
if __name__ == "__main__":
    # Bước 1: Thực hiện quét và xử lý bài báo từ các sitemap của VnEconomy
    # Mục tiêu: Thu thập dữ liệu bài báo liên quan đến nghiên cứu về căng thẳng thương mại
    approved_articles, all_article_infos = crawl_and_process_articles()

    # Bước 2: Lưu danh sách các bài báo được duyệt vào file JSON để lưu trữ dữ liệu thô,
    # hỗ trợ tái tạo và kiểm tra trong nghiên cứu
    with open("approved_articles_vneconomy.json", "w", encoding="utf-8") as f:
        # Sử dụng JSON để lưu trữ dữ liệu với định dạng dễ đọc, hỗ trợ tiếng Việt
        json.dump(approved_articles, f, ensure_ascii=False, indent=4)
    # In thông báo xác nhận đã lưu file
    print("\nĐã lưu kết quả vào file 'approved_articles_vneconomy.json'.")

    # Bước 3: Tạo bảng tổng hợp số bài báo theo tháng (tổng số và số được duyệt)
    # Mục tiêu: Phân tích xu hướng bài báo theo thời gian theo tháng từ VnEconomy
    month_summary_df = create_month_summary(approved_articles, all_article_infos)
    # In bảng tổng hợp để theo dõi kết quả nghiên cứu
    print("\nBảng tổng hợp theo tháng (Approved / Total):")
    print(month_summary_df)

    # Bước 4: Tạo bảng tần suất từ khóa theo tháng để phân tích xu hướng từ khóa
    # Danh sách từ khóa được định nghĩa dựa trên các chủ đề nghiên cứu về thương mại và căng thẳng
    keyword_list = [ "tăng thuế", "áp thuế", "tiền tệ", "wto", "tổ chức thương mại thế giới",
        "ngân khố", "nhà máy", "thặng dư", "nợ", "thuế quan",
        "rào cản", "khoản vay", "thâm hụt", "nhân dân tệ",
        "trợ cấp", "đô la", "xuất khẩu", "sao chép bất hợp pháp", "hack", "Huawei", "mạng", "đại sứ quán", "Hồng Kông",
        "trộm cắp", "nhập khẩu", "viễn thông", "vi phạm bản quyền", "lãnh sự quán", "nhập cư", "visa", "thẻ xanh",
        "khí hậu", "năng lượng", "bán phá giá", "bản quyền", "bằng sáng chế", "giam giữ", "biểu tình", "tự trị", "bắt giữ", "bầu cử", "đàn áp", "độc tài", "biểu tình", "phong trào",
        "quyền sở hữu trí tuệ", "tấn công mạng", "bảo hộ thương mại", "tỷ giá hối đoái", "môi trường", "tôn giáo", "Tây Tạng", "Đạt Lai", "quân sự", "an ninh", "quốc phòng", "khủng bố",
        "Lầu Năm Góc", "rocket", "tàu sân bay", "tàu chiến", "đánh chìm", "Triều Tiên", "tuần tra", "máy bay phản lực", "khẩn cấp", "kho vũ khí", "tình báo", "tàu khu trục", "khiêu khích",
        "hạt nhân", "trinh sát", "vũ khí", "tàu", "giám sát", "hạm đội", "sự cố", "liên minh", "đối thủ", "Liên Hợp Quốc", "bảo mật", "gián điệp", "radar", "tàu ngầm",
        "chuyến bay", "NATO", "phi công", "hiệp ước", "Thái Bình Dương", "lãnh thổ", "vệ tinh", "hàng hải", "quân đội", "Đài Loan", "bom",
        "Đài Bắc", "bạo lực", "Biển Đông", "đảo", "an ninh thông tin", "ô nhiễm", "tài chính", "giao dịch", "chuỗi cung ứng",
        "bảo hộ thương mại", "chủ nghĩa bảo hộ", "máy bay", "đạn đạo", "xây dựng lực lượng", "lưu vong", "đầu đạn", "tên lửa", "hải quân", "triển khai",
        "dược phẩm", "nông nghiệp", "công nghệ", "đơn phương", "đại dịch", "covid", "coronavirus", "viêm phổi", "song phương", "bán dẫn",
        "nhân quyền", "thương chiến", "kiểm soát xuất khẩu", "địa chính trị", "hàng rào thuế quan", "hạn chế xuất khẩu", "cuộc chiến thương mại",
        "đe dọa", "chiến tranh", "thù địch", "gia tăng căng thẳng", "leo thang căng thẳng", "giảm căng thẳng", "vi phạm", "kiện tụng", "trừng phạt",
        "khủng bố", "phá hủy", "tổn hại", "cấm vận", "lệnh cấm", "lệnh hạn chế", "giả mạo", "cáo buộc", "thao túng", "leo thang",
        "âm mưu", "đàn áp", "lạm dụng", "kìm hãm", "phản đối", "va chạm", "ngăn chặn", "khiêu khích", "mâu thuẫn", "chia rẽ",
        "đối thủ cạnh tranh", "đình trệ", "xung đột", "cảnh báo", "trộm cắp", "làm suy yếu", "áp đặt", "tranh cãi", "đáp trả", "nổ ra",
        "hung hăng", "kẻ thù", "phòng thủ", "chiến đấu", "xấu đi", "tranh chấp", "leo thang", "trả đũa", "phân cực", "chống lại", "đối trọng", "danh sách đen",
        "tồi tệ hơn", "đối kháng", "đối đầu", "giám sát", "cản trở", "chỉ trích", "nhượng bộ", "chi phối", "xung đột", "thống trị"
    ]
    # Tạo bảng tần suất từ khóa theo tháng
    keyword_freq_df = create_keyword_frequency_df(approved_articles, keyword_list)
    # In bảng tần suất để theo dõi kết quả nghiên cứu
    print("\nBảng tần suất từ khóa theo tháng:")
    print(keyword_freq_df)

    #############################################
    # Bước 5: Lưu dữ liệu ra 3 file Excel tự động và tải về (nếu chạy trên Google Colab)
# Kết nối Google Drive
drive.mount('/content/drive')

# Định nghĩa đường dẫn thư mục trong Google Drive
drive_folder = "/content/drive/My Drive/BAIBAO_Reports/"

# Kiểm tra và tạo thư mục nếu chưa tồn tại
if not os.path.exists(drive_folder):
    os.makedirs(drive_folder)

# Lưu File 1: approved_articles
approved_articles_df = pd.DataFrame(approved_articles)
file_path1 = os.path.join(drive_folder, "approved_articles_vneconomy.xlsx")
approved_articles_df.to_excel(file_path1, index=False)
print(f"\nĐã lưu thông tin các bài báo được duyệt vào: {file_path1}")

# Lưu File 2: month_summary
file_path2 = os.path.join(drive_folder, "month_summary_vneconomy.xlsx")
month_summary_df.to_excel(file_path2, index=False)
print(f"Đã lưu bảng tổng hợp theo tháng vào: {file_path2}")

# Lưu File 3: keyword_frequency
file_path3 = os.path.join(drive_folder, "keyword_frequency_vneconomy.xlsx")
keyword_freq_df.to_excel(file_path3, index=False)
print(f"Đã lưu bảng tần suất từ khóa theo tháng vào: {file_path3}")
```

```python
# Import các thư viện cần thiết cho quá trình scraping, xử lý dữ liệu, và xuất file
import os  # Thư viện để xử lý hệ thống tệp và thư mục, hỗ trợ tạo và quản lý thư mục lưu bài báo
import re  # Thư viện để xử lý biểu thức chính quy (regular expressions), dùng để làm sạch dữ liệu như tên file
import time  # Thư viện để thêm độ trễ (delay) trong các yêu cầu HTTP, tránh bị chặn bởi server
import datetime  # Thư viện để xử lý thời gian và ngày tháng, hỗ trợ định dạng và so sánh ngày trong nghiên cứu
import random  # Thư viện để tạo số ngẫu nhiên, có thể sử dụng trong việc quản lý độ trễ ngẫu nhiên nếu cần
import json  # Thư viện để đọc/ghi dữ liệu dạng JSON, hỗ trợ lưu trữ cấu trúc dữ liệu phức tạp cho bài báo
import requests  # Thư viện để thực hiện các yêu cầu HTTP (GET, POST) đến các trang web của VnEconomy
import nltk  # Thư viện Natural Language Toolkit để xử lý ngôn ngữ tự nhiên, như tóm tắt bài báo
import pandas as pd  # Thư viện để xử lý và phân tích dữ liệu dạng bảng (DataFrame) cho phân tích thống kê
from bs4 import BeautifulSoup  # Thư viện để parse HTML/XML, trích xuất dữ liệu từ trang web
from newspaper import Article  # Thư viện để trích xuất nội dung bài báo (tiêu đề, nội dung, tóm tắt)
from selenium import webdriver  # Thư viện để tự động hóa trình duyệt, hỗ trợ scraping động cho các trang phức tạp
from selenium.webdriver.chrome.options import Options  # Cấu hình tùy chọn cho trình duyệt Chrome
from selenium.webdriver.common.by import By  # Hỗ trợ tìm phần tử trong DOM bằng cách chỉ định (ID, class, v.v.)
from selenium.common.exceptions import NoSuchElementException, TimeoutException  # Xử lý lỗi khi scraping
from concurrent.futures import ThreadPoolExecutor, as_completed  # Thư viện để xử lý song song (multithreading)
from requests.adapters import HTTPAdapter  # Tăng cường khả năng kết nối HTTP với retry logic
from datetime import date, timedelta  # Hỗ trợ xử lý ngày tháng cụ thể, như tính toán khoảng thời gian
import sys

if 'google.colab' in sys.modules:
    from google.colab import drive
    drive.mount('/content/drive')
else:
    print("Đang chạy ngoài Google Colab – bỏ qua gắn Google Drive.")

# Bắt đầu phần chú thích chi tiết cho từng phần của code, nhằm đảm bảo tính minh bạch và tái tạo được trong nghiên cứu khoa học

#############################################
# PHẦN 1: XỬ LÝ BÀI BÁO (Scraping và kiểm tra từ khóa)
# Mục đích: Thu thập và phân tích nội dung bài báo từ trang web VnEconomy, kiểm tra từ khóa liên quan đến nghiên cứu
#############################################

def configure_driver():
    """
    Cấu hình trình duyệt Chrome để chạy ở chế độ không hiển thị giao diện (headless), nhằm tối ưu hóa tài nguyên
    và phù hợp với môi trường server hoặc nghiên cứu tự động hóa quy mô lớn trên kinh te do thi.

    Returns:
        webdriver.Chrome: Đối tượng trình duyệt Chrome đã được cấu hình, sử dụng cho Selenium.
    """
    # Khởi tạo tùy chọn cho trình duyệt Chrome
    chrome_options = Options()
    # Thêm tham số để chạy ở chế độ không giao diện (headless), giảm sử dụng tài nguyên
    chrome_options.add_argument("--no-sandbox")  # Cho phép chạy trên môi trường không có sandbox (như server Linux)
    chrome_options.add_argument("--headless")  # Chạy không hiển thị giao diện, phù hợp cho scraping tự động
    chrome_options.add_argument("--disable-dev-shm-usage")  # Tối ưu hóa sử dụng bộ nhớ trên hệ thống Linux
    # Trả về trình duyệt Chrome đã cấu hình để sử dụng trong scraping
    return webdriver.Chrome(options=chrome_options)

def save_to_text_file(title, link, date, text, summary, folder_name="articles"):
    """
    Lưu nội dung bài báo vào file .txt để lưu trữ dữ liệu thô, hỗ trợ tái tạo và kiểm tra trong nghiên cứu khoa học.

    Args:
        title (str): Tiêu đề bài báo, được sử dụng làm tên file và đại diện cho nội dung nghiên cứu.
        link (str): Đường dẫn URL của bài báo, dùng để tham chiếu nguồn gốc từ VnEconomy.
        date (str or datetime.date): Ngày xuất bản bài báo, định dạng theo chuẩn ISO hoặc chuỗi.
        text (str): Nội dung đầy đủ của bài báo, dùng để phân tích từ khóa.
        summary (str): Tóm tắt nội dung bài báo, tạo bởi thư viện Newspaper, hỗ trợ nghiên cứu nhanh.
        folder_name (str, optional): Tên thư mục lưu trữ file, mặc định là "articles". Defaults to "articles".
    """
    # Tạo tên file từ tiêu đề, loại bỏ các ký tự không hợp lệ trong tên file (như /, :, *, ?, ", <, >, |)
    filename = os.path.join(folder_name, re.sub(r'[\/:*?"<>|]', '', title) + '.txt')
    # Tạo thư mục nếu chưa tồn tại, đảm bảo không ghi đè dữ liệu
    os.makedirs(folder_name, exist_ok=True)
    # Mở file ở chế độ ghi (write), sử dụng mã hóa UTF-8 để hỗ trợ tiếng Việt
    with open(filename, 'w', encoding="utf-8") as file:
        # Ghi các thông tin bài báo theo định dạng rõ ràng, dễ đọc, phù hợp với tài liệu nghiên cứu
        file.write(f"Title: {title}\n\n")  # Ghi tiêu đề bài báo
        file.write(f"Article Link: {link}\n\n")  # Ghi URL bài báo từ VnEconomy
        file.write(f"Publish Date: {date}\n\n")  # Ghi ngày xuất bản
        file.write(f"Article:\n{text}\n\n")  # Ghi nội dung bài báo
        file.write(f"Summary: {summary}\n\n")  # Ghi tóm tắt bài báo
    # In thông báo xác nhận đã lưu file, hỗ trợ theo dõi tiến trình nghiên cứu
    print(f"Đã lưu bài báo: {filename}")

def is_within_date_range(date_obj):
    """
    Kiểm tra xem ngày của bài báo có nằm trong khoảng thời gian từ 01/01/2014 trở đi hay không,
    nhằm đảm bảo dữ liệu chỉ thu thập trong phạm vi nghiên cứu đã định từ kinh te do thi.

    Args:
        date_obj (datetime.date or str or None): Đối tượng ngày hoặc chuỗi đại diện cho ngày xuất bản từ bài báo.

    Returns:
        bool: True nếu ngày nằm trong khoảng từ 01/01/2014, False nếu không hoặc nếu ngày không hợp lệ.
    """
    # Kiểm tra nếu ngày là "Unknown" hoặc None, trả về False để loại bỏ bài báo không có ngày hợp lệ
    if date_obj == "Unknown" or date_obj is None:
        return False
    # Định nghĩa ngày bắt đầu của nghiên cứu (01/01/2014) để phù hợp với phạm vi nghiên cứu
    start_date = datetime.date(2014, 1, 1)
    # So sánh ngày bài báo với ngày bắt đầu, trả về True nếu bài báo đủ điều kiện
    return start_date <= date_obj

def get_matching_keywords(text):
    """
    Xác định các từ khóa trong văn bản dựa trên 4 danh mục (Hoa Kỳ, Trung Quốc, chủ đề, và căng thẳng),
    đồng thời loại bỏ bài viết nếu chứa từ khóa không liên quan.

    Args:
        text (str): Văn bản của bài báo (tiêu đề hoặc nội dung) để kiểm tra từ khóa.

    Returns:
        list: Danh sách các từ khóa khớp nếu có ít nhất một từ từ mỗi danh mục và không chứa từ không liên quan,
              ngược lại trả về danh sách rỗng.
    """
    # Danh sách từ khóa không liên quan
    unrelated_words = [
        # Phim/Điện ảnh
        "phim", "diễn viên", "đạo diễn","showbiz",
        "bóng đá", "cầu thủ", "trận đấu","tình yêu", "mối quan hệ", "hẹn hò",
        "người yêu", "chia tay", "hôn nhân",
        "game thủ",  "âm nhạc", "ca sĩ", "bài hát","ẩm thực", "món ăn",
        "nghỉ dưỡng","trẻ em", "cha mẹ","lễ cưới","đám tang"
    ]

    # Danh sách từ khóa chính
    us_words = ["hoa kỳ", "mỹ", "washington", "nhà trắng"]
    china_words = ["trung quốc", "bắc kinh"]
    topic_words = [
        "tăng thuế", "áp thuế", "tiền tệ", "wto", "tổ chức thương mại thế giới",
        "ngân khố", "nhà máy", "thặng dư", "nợ", "thuế quan",
        "rào cản", "khoản vay", "thâm hụt", "nhân dân tệ",
        "trợ cấp", "đô la", "xuất khẩu", "sao chép bất hợp pháp", "hack", "Huawei", "mạng", "đại sứ quán", "Hồng Kông",
        "trộm cắp", "nhập khẩu", "viễn thông", "vi phạm bản quyền", "lãnh sự quán", "nhập cư", "visa", "thẻ xanh",
        "khí hậu", "năng lượng", "bán phá giá", "bản quyền", "bằng sáng chế", "giam giữ", "biểu tình", "tự trị", "bắt giữ", "bầu cử", "đàn áp", "độc tài", "biểu tình", "phong trào",
        "quyền sở hữu trí tuệ", "tấn công mạng", "bảo hộ thương mại", "tỷ giá hối đoái", "môi trường", "tôn giáo", "Tây Tạng", "Đạt Lai", "quân sự", "an ninh", "quốc phòng", "khủng bố",
        "Lầu Năm Góc", "rocket", "tàu sân bay", "tàu chiến", "đánh chìm", "Triều Tiên", "tuần tra", "máy bay phản lực", "khẩn cấp", "kho vũ khí", "tình báo", "tàu khu trục", "khiêu khích",
        "hạt nhân", "trinh sát", "vũ khí", "tàu", "giám sát", "hạm đội", "sự cố", "liên minh", "đối thủ", "Liên Hợp Quốc", "bảo mật", "gián điệp", "radar", "tàu ngầm",
        "chuyến bay", "NATO", "phi công", "hiệp ước", "Thái Bình Dương", "lãnh thổ", "vệ tinh", "hàng hải", "quân đội", "Đài Loan", "bom",
        "Đài Bắc", "bạo lực", "Biển Đông", "đảo", "an ninh thông tin", "ô nhiễm", "tài chính", "giao dịch", "chuỗi cung ứng",
        "bảo hộ thương mại", "chủ nghĩa bảo hộ", "máy bay", "đạn đạo", "xây dựng lực lượng", "lưu vong", "đầu đạn", "tên lửa", "hải quân", "triển khai",
        "dược phẩm", "nông nghiệp", "công nghệ", "đơn phương", "đại dịch", "covid", "coronavirus", "viêm phổi", "song phương", "bán dẫn",
        "nhân quyền", "thương chiến", "kiểm soát xuất khẩu", "địa chính trị", "hàng rào thuế quan", "hạn chế xuất khẩu", "cuộc chiến thương mại"
    ]
    tension_words = [
        "đe dọa", "chiến tranh", "thù địch", "gia tăng căng thẳng", "leo thang căng thẳng", "giảm căng thẳng", "vi phạm", "kiện tụng", "trừng phạt",
        "khủng bố", "phá hủy", "tổn hại", "cấm vận", "lệnh cấm", "lệnh hạn chế", "giả mạo", "cáo buộc", "thao túng", "leo thang",
        "âm mưu", "đàn áp", "lạm dụng", "kìm hãm", "phản đối", "va chạm", "ngăn chặn", "khiêu khích", "mâu thuẫn", "chia rẽ",
        "đối thủ cạnh tranh", "đình trệ", "xung đột", "cảnh báo", "trộm cắp", "làm suy yếu", "áp đặt", "tranh cãi", "đáp trả", "nổ ra",
        "hung hăng", "kẻ thù", "phòng thủ", "chiến đấu", "xấu đi", "tranh chấp", "leo thang", "trả đũa", "phân cực", "chống lại", "đối trọng", "danh sách đen",
        "tồi tệ hơn", "đối kháng", "đối đầu", "giám sát", "cản trở", "chỉ trích", "nhượng bộ", "chi phối", "xung đột", "thống trị"
    ]

    text_lower = text.lower()

    # Kiểm tra từ khóa không liên quan trước
    for word in unrelated_words:
        if word in text_lower:
            return []  # Nếu tìm thấy từ không liên quan, trả về rỗng để loại bỏ bài viết

    # Kiểm tra từ khóa chính
    found_us = [word for word in us_words if word in text_lower]
    found_china = [word for word in china_words if word in text_lower]
    found_topic = [word for word in topic_words if word in text_lower]
    found_tension = [word for word in tension_words if word in text_lower]

    # Chỉ trả về kết quả nếu có ít nhất một từ từ mỗi danh mục
    if found_us and found_china and found_topic and found_tension:
        return list(set(found_us + found_china + found_topic + found_tension))
    return []

#############################################
# PHẦN 2: XỬ LÝ SITEMAP VÀ TRUY CẬP BÀI BÁO
# Mục đích: Thu thập danh sách URL bài báo từ sitemap của kinh te do thi và xử lý từng bài báo để trích xuất dữ liệu
#############################################

def get_sitemap_urls(main_sitemap_url="https://kinhtedothi.vn/sitemap.xml"):
    """
    Lấy danh sách URL của các sitemap từ trang web Kinh Tế Đô Thị, lọc chỉ lấy các sitemap
    từ https://kinhtedothi.vn/sitemap/news-2023-12.xml đến https://kinhtedothi.vn/sitemap/news-2014-1.xml,
    phù hợp với phạm vi thời gian nghiên cứu.

    Args:
        main_sitemap_url (str, optional): URL của sitemap chính của Kinh Tế Đô Thị, mặc định là "https://kinhtedothi.vn/sitemap.xml". Defaults to "https://kinhtedothi.vn/sitemap.xml".

    Returns:
        list: Danh sách các URL sitemap được lọc theo năm và tháng trong phạm vi nghiên cứu.
    """
    # Thực hiện yêu cầu HTTP GET để lấy nội dung sitemap chính
    r = requests.get(main_sitemap_url)
    # Sử dụng BeautifulSoup để parse nội dung XML của sitemap
    soup = BeautifulSoup(r.content, "xml")  # Sử dụng parser XML để đảm bảo tính chính xác
    # Tìm tất cả các thẻ <sitemap> trong sitemap chính
    sitemap_tags = soup.find_all("sitemap")
    # Trích xuất URL từ các thẻ <loc> trong sitemap
    sitemap_urls = [tag.find("loc").text for tag in sitemap_tags]

    # Định nghĩa URL bắt đầu và kết thúc theo phạm vi nghiên cứu
    start_url = "https://kinhtedothi.vn/sitemaps/sitemap-article-2024-12-30.xml"  # Sitemap mới nhất (tháng 2/2025)
    end_url = "https://kinhtedothi.vn/sitemaps/sitemap-article-2024-12-20.xml"  # Sitemap đầu tiên (tháng 1/2014)
    # Kiểm tra xem các URL bắt đầu và kết thúc có tồn tại trong danh sách không
    if start_url in sitemap_urls and end_url in sitemap_urls:
        # Tìm chỉ số của URL bắt đầu và kết thúc trong danh sách
        start_index = sitemap_urls.index(start_url)
        end_index = sitemap_urls.index(end_url)
        # Giả sử danh sách được sắp xếp từ mới nhất đến cũ nhất (2025 đến 2014)
        # Nếu start_index < end_index, lấy slice từ start_index đến end_index + 1, ngược lại lấy slice đảo
        if start_index <= end_index:
            filtered = sitemap_urls[start_index:end_index + 1]  # Lấy từ 2025-2 đến 2014-1
        else:
            filtered = sitemap_urls[end_index:start_index + 1]  # Lấy theo thứ tự ngược lại nếu cần
        return filtered  # Trả về danh sách sitemap đã lọc
    else:
        # Nếu không tìm thấy URL bắt đầu hoặc kết thúc, trả về danh sách rỗng
        print(f"Không tìm thấy {start_url} hoặc {end_url} trong sitemap chính.")
        return []

def get_article_urls_from_sitemap(sitemap_url):
    """
    Lấy danh sách URL bài báo từ một sitemap cụ thể của VnEconomy, bao gồm cả ngày chỉnh sửa cuối (lastmod),
    để đảm bảo thu thập dữ liệu bài báo chính xác và có thời gian cập nhật.

    Args:
        sitemap_url (str): URL của file sitemap cần parse từ VnEconomy.

    Returns:
        list: Danh sách các dictionary chứa thông tin URL và ngày chỉnh sửa cuối của bài báo.
    """
    # Thực hiện yêu cầu HTTP GET để lấy nội dung sitemap
    r = requests.get(sitemap_url)
    # Sử dụng BeautifulSoup để parse nội dung XML của sitemap
    soup = BeautifulSoup(r.content, "xml")  # Sử dụng parser XML để đảm bảo tính chính xác
    # Tìm tất cả các thẻ <url> trong sitemap
    url_tags = soup.find_all("url")
    # Khởi tạo danh sách để lưu thông tin bài báo
    articles = []
    # Lặp qua từng thẻ URL để trích xuất thông tin
    for tag in url_tags:
        # Lấy URL từ thẻ <loc>
        loc = tag.find("loc").text
        # Lấy ngày chỉnh sửa cuối (lastmod) nếu có
        lastmod = tag.find("lastmod").text if tag.find("lastmod") else None
        # Thêm thông tin bài báo vào danh sách
        articles.append({"url": loc, "lastmod": lastmod})
    # Trả về danh sách các bài báo từ sitemap
    return articles

def crawl_articles_from_sitemaps():
    """
    Quét qua tất cả các sitemap đã lọc từ kinh te do thi để thu thập thông tin bài báo,
    lọc chỉ lấy các bài báo có ngày từ 01/01/2014 trở đi, phù hợp với phạm vi nghiên cứu.

    Returns:
        list: Danh sách các dictionary chứa thông tin URL và ngày của bài báo thỏa mãn điều kiện.
    """
    # Lấy danh sách URL sitemap từ hàm get_sitemap_urls
    sitemap_urls = get_sitemap_urls()
    # In số lượng sitemap tìm thấy để theo dõi tiến trình nghiên cứu
    print(f"Đã tìm thấy {len(sitemap_urls)} sitemap từ kinh te do thi.")
    # Khởi tạo danh sách để lưu thông tin bài báo
    all_article_infos = []
    # Lặp qua từng sitemap để xử lý
    for sitemap_url in sitemap_urls:
        print(f"Đang xử lý sitemap: {sitemap_url}")
        # Lấy danh sách bài báo từ sitemap hiện tại
        articles = get_article_urls_from_sitemap(sitemap_url)
        # Xử lý từng bài báo trong sitemap
        for art in articles:
            # Chuyển đổi ngày lastmod thành đối tượng date nếu có
            if art["lastmod"]:
                try:
                    # Chuyển đổi chuỗi ISO 8601 thành đối tượng date, xử lý trường hợp kết thúc bằng 'Z'
                    art_date = datetime.datetime.fromisoformat(art["lastmod"].replace("Z", "+00:00")).date()
                except Exception as e:
                    # Nếu có lỗi trong quá trình chuyển đổi, gán date là None
                    art_date = None
            else:
                # Nếu không có lastmod, gán date là None
                art_date = None
            # Thêm ngày vào thông tin bài báo
            art["date"] = art_date
            # Kiểm tra và thêm bài báo vào danh sách nếu ngày >= 01/01/2014
            if art_date and art_date >= datetime.date(2014, 1, 1):
                all_article_infos.append(art)
    # In tổng số bài báo thu thập được để theo dõi
    print(f"Tổng số bài báo (theo sitemap) từ 01/01/2014: {len(all_article_infos)}")
    # Trả về danh sách bài báo thỏa mãn điều kiện nghiên cứu
    return all_article_infos

def process_article(url, sitemap_date, processed_urls, max_retries=3):
    """
    Xử lý một bài báo cụ thể từ kinh te do thi: trích xuất nội dung, kiểm tra từ khóa, và lưu nếu đáp ứng tiêu chí nghiên cứu.

    Args:
        url (str): URL của bài báo cần xử lý từ kinh te do thi.
        sitemap_date (datetime.date or None): Ngày chỉnh sửa cuối từ sitemap (lastmod).
        processed_urls (set): Tập hợp URL đã xử lý để tránh lặp lại trong nghiên cứu.
        max_retries (int, optional): Số lần thử lại khi gặp lỗi kết nối, mặc định là 3. Defaults to 3.

    Returns:
        dict or None: Thông tin bài báo nếu đáp ứng tiêu chí từ khóa, hoặc None nếu không thành công.
    """
    # Kiểm tra xem URL đã được xử lý chưa để tránh trùng lặp dữ liệu trong nghiên cứu
    if url in processed_urls:
        return None
    # Thêm URL vào tập hợp đã xử lý
    processed_urls.add(url)

    # Thử tải nội dung bài báo với cơ chế retry để xử lý lỗi kết nối, đảm bảo tính ổn định trong scraping
    for attempt in range(max_retries):
        try:
            # Thêm độ trễ 0.5 giây để tránh bị chặn bởi server của kinh te do thi
            time.sleep(0.5)
            # Thực hiện yêu cầu GET với timeout 30 giây để tránh treo chương trình
            response = requests.get(url, timeout=30)
            # Lấy nội dung HTML của bài báo
            html = response.text
            break  # Thoát vòng lặp nếu thành công
        except requests.exceptions.ReadTimeout as e:
            # Xử lý lỗi timeout, in thông báo và thử lại
            print(f"Timeout khi truy cập {url}. Thử lại ({attempt+1}/{max_retries})...")
            if attempt == max_retries - 1:
                # Nếu hết số lần thử, trả về None để bỏ qua bài báo
                return None
        except Exception as e:
            # Xử lý các lỗi khác (ví dụ: không kết nối được), trả về None
            return None

    try:
        # Không thu thập thông tin tác giả nữa để đơn giản hóa và tập trung vào nội dung nghiên cứu
        # Sử dụng thư viện Newspaper để trích xuất nội dung bài báo từ HTML
        article = Article(url, language="vi")  # Thiết lập ngôn ngữ tiếng Việt
        article.set_html(html)  # Gán nội dung HTML đã tải
        article.parse()  # Phân tích và trích xuất nội dung, tiêu đề
        try:
            # Tạo tóm tắt bài báo bằng NLP để hỗ trợ phân tích nhanh
            article.nlp()
        except LookupError as e:
            # Nếu chưa tải gói punkt của NLTK, tải xuống để xử lý tóm tắt
            nltk.download("punkt")
            article.nlp()

        # Sử dụng ngày từ sitemap (lastmod) làm ngày xuất bản, đảm bảo tính nhất quán với dữ liệu nguồn từ VnEconomy
        date_obj = sitemap_date if sitemap_date else "Unknown"
        # Kiểm tra xem ngày có nằm trong phạm vi nghiên cứu không
        if not is_within_date_range(date_obj):
            return None

        # Lấy nội dung và tóm tắt từ bài báo
        text = article.text  # Nội dung đầy đủ để phân tích từ khóa
        summary = article.summary  # Tóm tắt nội dung để nghiên cứu nhanh
        title = article.title if article.title else ""  # Tiêu đề, mặc định rỗng nếu không có

        # Kiểm tra từ khóa trong tiêu đề hoặc nội dung để xác định bài báo có liên quan đến nghiên cứu không
        matches = get_matching_keywords(title) or get_matching_keywords(text)
        if matches:
            # Nếu bài báo chứa đủ từ khóa, lưu vào file .txt để lưu trữ dữ liệu thô
            save_to_text_file(title, url, date_obj, text, summary)
            # In thông báo để theo dõi tiến trình, bao gồm từ khóa khớp
            print(f"Đã duyệt bài: \"{title}\" (Ngày: {date_obj}) | Từ khóa: {matches}")
            # Trả về thông tin bài báo để tổng hợp sau này
            return {
                "title": title,  # Tiêu đề bài báo
                "url": url,  # Đường dẫn bài báo từ VnEconomy
                "date": str(date_obj),  # Ngày xuất bản dưới dạng chuỗi
                "text": text,  # Nội dung đầy đủ
                "summary": summary,  # Tóm tắt
                "matching": matches  # Danh sách từ khóa khớp
            }
        else:
            # Nếu không có từ khóa phù hợp, trả về None để bỏ qua bài báo
            return None

    except Exception as e:
        # Xử lý lỗi trong quá trình trích xuất nội dung, trả về None và in lỗi để theo dõi
        print(f"Lỗi khi xử lý bài báo {url}: {e}")
        return None

def process_article_wrapper(article_info, processed_urls, idx):
    """
    Hàm bao bọc để xử lý bài báo, in tiến trình và xử lý các trường hợp ngoại lệ,
    đảm bảo tính theo dõi trong nghiên cứu quy mô lớn từ VnEconomy.

    Args:
        article_info (dict): Thông tin bài báo từ sitemap (URL và ngày).
        processed_urls (set): Tập hợp URL đã xử lý.
        idx (int): Chỉ số thứ tự của bài báo để theo dõi tiến trình nghiên cứu.

    Returns:
        dict or None: Kết quả xử lý bài báo hoặc thông tin URL và ngày nếu không đạt tiêu chí từ khóa.
    """
    # In thông báo để theo dõi tiến trình xử lý từng bài báo từ kinh te do thi
    print(f"Đang xử lý bài báo {idx}: {article_info['url']}")
    # Gọi hàm xử lý bài báo với thông tin ngày từ sitemap
    result = process_article(article_info["url"], article_info["date"], processed_urls)
    # Nếu bài báo được duyệt, in thông tin ngày và từ khóa
    if result and "date" in result:
        print(f"Bài báo {idx} có ngày {result['date']} và chứa từ khóa: {result.get('matching', [])}")
    # Nếu bài báo không đạt tiêu chí nhưng có ngày, trả về thông tin URL và ngày
    if result is None and article_info["date"]:
        result = {"url": article_info["url"], "date": str(article_info["date"])}
    # Trả về kết quả xử lý
    return result

def crawl_and_process_articles():
    """
    Thực hiện quét và xử lý tất cả bài báo từ các sitemap của kinh te do thi,
    sử dụng xử lý song song để tăng hiệu suất và thu thập dữ liệu nghiên cứu.

    Returns:
        tuple: Danh sách bài báo được duyệt (approved_articles) và danh sách tất cả bài báo từ sitemap.
    """
    # Lấy thông tin bài báo từ các sitemap của VnEconomy
    article_infos = crawl_articles_from_sitemaps()
    # Khởi tạo tập hợp để lưu URL đã xử lý, tránh trùng lặp
    processed_urls = set()
    # Khởi tạo danh sách để lưu bài báo được duyệt
    approved_articles = []

    # Sử dụng ThreadPoolExecutor để xử lý song song, tối ưu hóa thời gian thực thi trên kinh te do thi
    with ThreadPoolExecutor(max_workers=20) as executor:  # Sử dụng 20 worker để cân bằng hiệu suất và tải server
        # Tạo danh sách các tác vụ (futures) cho từng bài báo
        futures = {executor.submit(process_article_wrapper, info, processed_urls, idx): info
                   for idx, info in enumerate(article_infos, start=1)}  # Số thứ tự bắt đầu từ 1
        # Theo dõi và thu thập kết quả khi các tác vụ hoàn thành
        for future in as_completed(futures):
            # Lấy kết quả từ mỗi tác vụ
            result = future.result()
            # Nếu bài báo được duyệt (có tiêu đề), thêm vào danh sách approved_articles
            if result and "title" in result:
                approved_articles.append(result)
    # In số lượng bài báo được duyệt để theo dõi kết quả nghiên cứu
    print(f"Đã duyệt {len(approved_articles)} bài báo (đáp ứng tiêu chí từ khóa) từ sitemap của kinh te do thi.")
    # Trả về danh sách bài báo được duyệt và danh sách tất cả bài báo từ sitemap
    return approved_articles, article_infos

#############################################
# PHẦN 3: TẠO BẢNG DATAFRAME THEO THÁNG
# Mục đích: Tổng hợp dữ liệu bài báo và từ khóa từ VnEconomy theo từng tháng để phân tích thống kê
#############################################

def get_month(date_obj):
    """
    Chuyển đổi đối tượng ngày thành chuỗi định dạng "Năm-MM" (ví dụ: "2021-03"),
    phù hợp với yêu cầu phân tích dữ liệu theo tháng trong nghiên cứu từ kinh te do thi.

    Args:
        date_obj (datetime.date): Đối tượng ngày cần chuyển đổi.

    Returns:
        str: Chuỗi định dạng "YYYY-MM" đại diện cho tháng, dùng để nhóm dữ liệu.
    """
    # Sử dụng định dạng "YYYY-MM" với tháng có 2 chữ số (padding 0) để đảm bảo thống nhất
    return f"{date_obj.year}-{date_obj.month:02d}"

def create_month_summary(approved_articles, all_article_infos):
    """
    Tạo bảng tổng hợp dữ liệu bài báo từ VnEconomy theo tháng, bao gồm tổng số bài báo, số bài báo được duyệt,
    và tỷ lệ bài báo được duyệt so với tổng số, phục vụ phân tích xu hướng theo thời gian.

    Args:
        approved_articles (list): Danh sách các bài báo được duyệt theo tiêu chí từ khóa từ kinh te do thi.
        all_article_infos (list): Danh sách tất cả bài báo thu thập từ sitemap của kinh te do thi.

    Returns:
        pd.DataFrame: Bảng dữ liệu tổng hợp theo tháng với các cột: 'month', 'total_count', 'approved_count', 'approved_ratio'.
    """
    # Chuẩn bị dữ liệu cho tất cả bài báo từ sitemap
    all_data = []
    for info in all_article_infos:
        if info["date"]:  # Chỉ lấy bài báo có ngày hợp lệ
            all_data.append({"date": info["date"]})
    # Tạo DataFrame từ danh sách dữ liệu
    all_df = pd.DataFrame(all_data)
    if all_df.empty:  # Kiểm tra nếu không có dữ liệu
        return pd.DataFrame()
    # Chuyển đổi cột 'date' thành định dạng datetime để xử lý
    all_df['date'] = pd.to_datetime(all_df['date'])
    # Thêm cột 'month' theo định dạng "YYYY-MM"
    all_df['month'] = all_df['date'].dt.to_period("M").astype(str)

    # Chuẩn bị dữ liệu cho bài báo được duyệt
    approved_data = []
    for art in approved_articles:
        try:
            # Chuyển đổi ngày từ chuỗi "%Y-%m-%d" thành đối tượng datetime
            d = datetime.datetime.strptime(art["date"], "%Y-%m-%d")
            approved_data.append({"date": d})
        except Exception as e:
            # Bỏ qua bài báo nếu ngày không hợp lệ, ghi log lỗi nếu cần cho nghiên cứu
            print(f"Lỗi chuyển đổi ngày cho bài báo {art.get('url', 'Unknown')} từ kinh te do thi: {e}")
            continue
    # Tạo DataFrame từ danh sách bài báo được duyệt
    approved_df = pd.DataFrame(approved_data)
    if not approved_df.empty:
        # Thêm cột 'month' cho bài báo được duyệt
        approved_df['month'] = approved_df['date'].dt.to_period("M").astype(str)

    # Tính tổng số bài báo theo tháng
    total_summary = all_df.groupby("month").size().reset_index(name="total_count")
    # Tính số bài báo được duyệt theo tháng, nếu không có thì tạo DataFrame rỗng với cột phù hợp
    approved_summary = approved_df.groupby("month").size().reset_index(name="approved_count") if not approved_df.empty else pd.DataFrame(columns=["month", "approved_count"])
    # Gộp dữ liệu tổng và dữ liệu duyệt, điền giá trị thiếu bằng 0
    summary = pd.merge(total_summary, approved_summary, on="month", how="left")
    # Đảm bảo cột approved_count là số nguyên, điền giá trị 0 nếu thiếu
    summary["approved_count"] = summary["approved_count"].fillna(0).astype(int)
    # Tính tỷ lệ bài báo được duyệt so với tổng số bài báo
    summary["approved_ratio"] = summary["approved_count"] / summary["total_count"]
    # Trả về bảng tổng hợp theo tháng
    return summary

def create_keyword_frequency_df(approved_articles, keyword_list):
    """
    Tạo bảng tần suất xuất hiện của các từ khóa từ kinh te do thi theo tháng, tính toán tỷ lệ xuất hiện dựa trên số bài báo trong tháng,
    phục vụ phân tích tần suất và xu hướng từ khóa trong nghiên cứu.

    Args:
        approved_articles (list): Danh sách các bài báo được duyệt theo tiêu chí từ khóa từ kinh te do thi.
        keyword_list (list): Danh sách các từ khóa cần phân tích tần suất, dựa trên các chủ đề nghiên cứu.

    Returns:
        pd.DataFrame: Bảng dữ liệu tần suất từ khóa theo tháng với cột 'month' và các cột từ khóa.
    """
    # Chuẩn bị dữ liệu từ các bài báo được duyệt
    data = []
    for art in approved_articles:
        try:
            # Chuyển đổi ngày từ chuỗi "%Y-%m-%d" thành đối tượng datetime
            d = datetime.datetime.strptime(art["date"], "%Y-%m-%d")
            # Thêm thông tin ngày và danh sách từ khóa khớp vào danh sách dữ liệu
            data.append({"date": d, "matching": art.get("matching", [])})
        except Exception as e:
            # Bỏ qua bài báo nếu ngày không hợp lệ, ghi log lỗi nếu cần
            print(f"Lỗi chuyển đổi ngày cho bài báo {art.get('url', 'Unknown')} từ kinh te do thi: {e}")
            continue
    # Tạo DataFrame từ danh sách dữ liệu
    df = pd.DataFrame(data)
    if df.empty:  # Kiểm tra nếu không có dữ liệu
        return pd.DataFrame()
    # Thêm cột 'month' theo định dạng "YYYY-MM"
    df['month'] = df['date'].dt.to_period("M").astype(str)
    # Tạo cột nhị phân (0 hoặc 1) cho từng từ khóa, 1 nếu từ khóa xuất hiện trong danh sách khớp
    for kw in keyword_list:
        df[kw] = df['matching'].apply(lambda lst: 1 if kw in lst else 0)
    # Tính tổng số lần xuất hiện của từng từ khóa theo tháng
    freq = df.groupby('month').agg({kw: 'sum' for kw in keyword_list})
    # Tính tổng số bài báo theo tháng
    total = df.groupby('month').size()
    # Tính tỷ lệ xuất hiện của từng từ khóa (số lần xuất hiện / tổng số bài báo trong tháng)
    freq = freq.divide(total, axis=0)
    # Đặt lại chỉ số và chuyển thành DataFrame với cột 'month'
    freq.reset_index(inplace=True)
    # Trả về bảng tần suất từ khóa theo tháng
    return freq

#############################################
# PHẦN CHÍNH (MAIN): THỰC THI CHƯƠNG TRÌNH
# Mục đích: Kết hợp các bước scraping, xử lý, và tổng hợp dữ liệu từ kinh te do thi để tạo ra kết quả nghiên cứu
#############################################
if __name__ == "__main__":
    # Bước 1: Thực hiện quét và xử lý bài báo từ các sitemap của kinh te do thi
    # Mục tiêu: Thu thập dữ liệu bài báo liên quan đến nghiên cứu về căng thẳng thương mại
    approved_articles, all_article_infos = crawl_and_process_articles()

    # Bước 2: Lưu danh sách các bài báo được duyệt vào file JSON để lưu trữ dữ liệu thô,
    # hỗ trợ tái tạo và kiểm tra trong nghiên cứu
    with open("approved_articles_kinhtedothi.json", "w", encoding="utf-8") as f:
        # Sử dụng JSON để lưu trữ dữ liệu với định dạng dễ đọc, hỗ trợ tiếng Việt
        json.dump(approved_articles, f, ensure_ascii=False, indent=4)
    # In thông báo xác nhận đã lưu file
    print("\nĐã lưu kết quả vào file 'approved_articles_kinhtedothi.json'.")

    # Bước 3: Tạo bảng tổng hợp số bài báo theo tháng (tổng số và số được duyệt)
    # Mục tiêu: Phân tích xu hướng bài báo theo thời gian theo tháng từ kinhtedothi
    month_summary_df = create_month_summary(approved_articles, all_article_infos)
    # In bảng tổng hợp để theo dõi kết quả nghiên cứu
    print("\nBảng tổng hợp theo tháng (Approved / Total):")
    print(month_summary_df)

    # Bước 4: Tạo bảng tần suất từ khóa theo tháng để phân tích xu hướng từ khóa
    # Danh sách từ khóa được định nghĩa dựa trên các chủ đề nghiên cứu về thương mại và căng thẳng
    keyword_list = [ "tăng thuế", "áp thuế", "tiền tệ", "wto", "tổ chức thương mại thế giới",
        "ngân khố", "nhà máy", "thặng dư", "nợ", "thuế quan",
        "rào cản", "khoản vay", "thâm hụt", "nhân dân tệ",
        "trợ cấp", "đô la", "xuất khẩu", "sao chép bất hợp pháp", "hack", "Huawei", "mạng", "đại sứ quán", "Hồng Kông",
        "trộm cắp", "nhập khẩu", "viễn thông", "vi phạm bản quyền", "lãnh sự quán", "nhập cư", "visa", "thẻ xanh",
        "khí hậu", "năng lượng", "bán phá giá", "bản quyền", "bằng sáng chế", "giam giữ", "biểu tình", "tự trị", "bắt giữ", "bầu cử", "đàn áp", "độc tài", "biểu tình", "phong trào",
        "quyền sở hữu trí tuệ", "tấn công mạng", "bảo hộ thương mại", "tỷ giá hối đoái", "môi trường", "tôn giáo", "Tây Tạng", "Đạt Lai", "quân sự", "an ninh", "quốc phòng", "khủng bố",
        "Lầu Năm Góc", "rocket", "tàu sân bay", "tàu chiến", "đánh chìm", "Triều Tiên", "tuần tra", "máy bay phản lực", "khẩn cấp", "kho vũ khí", "tình báo", "tàu khu trục", "khiêu khích",
        "hạt nhân", "trinh sát", "vũ khí", "tàu", "giám sát", "hạm đội", "sự cố", "liên minh", "đối thủ", "Liên Hợp Quốc", "bảo mật", "gián điệp", "radar", "tàu ngầm",
        "chuyến bay", "NATO", "phi công", "hiệp ước", "Thái Bình Dương", "lãnh thổ", "vệ tinh", "hàng hải", "quân đội", "Đài Loan", "bom",
        "Đài Bắc", "bạo lực", "Biển Đông", "đảo", "an ninh thông tin", "ô nhiễm", "tài chính", "giao dịch", "chuỗi cung ứng",
        "bảo hộ thương mại", "chủ nghĩa bảo hộ", "máy bay", "đạn đạo", "xây dựng lực lượng", "lưu vong", "đầu đạn", "tên lửa", "hải quân", "triển khai",
        "dược phẩm", "nông nghiệp", "công nghệ", "đơn phương", "đại dịch", "covid", "coronavirus", "viêm phổi", "song phương", "bán dẫn",
        "nhân quyền", "thương chiến", "kiểm soát xuất khẩu", "địa chính trị", "hàng rào thuế quan", "hạn chế xuất khẩu", "cuộc chiến thương mại",
        "đe dọa", "chiến tranh", "thù địch", "gia tăng căng thẳng", "leo thang căng thẳng", "giảm căng thẳng", "vi phạm", "kiện tụng", "trừng phạt",
        "khủng bố", "phá hủy", "tổn hại", "cấm vận", "lệnh cấm", "lệnh hạn chế", "giả mạo", "cáo buộc", "thao túng", "leo thang",
        "âm mưu", "đàn áp", "lạm dụng", "kìm hãm", "phản đối", "va chạm", "ngăn chặn", "khiêu khích", "mâu thuẫn", "chia rẽ",
        "đối thủ cạnh tranh", "đình trệ", "xung đột", "cảnh báo", "trộm cắp", "làm suy yếu", "áp đặt", "tranh cãi", "đáp trả", "nổ ra",
        "hung hăng", "kẻ thù", "phòng thủ", "chiến đấu", "xấu đi", "tranh chấp", "leo thang", "trả đũa", "phân cực", "chống lại", "đối trọng", "danh sách đen",
        "tồi tệ hơn", "đối kháng", "đối đầu", "giám sát", "cản trở", "chỉ trích", "nhượng bộ", "chi phối", "xung đột", "thống trị"
    ]
    # Tạo bảng tần suất từ khóa theo tháng
    keyword_freq_df = create_keyword_frequency_df(approved_articles, keyword_list)
    # In bảng tần suất để theo dõi kết quả nghiên cứu
    print("\nBảng tần suất từ khóa theo tháng:")
    print(keyword_freq_df)

    #############################################
    # Bước 5: Lưu dữ liệu ra 3 file Excel tự động và tải về (nếu chạy trên Google Colab)
# Kết nối Google Drive
drive.mount('/content/drive')

# Định nghĩa đường dẫn thư mục trong Google Drive
drive_folder = "/content/drive/My Drive/BAIBAO_Reports/"

# Kiểm tra và tạo thư mục nếu chưa tồn tại
if not os.path.exists(drive_folder):
    os.makedirs(drive_folder)

# Lưu File 1: approved_articles
approved_articles_df = pd.DataFrame(approved_articles)
file_path1 = os.path.join(drive_folder, "approved_articles_kinhtedothi.xlsx")
approved_articles_df.to_excel(file_path1, index=False)
print(f"\nĐã lưu thông tin các bài báo được duyệt vào: {file_path1}")

# Lưu File 2: month_summary
file_path2 = os.path.join(drive_folder, "month_summary_kinhtedothi.xlsx")
month_summary_df.to_excel(file_path2, index=False)
print(f"Đã lưu bảng tổng hợp theo tháng vào: {file_path2}")

# Lưu File 3: keyword_frequency
file_path3 = os.path.join(drive_folder, "keyword_frequency_kinhtedothi.xlsx")
keyword_freq_df.to_excel(file_path3, index=False)
print(f"Đã lưu bảng tần suất từ khóa theo tháng vào: {file_path3}")
```

```python
# Import các thư viện cần thiết cho quá trình scraping, xử lý dữ liệu, và xuất file
import os  # Thư viện để xử lý hệ thống tệp và thư mục, hỗ trợ tạo và quản lý thư mục lưu bài báo
import re  # Thư viện để xử lý biểu thức chính quy (regular expressions), dùng để làm sạch dữ liệu như tên file
import time  # Thư viện để thêm độ trễ (delay) trong các yêu cầu HTTP, tránh bị chặn bởi server
import datetime  # Thư viện để xử lý thời gian và ngày tháng, hỗ trợ định dạng và so sánh ngày trong nghiên cứu
import random  # Thư viện để tạo số ngẫu nhiên, có thể sử dụng trong việc quản lý độ trễ ngẫu nhiên nếu cần
import json  # Thư viện để đọc/ghi dữ liệu dạng JSON, hỗ trợ lưu trữ cấu trúc dữ liệu phức tạp cho bài báo
import requests  # Thư viện để thực hiện các yêu cầu HTTP (GET, POST) đến các trang web của VnEconomy
import nltk  # Thư viện Natural Language Toolkit để xử lý ngôn ngữ tự nhiên, như tóm tắt bài báo
import pandas as pd  # Thư viện để xử lý và phân tích dữ liệu dạng bảng (DataFrame) cho phân tích thống kê
from bs4 import BeautifulSoup  # Thư viện để parse HTML/XML, trích xuất dữ liệu từ trang web
from newspaper import Article  # Thư viện để trích xuất nội dung bài báo (tiêu đề, nội dung, tóm tắt)
from selenium import webdriver  # Thư viện để tự động hóa trình duyệt, hỗ trợ scraping động cho các trang phức tạp
from selenium.webdriver.chrome.options import Options  # Cấu hình tùy chọn cho trình duyệt Chrome
from selenium.webdriver.common.by import By  # Hỗ trợ tìm phần tử trong DOM bằng cách chỉ định (ID, class, v.v.)
from selenium.common.exceptions import NoSuchElementException, TimeoutException  # Xử lý lỗi khi scraping
from concurrent.futures import ThreadPoolExecutor, as_completed  # Thư viện để xử lý song song (multithreading)
from requests.adapters import HTTPAdapter  # Tăng cường khả năng kết nối HTTP với retry logic
from datetime import date, timedelta  # Hỗ trợ xử lý ngày tháng cụ thể, như tính toán khoảng thời gian
import sys

if 'google.colab' in sys.modules:
    from google.colab import drive
    drive.mount('/content/drive')
else:
    print("Đang chạy ngoài Google Colab – bỏ qua gắn Google Drive.")

# Bắt đầu phần chú thích chi tiết cho từng phần của code, nhằm đảm bảo tính minh bạch và tái tạo được trong nghiên cứu khoa học

#############################################
# PHẦN 1: XỬ LÝ BÀI BÁO (Scraping và kiểm tra từ khóa)
# Mục đích: Thu thập và phân tích nội dung bài báo từ trang web doanhnhansaigon, kiểm tra từ khóa liên quan đến nghiên cứu
#############################################

def configure_driver():
    """
    Cấu hình trình duyệt Chrome để chạy ở chế độ không hiển thị giao diện (headless), nhằm tối ưu hóa tài nguyên
    và phù hợp với môi trường server hoặc nghiên cứu tự động hóa quy mô lớn trên kinh te do thi.

    Returns:
        webdriver.Chrome: Đối tượng trình duyệt Chrome đã được cấu hình, sử dụng cho Selenium.
    """
    # Khởi tạo tùy chọn cho trình duyệt Chrome
    chrome_options = Options()
    # Thêm tham số để chạy ở chế độ không giao diện (headless), giảm sử dụng tài nguyên
    chrome_options.add_argument("--no-sandbox")  # Cho phép chạy trên môi trường không có sandbox (như server Linux)
    chrome_options.add_argument("--headless")  # Chạy không hiển thị giao diện, phù hợp cho scraping tự động
    chrome_options.add_argument("--disable-dev-shm-usage")  # Tối ưu hóa sử dụng bộ nhớ trên hệ thống Linux
    # Trả về trình duyệt Chrome đã cấu hình để sử dụng trong scraping
    return webdriver.Chrome(options=chrome_options)

def save_to_text_file(title, link, date, text, summary, folder_name="articles"):
    """
    Lưu nội dung bài báo vào file .txt để lưu trữ dữ liệu thô, hỗ trợ tái tạo và kiểm tra trong nghiên cứu khoa học.

    Args:
        title (str): Tiêu đề bài báo, được sử dụng làm tên file và đại diện cho nội dung nghiên cứu.
        link (str): Đường dẫn URL của bài báo, dùng để tham chiếu nguồn gốc từ doanhnhansaigon.
        date (str or datetime.date): Ngày xuất bản bài báo, định dạng theo chuẩn ISO hoặc chuỗi.
        text (str): Nội dung đầy đủ của bài báo, dùng để phân tích từ khóa.
        summary (str): Tóm tắt nội dung bài báo, tạo bởi thư viện Newspaper, hỗ trợ nghiên cứu nhanh.
        folder_name (str, optional): Tên thư mục lưu trữ file, mặc định là "articles". Defaults to "articles".
    """
    # Tạo tên file từ tiêu đề, loại bỏ các ký tự không hợp lệ trong tên file (như /, :, *, ?, ", <, >, |)
    filename = os.path.join(folder_name, re.sub(r'[\/:*?"<>|]', '', title) + '.txt')
    # Tạo thư mục nếu chưa tồn tại, đảm bảo không ghi đè dữ liệu
    os.makedirs(folder_name, exist_ok=True)
    # Mở file ở chế độ ghi (write), sử dụng mã hóa UTF-8 để hỗ trợ tiếng Việt
    with open(filename, 'w', encoding="utf-8") as file:
        # Ghi các thông tin bài báo theo định dạng rõ ràng, dễ đọc, phù hợp với tài liệu nghiên cứu
        file.write(f"Title: {title}\n\n")  # Ghi tiêu đề bài báo
        file.write(f"Article Link: {link}\n\n")  # Ghi URL bài báo từ VnEconomy
        file.write(f"Publish Date: {date}\n\n")  # Ghi ngày xuất bản
        file.write(f"Article:\n{text}\n\n")  # Ghi nội dung bài báo
        file.write(f"Summary: {summary}\n\n")  # Ghi tóm tắt bài báo
    # In thông báo xác nhận đã lưu file, hỗ trợ theo dõi tiến trình nghiên cứu
    print(f"Đã lưu bài báo: {filename}")

def is_within_date_range(date_obj):
    """
    Kiểm tra xem ngày của bài báo có nằm trong khoảng thời gian từ 01/01/2014 trở đi hay không,
    nhằm đảm bảo dữ liệu chỉ thu thập trong phạm vi nghiên cứu đã định từ kinh te do thi.

    Args:
        date_obj (datetime.date or str or None): Đối tượng ngày hoặc chuỗi đại diện cho ngày xuất bản từ bài báo.

    Returns:
        bool: True nếu ngày nằm trong khoảng từ 01/01/2014, False nếu không hoặc nếu ngày không hợp lệ.
    """
    # Kiểm tra nếu ngày là "Unknown" hoặc None, trả về False để loại bỏ bài báo không có ngày hợp lệ
    if date_obj == "Unknown" or date_obj is None:
        return False
    # Định nghĩa ngày bắt đầu của nghiên cứu (01/01/2014) để phù hợp với phạm vi nghiên cứu
    start_date = datetime.date(2014, 1, 1)
    # So sánh ngày bài báo với ngày bắt đầu, trả về True nếu bài báo đủ điều kiện
    return start_date <= date_obj

def get_matching_keywords(text):
    """
    Xác định các từ khóa trong văn bản dựa trên 4 danh mục (Hoa Kỳ, Trung Quốc, chủ đề, và căng thẳng),
    đồng thời loại bỏ bài viết nếu chứa từ khóa không liên quan.

    Args:
        text (str): Văn bản của bài báo (tiêu đề hoặc nội dung) để kiểm tra từ khóa.

    Returns:
        list: Danh sách các từ khóa khớp nếu có ít nhất một từ từ mỗi danh mục và không chứa từ không liên quan,
              ngược lại trả về danh sách rỗng.
    """
    # Danh sách từ khóa không liên quan
    unrelated_words = [
        # Phim/Điện ảnh
        "phim", "diễn viên", "đạo diễn","showbiz",
        "bóng đá", "cầu thủ", "trận đấu","tình yêu", "mối quan hệ", "hẹn hò",
        "người yêu", "chia tay", "hôn nhân",
        "game thủ",  "âm nhạc", "ca sĩ", "bài hát","ẩm thực", "món ăn",
        "nghỉ dưỡng","trẻ em", "cha mẹ","lễ cưới","đám tang"
    ]

    # Danh sách từ khóa chính
    us_words = ["hoa kỳ", "mỹ", "washington", "nhà trắng"]
    china_words = ["trung quốc", "bắc kinh"]
    topic_words = [
        "tăng thuế", "áp thuế", "tiền tệ", "wto", "tổ chức thương mại thế giới",
        "ngân khố", "nhà máy", "thặng dư", "nợ", "thuế quan",
        "rào cản", "khoản vay", "thâm hụt", "nhân dân tệ",
        "trợ cấp", "đô la", "xuất khẩu", "sao chép bất hợp pháp", "hack", "Huawei", "mạng", "đại sứ quán", "Hồng Kông",
        "trộm cắp", "nhập khẩu", "viễn thông", "vi phạm bản quyền", "lãnh sự quán", "nhập cư", "visa", "thẻ xanh",
        "khí hậu", "năng lượng", "bán phá giá", "bản quyền", "bằng sáng chế", "giam giữ", "biểu tình", "tự trị", "bắt giữ", "bầu cử", "đàn áp", "độc tài", "biểu tình", "phong trào",
        "quyền sở hữu trí tuệ", "tấn công mạng", "bảo hộ thương mại", "tỷ giá hối đoái", "môi trường", "tôn giáo", "Tây Tạng", "Đạt Lai", "quân sự", "an ninh", "quốc phòng", "khủng bố",
        "Lầu Năm Góc", "rocket", "tàu sân bay", "tàu chiến", "đánh chìm", "Triều Tiên", "tuần tra", "máy bay phản lực", "khẩn cấp", "kho vũ khí", "tình báo", "tàu khu trục", "khiêu khích",
        "hạt nhân", "trinh sát", "vũ khí", "tàu", "giám sát", "hạm đội", "sự cố", "liên minh", "đối thủ", "Liên Hợp Quốc", "bảo mật", "gián điệp", "radar", "tàu ngầm",
        "chuyến bay", "NATO", "phi công", "hiệp ước", "Thái Bình Dương", "lãnh thổ", "vệ tinh", "hàng hải", "quân đội", "Đài Loan", "bom",
        "Đài Bắc", "bạo lực", "Biển Đông", "đảo", "an ninh thông tin", "ô nhiễm", "tài chính", "giao dịch", "chuỗi cung ứng",
        "bảo hộ thương mại", "chủ nghĩa bảo hộ", "máy bay", "đạn đạo", "xây dựng lực lượng", "lưu vong", "đầu đạn", "tên lửa", "hải quân", "triển khai",
        "dược phẩm", "nông nghiệp", "công nghệ", "đơn phương", "đại dịch", "covid", "coronavirus", "viêm phổi", "song phương", "bán dẫn",
        "nhân quyền", "thương chiến", "kiểm soát xuất khẩu", "địa chính trị", "hàng rào thuế quan", "hạn chế xuất khẩu", "cuộc chiến thương mại"
    ]
    tension_words = [
        "đe dọa", "chiến tranh", "thù địch", "gia tăng căng thẳng", "leo thang căng thẳng", "giảm căng thẳng", "vi phạm", "kiện tụng", "trừng phạt",
        "khủng bố", "phá hủy", "tổn hại", "cấm vận", "lệnh cấm", "lệnh hạn chế", "giả mạo", "cáo buộc", "thao túng", "leo thang",
        "âm mưu", "đàn áp", "lạm dụng", "kìm hãm", "phản đối", "va chạm", "ngăn chặn", "khiêu khích", "mâu thuẫn", "chia rẽ",
        "đối thủ cạnh tranh", "đình trệ", "xung đột", "cảnh báo", "trộm cắp", "làm suy yếu", "áp đặt", "tranh cãi", "đáp trả", "nổ ra",
        "hung hăng", "kẻ thù", "phòng thủ", "chiến đấu", "xấu đi", "tranh chấp", "leo thang", "trả đũa", "phân cực", "chống lại", "đối trọng", "danh sách đen",
        "tồi tệ hơn", "đối kháng", "đối đầu", "giám sát", "cản trở", "chỉ trích", "nhượng bộ", "chi phối", "xung đột", "thống trị"
    ]

    text_lower = text.lower()

    # Kiểm tra từ khóa không liên quan trước
    for word in unrelated_words:
        if word in text_lower:
            return []  # Nếu tìm thấy từ không liên quan, trả về rỗng để loại bỏ bài viết

    # Kiểm tra từ khóa chính
    found_us = [word for word in us_words if word in text_lower]
    found_china = [word for word in china_words if word in text_lower]
    found_topic = [word for word in topic_words if word in text_lower]
    found_tension = [word for word in tension_words if word in text_lower]

    # Chỉ trả về kết quả nếu có ít nhất một từ từ mỗi danh mục
    if found_us and found_china and found_topic and found_tension:
        return list(set(found_us + found_china + found_topic + found_tension))
    return []

#############################################
# PHẦN 2: XỬ LÝ SITEMAP VÀ TRUY CẬP BÀI BÁO
# Mục đích: Thu thập danh sách URL bài báo từ sitemap của kinh te do thi và xử lý từng bài báo để trích xuất dữ liệu
#############################################

def get_sitemap_urls():
    """
    Trả về danh sách URL của các sitemap bài viết theo ngày từ Doanh Nhân Sài Gòn,
    từ 28/02/2025 đến 01/01/2014 (giảm dần), phù hợp với phạm vi thời gian nghiên cứu.

    Returns:
        list: Danh sách các URL sitemap được tạo theo ngày, từ mới nhất đến cũ nhất.
    """
    # Khởi tạo danh sách để lưu URL các sitemap
    sitemap_urls = []
    # Định nghĩa ngày bắt đầu (28/02/2025) và ngày kết thúc (01/01/2014) của nghiên cứu
    start_date = date(2023,12,31)  # Ngày mới nhất theo yêu cầu
    end_date = date(2023,10,10)  # Ngày bắt đầu của nghiên cứu

    # Tính khoảng thời gian giữa ngày bắt đầu và kết thúc (1 ngày giảm dần)
    delta = timedelta(days=1)
    current_date = start_date  # Bắt đầu từ ngày mới nhất

    # Lặp qua các ngày từ 28/02/2025 đến 01/01/2014, tạo URL sitemap cho từng ngày
    while current_date >= end_date:
        # Tạo URL sitemap theo định dạng của Doanh Nhân Sài Gòn
        url = f"https://doanhnhansaigon.vn/sitemap-article-{current_date.year}-{current_date.month:02d}-{current_date.day:02d}.xml"
        sitemap_urls.append(url)
        # Giảm ngày đi 1 ngày để tiếp tục vòng lặp
        current_date -= delta
    # Trả về danh sách các URL sitemap để xử lý tiếp
    return sitemap_urls

def get_article_urls_from_sitemap(sitemap_url):
    """
    Lấy danh sách URL bài báo từ một sitemap cụ thể của doanhnhansaigon, bao gồm cả ngày chỉnh sửa cuối (lastmod),
    để đảm bảo thu thập dữ liệu bài báo chính xác và có thời gian cập nhật.

    Args:
        sitemap_url (str): URL của file sitemap cần parse từ doanhnhansaigon.

    Returns:
        list: Danh sách các dictionary chứa thông tin URL và ngày chỉnh sửa cuối của bài báo.
    """
    # Thực hiện yêu cầu HTTP GET để lấy nội dung sitemap
    r = requests.get(sitemap_url)
    # Sử dụng BeautifulSoup để parse nội dung XML của sitemap
    soup = BeautifulSoup(r.content, "xml")  # Sử dụng parser XML để đảm bảo tính chính xác
    # Tìm tất cả các thẻ <url> trong sitemap
    url_tags = soup.find_all("url")
    # Khởi tạo danh sách để lưu thông tin bài báo
    articles = []
    # Lặp qua từng thẻ URL để trích xuất thông tin
    for tag in url_tags:
        # Lấy URL từ thẻ <loc>
        loc = tag.find("loc").text
        # Lấy ngày chỉnh sửa cuối (lastmod) nếu có
        lastmod = tag.find("lastmod").text if tag.find("lastmod") else None
        # Thêm thông tin bài báo vào danh sách
        articles.append({"url": loc, "lastmod": lastmod})
    # Trả về danh sách các bài báo từ sitemap
    return articles

def crawl_articles_from_sitemaps():
    """
    Quét qua tất cả các sitemap đã lọc từ kinh te do thi để thu thập thông tin bài báo,
    lọc chỉ lấy các bài báo có ngày từ 01/01/2014 trở đi, phù hợp với phạm vi nghiên cứu.

    Returns:
        list: Danh sách các dictionary chứa thông tin URL và ngày của bài báo thỏa mãn điều kiện.
    """
    # Lấy danh sách URL sitemap từ hàm get_sitemap_urls
    sitemap_urls = get_sitemap_urls()
    # In số lượng sitemap tìm thấy để theo dõi tiến trình nghiên cứu
    print(f"Đã tìm thấy {len(sitemap_urls)} sitemap từ doanhnhansaigon.")
    # Khởi tạo danh sách để lưu thông tin bài báo
    all_article_infos = []
    # Lặp qua từng sitemap để xử lý
    for sitemap_url in sitemap_urls:
        print(f"Đang xử lý sitemap: {sitemap_url}")
        # Lấy danh sách bài báo từ sitemap hiện tại
        articles = get_article_urls_from_sitemap(sitemap_url)
        # Xử lý từng bài báo trong sitemap
        for art in articles:
            # Chuyển đổi ngày lastmod thành đối tượng date nếu có
            if art["lastmod"]:
                try:
                    # Chuyển đổi chuỗi ISO 8601 thành đối tượng date, xử lý trường hợp kết thúc bằng 'Z'
                    art_date = datetime.datetime.fromisoformat(art["lastmod"].replace("Z", "+00:00")).date()
                except Exception as e:
                    # Nếu có lỗi trong quá trình chuyển đổi, gán date là None
                    art_date = None
            else:
                # Nếu không có lastmod, gán date là None
                art_date = None
            # Thêm ngày vào thông tin bài báo
            art["date"] = art_date
            # Kiểm tra và thêm bài báo vào danh sách nếu ngày >= 01/01/2014
            if art_date and art_date >= datetime.date(2014, 1, 1):
                all_article_infos.append(art)
    # In tổng số bài báo thu thập được để theo dõi
    print(f"Tổng số bài báo (theo sitemap) từ 01/01/2014: {len(all_article_infos)}")
    # Trả về danh sách bài báo thỏa mãn điều kiện nghiên cứu
    return all_article_infos

def process_article(url, sitemap_date, processed_urls, max_retries=3):
    """
    Xử lý một bài báo cụ thể từ kinh te do thi: trích xuất nội dung, kiểm tra từ khóa, và lưu nếu đáp ứng tiêu chí nghiên cứu.

    Args:
        url (str): URL của bài báo cần xử lý từ kinh te do thi.
        sitemap_date (datetime.date or None): Ngày chỉnh sửa cuối từ sitemap (lastmod).
        processed_urls (set): Tập hợp URL đã xử lý để tránh lặp lại trong nghiên cứu.
        max_retries (int, optional): Số lần thử lại khi gặp lỗi kết nối, mặc định là 3. Defaults to 3.

    Returns:
        dict or None: Thông tin bài báo nếu đáp ứng tiêu chí từ khóa, hoặc None nếu không thành công.
    """
    # Kiểm tra xem URL đã được xử lý chưa để tránh trùng lặp dữ liệu trong nghiên cứu
    if url in processed_urls:
        return None
    # Thêm URL vào tập hợp đã xử lý
    processed_urls.add(url)

    # Thử tải nội dung bài báo với cơ chế retry để xử lý lỗi kết nối, đảm bảo tính ổn định trong scraping
    for attempt in range(max_retries):
        try:
            # Thêm độ trễ 0.5 giây để tránh bị chặn bởi server của doanhnhansaigon
            time.sleep(0.5)
            # Thực hiện yêu cầu GET với timeout 30 giây để tránh treo chương trình
            response = requests.get(url, timeout=30)
            # Lấy nội dung HTML của bài báo
            html = response.text
            break  # Thoát vòng lặp nếu thành công
        except requests.exceptions.ReadTimeout as e:
            # Xử lý lỗi timeout, in thông báo và thử lại
            print(f"Timeout khi truy cập {url}. Thử lại ({attempt+1}/{max_retries})...")
            if attempt == max_retries - 1:
                # Nếu hết số lần thử, trả về None để bỏ qua bài báo
                return None
        except Exception as e:
            # Xử lý các lỗi khác (ví dụ: không kết nối được), trả về None
            return None

    try:
        # Không thu thập thông tin tác giả nữa để đơn giản hóa và tập trung vào nội dung nghiên cứu
        # Sử dụng thư viện Newspaper để trích xuất nội dung bài báo từ HTML
        article = Article(url, language="vi")  # Thiết lập ngôn ngữ tiếng Việt
        article.set_html(html)  # Gán nội dung HTML đã tải
        article.parse()  # Phân tích và trích xuất nội dung, tiêu đề
        try:
            # Tạo tóm tắt bài báo bằng NLP để hỗ trợ phân tích nhanh
            article.nlp()
        except LookupError as e:
            # Nếu chưa tải gói punkt của NLTK, tải xuống để xử lý tóm tắt
            nltk.download("punkt")
            article.nlp()

        # Sử dụng ngày từ sitemap (lastmod) làm ngày xuất bản, đảm bảo tính nhất quán với dữ liệu nguồn từ doanhnhansaigon
        date_obj = sitemap_date if sitemap_date else "Unknown"
        # Kiểm tra xem ngày có nằm trong phạm vi nghiên cứu không
        if not is_within_date_range(date_obj):
            return None

        # Lấy nội dung và tóm tắt từ bài báo
        text = article.text  # Nội dung đầy đủ để phân tích từ khóa
        summary = article.summary  # Tóm tắt nội dung để nghiên cứu nhanh
        title = article.title if article.title else ""  # Tiêu đề, mặc định rỗng nếu không có

        # Kiểm tra từ khóa trong tiêu đề hoặc nội dung để xác định bài báo có liên quan đến nghiên cứu không
        matches = get_matching_keywords(title) or get_matching_keywords(text)
        if matches:
            # Nếu bài báo chứa đủ từ khóa, lưu vào file .txt để lưu trữ dữ liệu thô
            save_to_text_file(title, url, date_obj, text, summary)
            # In thông báo để theo dõi tiến trình, bao gồm từ khóa khớp
            print(f"Đã duyệt bài: \"{title}\" (Ngày: {date_obj}) | Từ khóa: {matches}")
            # Trả về thông tin bài báo để tổng hợp sau này
            return {
                "title": title,  # Tiêu đề bài báo
                "url": url,  # Đường dẫn bài báo từ VnEconomy
                "date": str(date_obj),  # Ngày xuất bản dưới dạng chuỗi
                "text": text,  # Nội dung đầy đủ
                "summary": summary,  # Tóm tắt
                "matching": matches  # Danh sách từ khóa khớp
            }
        else:
            # Nếu không có từ khóa phù hợp, trả về None để bỏ qua bài báo
            return None

    except Exception as e:
        # Xử lý lỗi trong quá trình trích xuất nội dung, trả về None và in lỗi để theo dõi
        print(f"Lỗi khi xử lý bài báo {url}: {e}")
        return None

def process_article_wrapper(article_info, processed_urls, idx):
    """
    Hàm bao bọc để xử lý bài báo, in tiến trình và xử lý các trường hợp ngoại lệ,
    đảm bảo tính theo dõi trong nghiên cứu quy mô lớn từ doanhnhansaigon.

    Args:
        article_info (dict): Thông tin bài báo từ sitemap (URL và ngày).
        processed_urls (set): Tập hợp URL đã xử lý.
        idx (int): Chỉ số thứ tự của bài báo để theo dõi tiến trình nghiên cứu.

    Returns:
        dict or None: Kết quả xử lý bài báo hoặc thông tin URL và ngày nếu không đạt tiêu chí từ khóa.
    """
    # In thông báo để theo dõi tiến trình xử lý từng bài báo từ kinh te do thi
    print(f"Đang xử lý bài báo {idx}: {article_info['url']}")
    # Gọi hàm xử lý bài báo với thông tin ngày từ sitemap
    result = process_article(article_info["url"], article_info["date"], processed_urls)
    # Nếu bài báo được duyệt, in thông tin ngày và từ khóa
    if result and "date" in result:
        print(f"Bài báo {idx} có ngày {result['date']} và chứa từ khóa: {result.get('matching', [])}")
    # Nếu bài báo không đạt tiêu chí nhưng có ngày, trả về thông tin URL và ngày
    if result is None and article_info["date"]:
        result = {"url": article_info["url"], "date": str(article_info["date"])}
    # Trả về kết quả xử lý
    return result

def crawl_and_process_articles():
    """
    Thực hiện quét và xử lý tất cả bài báo từ các sitemap của doanhnhansaigon,
    sử dụng xử lý song song để tăng hiệu suất và thu thập dữ liệu nghiên cứu.

    Returns:
        tuple: Danh sách bài báo được duyệt (approved_articles) và danh sách tất cả bài báo từ sitemap.
    """
    # Lấy thông tin bài báo từ các sitemap của
    article_infos = crawl_articles_from_sitemaps()
    # Khởi tạo tập hợp để lưu URL đã xử lý, tránh trùng lặp
    processed_urls = set()
    # Khởi tạo danh sách để lưu bài báo được duyệt
    approved_articles = []

    # Sử dụng ThreadPoolExecutor để xử lý song song, tối ưu hóa thời gian thực thi trên kinh te do thi
    with ThreadPoolExecutor(max_workers=20) as executor:  # Sử dụng 20 worker để cân bằng hiệu suất và tải server
        # Tạo danh sách các tác vụ (futures) cho từng bài báo
        futures = {executor.submit(process_article_wrapper, info, processed_urls, idx): info
                   for idx, info in enumerate(article_infos, start=1)}  # Số thứ tự bắt đầu từ 1
        # Theo dõi và thu thập kết quả khi các tác vụ hoàn thành
        for future in as_completed(futures):
            # Lấy kết quả từ mỗi tác vụ
            result = future.result()
            # Nếu bài báo được duyệt (có tiêu đề), thêm vào danh sách approved_articles
            if result and "title" in result:
                approved_articles.append(result)
    # In số lượng bài báo được duyệt để theo dõi kết quả nghiên cứu
    print(f"Đã duyệt {len(approved_articles)} bài báo (đáp ứng tiêu chí từ khóa) từ sitemap của doanhnhansaigon.")
    # Trả về danh sách bài báo được duyệt và danh sách tất cả bài báo từ sitemap
    return approved_articles, article_infos

#############################################
# PHẦN 3: TẠO BẢNG DATAFRAME THEO THÁNG
# Mục đích: Tổng hợp dữ liệu bài báo và từ khóa từ doanhnhansaigon theo từng tháng để phân tích thống kê
#############################################

def get_month(date_obj):
    """
    Chuyển đổi đối tượng ngày thành chuỗi định dạng "Năm-MM" (ví dụ: "2021-03"),
    phù hợp với yêu cầu phân tích dữ liệu theo tháng trong nghiên cứu từ doanhnhansaigon.

    Args:
        date_obj (datetime.date): Đối tượng ngày cần chuyển đổi.

    Returns:
        str: Chuỗi định dạng "YYYY-MM" đại diện cho tháng, dùng để nhóm dữ liệu.
    """
    # Sử dụng định dạng "YYYY-MM" với tháng có 2 chữ số (padding 0) để đảm bảo thống nhất
    return f"{date_obj.year}-{date_obj.month:02d}"

def create_month_summary(approved_articles, all_article_infos):
    """
    Tạo bảng tổng hợp dữ liệu bài báo từ doanhnhansaigon theo tháng, bao gồm tổng số bài báo, số bài báo được duyệt,
    và tỷ lệ bài báo được duyệt so với tổng số, phục vụ phân tích xu hướng theo thời gian.

    Args:
        approved_articles (list): Danh sách các bài báo được duyệt theo tiêu chí từ khóa từ kinh te do thi.
        all_article_infos (list): Danh sách tất cả bài báo thu thập từ sitemap của kinh te do thi.

    Returns:
        pd.DataFrame: Bảng dữ liệu tổng hợp theo tháng với các cột: 'month', 'total_count', 'approved_count', 'approved_ratio'.
    """
    # Chuẩn bị dữ liệu cho tất cả bài báo từ sitemap
    all_data = []
    for info in all_article_infos:
        if info["date"]:  # Chỉ lấy bài báo có ngày hợp lệ
            all_data.append({"date": info["date"]})
    # Tạo DataFrame từ danh sách dữ liệu
    all_df = pd.DataFrame(all_data)
    if all_df.empty:  # Kiểm tra nếu không có dữ liệu
        return pd.DataFrame()
    # Chuyển đổi cột 'date' thành định dạng datetime để xử lý
    all_df['date'] = pd.to_datetime(all_df['date'])
    # Thêm cột 'month' theo định dạng "YYYY-MM"
    all_df['month'] = all_df['date'].dt.to_period("M").astype(str)

    # Chuẩn bị dữ liệu cho bài báo được duyệt
    approved_data = []
    for art in approved_articles:
        try:
            # Chuyển đổi ngày từ chuỗi "%Y-%m-%d" thành đối tượng datetime
            d = datetime.datetime.strptime(art["date"], "%Y-%m-%d")
            approved_data.append({"date": d})
        except Exception as e:
            # Bỏ qua bài báo nếu ngày không hợp lệ, ghi log lỗi nếu cần cho nghiên cứu
            print(f"Lỗi chuyển đổi ngày cho bài báo {art.get('url', 'Unknown')} từ doanhnhansaigon: {e}")
            continue
    # Tạo DataFrame từ danh sách bài báo được duyệt
    approved_df = pd.DataFrame(approved_data)
    if not approved_df.empty:
        # Thêm cột 'month' cho bài báo được duyệt
        approved_df['month'] = approved_df['date'].dt.to_period("M").astype(str)

    # Tính tổng số bài báo theo tháng
    total_summary = all_df.groupby("month").size().reset_index(name="total_count")
    # Tính số bài báo được duyệt theo tháng, nếu không có thì tạo DataFrame rỗng với cột phù hợp
    approved_summary = approved_df.groupby("month").size().reset_index(name="approved_count") if not approved_df.empty else pd.DataFrame(columns=["month", "approved_count"])
    # Gộp dữ liệu tổng và dữ liệu duyệt, điền giá trị thiếu bằng 0
    summary = pd.merge(total_summary, approved_summary, on="month", how="left")
    # Đảm bảo cột approved_count là số nguyên, điền giá trị 0 nếu thiếu
    summary["approved_count"] = summary["approved_count"].fillna(0).astype(int)
    # Tính tỷ lệ bài báo được duyệt so với tổng số bài báo
    summary["approved_ratio"] = summary["approved_count"] / summary["total_count"]
    # Trả về bảng tổng hợp theo tháng
    return summary

def create_keyword_frequency_df(approved_articles, keyword_list):
    """
    Tạo bảng tần suất xuất hiện của các từ khóa từ kinh te do thi theo tháng, tính toán tỷ lệ xuất hiện dựa trên số bài báo trong tháng,
    phục vụ phân tích tần suất và xu hướng từ khóa trong nghiên cứu.

    Args:
        approved_articles (list): Danh sách các bài báo được duyệt theo tiêu chí từ khóa từ kinh te do thi.
        keyword_list (list): Danh sách các từ khóa cần phân tích tần suất, dựa trên các chủ đề nghiên cứu.

    Returns:
        pd.DataFrame: Bảng dữ liệu tần suất từ khóa theo tháng với cột 'month' và các cột từ khóa.
    """
    # Chuẩn bị dữ liệu từ các bài báo được duyệt
    data = []
    for art in approved_articles:
        try:
            # Chuyển đổi ngày từ chuỗi "%Y-%m-%d" thành đối tượng datetime
            d = datetime.datetime.strptime(art["date"], "%Y-%m-%d")
            # Thêm thông tin ngày và danh sách từ khóa khớp vào danh sách dữ liệu
            data.append({"date": d, "matching": art.get("matching", [])})
        except Exception as e:
            # Bỏ qua bài báo nếu ngày không hợp lệ, ghi log lỗi nếu cần
            print(f"Lỗi chuyển đổi ngày cho bài báo {art.get('url', 'Unknown')} từ doanhnhansaigon: {e}")
            continue
    # Tạo DataFrame từ danh sách dữ liệu
    df = pd.DataFrame(data)
    if df.empty:  # Kiểm tra nếu không có dữ liệu
        return pd.DataFrame()
    # Thêm cột 'month' theo định dạng "YYYY-MM"
    df['month'] = df['date'].dt.to_period("M").astype(str)
    # Tạo cột nhị phân (0 hoặc 1) cho từng từ khóa, 1 nếu từ khóa xuất hiện trong danh sách khớp
    for kw in keyword_list:
        df[kw] = df['matching'].apply(lambda lst: 1 if kw in lst else 0)
    # Tính tổng số lần xuất hiện của từng từ khóa theo tháng
    freq = df.groupby('month').agg({kw: 'sum' for kw in keyword_list})
    # Tính tổng số bài báo theo tháng
    total = df.groupby('month').size()
    # Tính tỷ lệ xuất hiện của từng từ khóa (số lần xuất hiện / tổng số bài báo trong tháng)
    freq = freq.divide(total, axis=0)
    # Đặt lại chỉ số và chuyển thành DataFrame với cột 'month'
    freq.reset_index(inplace=True)
    # Trả về bảng tần suất từ khóa theo tháng
    return freq

#############################################
# PHẦN CHÍNH (MAIN): THỰC THI CHƯƠNG TRÌNH
# Mục đích: Kết hợp các bước scraping, xử lý, và tổng hợp dữ liệu từ kinh te do thi để tạo ra kết quả nghiên cứu
#############################################
if __name__ == "__main__":
    # Bước 1: Thực hiện quét và xử lý bài báo từ các sitemap của doanhnhansaigon
    # Mục tiêu: Thu thập dữ liệu bài báo liên quan đến nghiên cứu về căng thẳng thương mại
    approved_articles, all_article_infos = crawl_and_process_articles()

    # Bước 2: Lưu danh sách các bài báo được duyệt vào file JSON để lưu trữ dữ liệu thô,
    # hỗ trợ tái tạo và kiểm tra trong nghiên cứu
    with open("approved_articles_doanhnhansaigon.json", "w", encoding="utf-8") as f:
        # Sử dụng JSON để lưu trữ dữ liệu với định dạng dễ đọc, hỗ trợ tiếng Việt
        json.dump(approved_articles, f, ensure_ascii=False, indent=4)
    # In thông báo xác nhận đã lưu file
    print("\nĐã lưu kết quả vào file 'approved_articles_doanhnhansaigon.json'.")

    # Bước 3: Tạo bảng tổng hợp số bài báo theo tháng (tổng số và số được duyệt)
    # Mục tiêu: Phân tích xu hướng bài báo theo thời gian theo tháng từ kinhtedothi
    month_summary_df = create_month_summary(approved_articles, all_article_infos)
    # In bảng tổng hợp để theo dõi kết quả nghiên cứu
    print("\nBảng tổng hợp theo tháng (Approved / Total):")
    print(month_summary_df)

    # Bước 4: Tạo bảng tần suất từ khóa theo tháng để phân tích xu hướng từ khóa
    # Danh sách từ khóa được định nghĩa dựa trên các chủ đề nghiên cứu về thương mại và căng thẳng
    keyword_list = [ "tăng thuế", "áp thuế", "tiền tệ", "wto", "tổ chức thương mại thế giới",
        "ngân khố", "nhà máy", "thặng dư", "nợ", "thuế quan",
        "rào cản", "khoản vay", "thâm hụt", "nhân dân tệ",
        "trợ cấp", "đô la", "xuất khẩu", "sao chép bất hợp pháp", "hack", "Huawei", "mạng", "đại sứ quán", "Hồng Kông",
        "trộm cắp", "nhập khẩu", "viễn thông", "vi phạm bản quyền", "lãnh sự quán", "nhập cư", "visa", "thẻ xanh",
        "khí hậu", "năng lượng", "bán phá giá", "bản quyền", "bằng sáng chế", "giam giữ", "biểu tình", "tự trị", "bắt giữ", "bầu cử", "đàn áp", "độc tài", "biểu tình", "phong trào",
        "quyền sở hữu trí tuệ", "tấn công mạng", "bảo hộ thương mại", "tỷ giá hối đoái", "môi trường", "tôn giáo", "Tây Tạng", "Đạt Lai", "quân sự", "an ninh", "quốc phòng", "khủng bố",
        "Lầu Năm Góc", "rocket", "tàu sân bay", "tàu chiến", "đánh chìm", "Triều Tiên", "tuần tra", "máy bay phản lực", "khẩn cấp", "kho vũ khí", "tình báo", "tàu khu trục", "khiêu khích",
        "hạt nhân", "trinh sát", "vũ khí", "tàu", "giám sát", "hạm đội", "sự cố", "liên minh", "đối thủ", "Liên Hợp Quốc", "bảo mật", "gián điệp", "radar", "tàu ngầm",
        "chuyến bay", "NATO", "phi công", "hiệp ước", "Thái Bình Dương", "lãnh thổ", "vệ tinh", "hàng hải", "quân đội", "Đài Loan", "bom",
        "Đài Bắc", "bạo lực", "Biển Đông", "đảo", "an ninh thông tin", "ô nhiễm", "tài chính", "giao dịch", "chuỗi cung ứng",
        "bảo hộ thương mại", "chủ nghĩa bảo hộ", "máy bay", "đạn đạo", "xây dựng lực lượng", "lưu vong", "đầu đạn", "tên lửa", "hải quân", "triển khai",
        "dược phẩm", "nông nghiệp", "công nghệ", "đơn phương", "đại dịch", "covid", "coronavirus", "viêm phổi", "song phương", "bán dẫn",
        "nhân quyền", "thương chiến", "kiểm soát xuất khẩu", "địa chính trị", "hàng rào thuế quan", "hạn chế xuất khẩu", "cuộc chiến thương mại",
        "đe dọa", "chiến tranh", "thù địch", "gia tăng căng thẳng", "leo thang căng thẳng", "giảm căng thẳng", "vi phạm", "kiện tụng", "trừng phạt",
        "khủng bố", "phá hủy", "tổn hại", "cấm vận", "lệnh cấm", "lệnh hạn chế", "giả mạo", "cáo buộc", "thao túng", "leo thang",
        "âm mưu", "đàn áp", "lạm dụng", "kìm hãm", "phản đối", "va chạm", "ngăn chặn", "khiêu khích", "mâu thuẫn", "chia rẽ",
        "đối thủ cạnh tranh", "đình trệ", "xung đột", "cảnh báo", "trộm cắp", "làm suy yếu", "áp đặt", "tranh cãi", "đáp trả", "nổ ra",
        "hung hăng", "kẻ thù", "phòng thủ", "chiến đấu", "xấu đi", "tranh chấp", "leo thang", "trả đũa", "phân cực", "chống lại", "đối trọng", "danh sách đen",
        "tồi tệ hơn", "đối kháng", "đối đầu", "giám sát", "cản trở", "chỉ trích", "nhượng bộ", "chi phối", "xung đột", "thống trị"
    ]
    # Tạo bảng tần suất từ khóa theo tháng
    keyword_freq_df = create_keyword_frequency_df(approved_articles, keyword_list)
    # In bảng tần suất để theo dõi kết quả nghiên cứu
    print("\nBảng tần suất từ khóa theo tháng:")
    print(keyword_freq_df)

    #############################################
    # Bước 5: Lưu dữ liệu ra 3 file Excel tự động và tải về (nếu chạy trên Google Colab)
# Kết nối Google Drive
drive.mount('/content/drive')

# Định nghĩa đường dẫn thư mục trong Google Drive
drive_folder = "/content/drive/My Drive/BAIBAO_Reports/"

# Kiểm tra và tạo thư mục nếu chưa tồn tại
if not os.path.exists(drive_folder):
    os.makedirs(drive_folder)

# Lưu File 1: approved_articles
approved_articles_df = pd.DataFrame(approved_articles)
file_path1 = os.path.join(drive_folder, "approved_articles_doanhnhansaigon.xlsx")
approved_articles_df.to_excel(file_path1, index=False)
print(f"\nĐã lưu thông tin các bài báo được duyệt vào: {file_path1}")

# Lưu File 2: month_summary
file_path2 = os.path.join(drive_folder, "month_summary_doanhnhansaigon.xlsx")
month_summary_df.to_excel(file_path2, index=False)
print(f"Đã lưu bảng tổng hợp theo tháng vào: {file_path2}")

# Lưu File 3: keyword_frequency
file_path3 = os.path.join(drive_folder, "keyword_frequency_doanhnhansaigon.xlsx")
keyword_freq_df.to_excel(file_path3, index=False)
print(f"Đã lưu bảng tần suất từ khóa theo tháng vào: {file_path3}")
```

```python
# Import các thư viện cần thiết cho quá trình scraping, xử lý dữ liệu, và xuất file
import os  # Thư viện để xử lý hệ thống tệp và thư mục, hỗ trợ tạo và quản lý thư mục lưu bài báo
import re  # Thư viện để xử lý biểu thức chính quy (regular expressions), dùng để làm sạch dữ liệu như tên file
import time  # Thư viện để thêm độ trễ (delay) trong các yêu cầu HTTP, tránh bị chặn bởi server
import datetime  # Thư viện để xử lý thời gian và ngày tháng, hỗ trợ định dạng và so sánh ngày trong nghiên cứu
import random  # Thư viện để tạo số ngẫu nhiên, có thể sử dụng trong việc quản lý độ trễ ngẫu nhiên nếu cần
import json  # Thư viện để đọc/ghi dữ liệu dạng JSON, hỗ trợ lưu trữ cấu trúc dữ liệu phức tạp cho bài báo
import requests  # Thư viện để thực hiện các yêu cầu HTTP (GET, POST) đến các trang web của VnEconomy
import nltk  # Thư viện Natural Language Toolkit để xử lý ngôn ngữ tự nhiên, như tóm tắt bài báo
import pandas as pd  # Thư viện để xử lý và phân tích dữ liệu dạng bảng (DataFrame) cho phân tích thống kê
from bs4 import BeautifulSoup  # Thư viện để parse HTML/XML, trích xuất dữ liệu từ trang web
from newspaper import Article  # Thư viện để trích xuất nội dung bài báo (tiêu đề, nội dung, tóm tắt)
from selenium import webdriver  # Thư viện để tự động hóa trình duyệt, hỗ trợ scraping động cho các trang phức tạp
from selenium.webdriver.chrome.options import Options  # Cấu hình tùy chọn cho trình duyệt Chrome
from selenium.webdriver.common.by import By  # Hỗ trợ tìm phần tử trong DOM bằng cách chỉ định (ID, class, v.v.)
from selenium.common.exceptions import NoSuchElementException, TimeoutException  # Xử lý lỗi khi scraping
from concurrent.futures import ThreadPoolExecutor, as_completed  # Thư viện để xử lý song song (multithreading)
from requests.adapters import HTTPAdapter  # Tăng cường khả năng kết nối HTTP với retry logic
from datetime import date, timedelta  # Hỗ trợ xử lý ngày tháng cụ thể, như tính toán khoảng thời gian
import sys

if 'google.colab' in sys.modules:
    from google.colab import drive
    drive.mount('/content/drive')
else:
    print("Đang chạy ngoài Google Colab – bỏ qua gắn Google Drive.")

# Bắt đầu phần chú thích chi tiết cho từng phần của code, nhằm đảm bảo tính minh bạch và tái tạo được trong nghiên cứu khoa học

#############################################
# PHẦN 1: XỬ LÝ BÀI BÁO (Scraping và kiểm tra từ khóa)
# Mục đích: Thu thập và phân tích nội dung bài báo từ trang web doanhnhansaigon, kiểm tra từ khóa liên quan đến nghiên cứu
#############################################

def configure_driver():
    """
    Cấu hình trình duyệt Chrome để chạy ở chế độ không hiển thị giao diện (headless), nhằm tối ưu hóa tài nguyên
    và phù hợp với môi trường server hoặc nghiên cứu tự động hóa quy mô lớn trên kinh te do thi.

    Returns:
        webdriver.Chrome: Đối tượng trình duyệt Chrome đã được cấu hình, sử dụng cho Selenium.
    """
    # Khởi tạo tùy chọn cho trình duyệt Chrome
    chrome_options = Options()
    # Thêm tham số để chạy ở chế độ không giao diện (headless), giảm sử dụng tài nguyên
    chrome_options.add_argument("--no-sandbox")  # Cho phép chạy trên môi trường không có sandbox (như server Linux)
    chrome_options.add_argument("--headless")  # Chạy không hiển thị giao diện, phù hợp cho scraping tự động
    chrome_options.add_argument("--disable-dev-shm-usage")  # Tối ưu hóa sử dụng bộ nhớ trên hệ thống Linux
    # Trả về trình duyệt Chrome đã cấu hình để sử dụng trong scraping
    return webdriver.Chrome(options=chrome_options)

def save_to_text_file(title, link, date, text, summary, folder_name="articles"):
    """
    Lưu nội dung bài báo vào file .txt để lưu trữ dữ liệu thô, hỗ trợ tái tạo và kiểm tra trong nghiên cứu khoa học.

    Args:
        title (str): Tiêu đề bài báo, được sử dụng làm tên file và đại diện cho nội dung nghiên cứu.
        link (str): Đường dẫn URL của bài báo, dùng để tham chiếu nguồn gốc từ doanhnhansaigon.
        date (str or datetime.date): Ngày xuất bản bài báo, định dạng theo chuẩn ISO hoặc chuỗi.
        text (str): Nội dung đầy đủ của bài báo, dùng để phân tích từ khóa.
        summary (str): Tóm tắt nội dung bài báo, tạo bởi thư viện Newspaper, hỗ trợ nghiên cứu nhanh.
        folder_name (str, optional): Tên thư mục lưu trữ file, mặc định là "articles". Defaults to "articles".
    """
    # Tạo tên file từ tiêu đề, loại bỏ các ký tự không hợp lệ trong tên file (như /, :, *, ?, ", <, >, |)
    filename = os.path.join(folder_name, re.sub(r'[\/:*?"<>|]', '', title) + '.txt')
    # Tạo thư mục nếu chưa tồn tại, đảm bảo không ghi đè dữ liệu
    os.makedirs(folder_name, exist_ok=True)
    # Mở file ở chế độ ghi (write), sử dụng mã hóa UTF-8 để hỗ trợ tiếng Việt
    with open(filename, 'w', encoding="utf-8") as file:
        # Ghi các thông tin bài báo theo định dạng rõ ràng, dễ đọc, phù hợp với tài liệu nghiên cứu
        file.write(f"Title: {title}\n\n")  # Ghi tiêu đề bài báo
        file.write(f"Article Link: {link}\n\n")  # Ghi URL bài báo từ VnEconomy
        file.write(f"Publish Date: {date}\n\n")  # Ghi ngày xuất bản
        file.write(f"Article:\n{text}\n\n")  # Ghi nội dung bài báo
        file.write(f"Summary: {summary}\n\n")  # Ghi tóm tắt bài báo
    # In thông báo xác nhận đã lưu file, hỗ trợ theo dõi tiến trình nghiên cứu
    print(f"Đã lưu bài báo: {filename}")

def is_within_date_range(date_obj):
    """
    Kiểm tra xem ngày của bài báo có nằm trong khoảng thời gian từ 01/01/2014 trở đi hay không,
    nhằm đảm bảo dữ liệu chỉ thu thập trong phạm vi nghiên cứu đã định từ kinh te do thi.

    Args:
        date_obj (datetime.date or str or None): Đối tượng ngày hoặc chuỗi đại diện cho ngày xuất bản từ bài báo.

    Returns:
        bool: True nếu ngày nằm trong khoảng từ 01/01/2014, False nếu không hoặc nếu ngày không hợp lệ.
    """
    # Kiểm tra nếu ngày là "Unknown" hoặc None, trả về False để loại bỏ bài báo không có ngày hợp lệ
    if date_obj == "Unknown" or date_obj is None:
        return False
    # Định nghĩa ngày bắt đầu của nghiên cứu (01/01/2014) để phù hợp với phạm vi nghiên cứu
    start_date = datetime.date(2014, 1, 1)
    # So sánh ngày bài báo với ngày bắt đầu, trả về True nếu bài báo đủ điều kiện
    return start_date <= date_obj

def get_matching_keywords(text):
    """
    Xác định các từ khóa trong văn bản dựa trên 4 danh mục (Hoa Kỳ, Trung Quốc, chủ đề, và căng thẳng),
    đồng thời loại bỏ bài viết nếu chứa từ khóa không liên quan.

    Args:
        text (str): Văn bản của bài báo (tiêu đề hoặc nội dung) để kiểm tra từ khóa.

    Returns:
        list: Danh sách các từ khóa khớp nếu có ít nhất một từ từ mỗi danh mục và không chứa từ không liên quan,
              ngược lại trả về danh sách rỗng.
    """
    # Danh sách từ khóa không liên quan
    unrelated_words = [
        # Phim/Điện ảnh
        "phim", "diễn viên", "đạo diễn","showbiz",
        "bóng đá", "cầu thủ", "trận đấu","tình yêu", "mối quan hệ", "hẹn hò",
        "người yêu", "chia tay", "hôn nhân",
        "game thủ",  "âm nhạc", "ca sĩ", "bài hát","ẩm thực", "món ăn",
        "nghỉ dưỡng","trẻ em", "cha mẹ","lễ cưới","đám tang"
    ]

    # Danh sách từ khóa chính
    us_words = ["hoa kỳ", "mỹ", "washington", "nhà trắng"]
    china_words = ["trung quốc", "bắc kinh"]
    topic_words = [
        "tăng thuế", "áp thuế", "tiền tệ", "wto", "tổ chức thương mại thế giới",
        "ngân khố", "nhà máy", "thặng dư", "nợ", "thuế quan",
        "rào cản", "khoản vay", "thâm hụt", "nhân dân tệ",
        "trợ cấp", "đô la", "xuất khẩu", "sao chép bất hợp pháp", "hack", "Huawei", "mạng", "đại sứ quán", "Hồng Kông",
        "trộm cắp", "nhập khẩu", "viễn thông", "vi phạm bản quyền", "lãnh sự quán", "nhập cư", "visa", "thẻ xanh",
        "khí hậu", "năng lượng", "bán phá giá", "bản quyền", "bằng sáng chế", "giam giữ", "biểu tình", "tự trị", "bắt giữ", "bầu cử", "đàn áp", "độc tài", "biểu tình", "phong trào",
        "quyền sở hữu trí tuệ", "tấn công mạng", "bảo hộ thương mại", "tỷ giá hối đoái", "môi trường", "tôn giáo", "Tây Tạng", "Đạt Lai", "quân sự", "an ninh", "quốc phòng", "khủng bố",
        "Lầu Năm Góc", "rocket", "tàu sân bay", "tàu chiến", "đánh chìm", "Triều Tiên", "tuần tra", "máy bay phản lực", "khẩn cấp", "kho vũ khí", "tình báo", "tàu khu trục", "khiêu khích",
        "hạt nhân", "trinh sát", "vũ khí", "tàu", "giám sát", "hạm đội", "sự cố", "liên minh", "đối thủ", "Liên Hợp Quốc", "bảo mật", "gián điệp", "radar", "tàu ngầm",
        "chuyến bay", "NATO", "phi công", "hiệp ước", "Thái Bình Dương", "lãnh thổ", "vệ tinh", "hàng hải", "quân đội", "Đài Loan", "bom",
        "Đài Bắc", "bạo lực", "Biển Đông", "đảo", "an ninh thông tin", "ô nhiễm", "tài chính", "giao dịch", "chuỗi cung ứng",
        "bảo hộ thương mại", "chủ nghĩa bảo hộ", "máy bay", "đạn đạo", "xây dựng lực lượng", "lưu vong", "đầu đạn", "tên lửa", "hải quân", "triển khai",
        "dược phẩm", "nông nghiệp", "công nghệ", "đơn phương", "đại dịch", "covid", "coronavirus", "viêm phổi", "song phương", "bán dẫn",
        "nhân quyền", "thương chiến", "kiểm soát xuất khẩu", "địa chính trị", "hàng rào thuế quan", "hạn chế xuất khẩu", "cuộc chiến thương mại"
    ]
    tension_words = [
        "đe dọa", "chiến tranh", "thù địch", "gia tăng căng thẳng", "leo thang căng thẳng", "giảm căng thẳng", "vi phạm", "kiện tụng", "trừng phạt",
        "khủng bố", "phá hủy", "tổn hại", "cấm vận", "lệnh cấm", "lệnh hạn chế", "giả mạo", "cáo buộc", "thao túng", "leo thang",
        "âm mưu", "đàn áp", "lạm dụng", "kìm hãm", "phản đối", "va chạm", "ngăn chặn", "khiêu khích", "mâu thuẫn", "chia rẽ",
        "đối thủ cạnh tranh", "đình trệ", "xung đột", "cảnh báo", "trộm cắp", "làm suy yếu", "áp đặt", "tranh cãi", "đáp trả", "nổ ra",
        "hung hăng", "kẻ thù", "phòng thủ", "chiến đấu", "xấu đi", "tranh chấp", "leo thang", "trả đũa", "phân cực", "chống lại", "đối trọng", "danh sách đen",
        "tồi tệ hơn", "đối kháng", "đối đầu", "giám sát", "cản trở", "chỉ trích", "nhượng bộ", "chi phối", "xung đột", "thống trị"
    ]

    text_lower = text.lower()

    # Kiểm tra từ khóa không liên quan trước
    for word in unrelated_words:
        if word in text_lower:
            return []  # Nếu tìm thấy từ không liên quan, trả về rỗng để loại bỏ bài viết

    # Kiểm tra từ khóa chính
    found_us = [word for word in us_words if word in text_lower]
    found_china = [word for word in china_words if word in text_lower]
    found_topic = [word for word in topic_words if word in text_lower]
    found_tension = [word for word in tension_words if word in text_lower]

    # Chỉ trả về kết quả nếu có ít nhất một từ từ mỗi danh mục
    if found_us and found_china and found_topic and found_tension:
        return list(set(found_us + found_china + found_topic + found_tension))
    return []

#############################################
# PHẦN 2: XỬ LÝ SITEMAP VÀ TRUY CẬP BÀI BÁO
# Mục đích: Thu thập danh sách URL bài báo từ sitemap của kinh te do thi và xử lý từng bài báo để trích xuất dữ liệu
#############################################

def get_sitemap_urls(main_sitemap_url="https://tapchitaichinh.vn/sitemap.xml"):
    """
    Lấy danh sách URL của các sitemap từ trang web tapchitaichinh, lọc chỉ lấy các sitemap
    từ https://tapchitaichinh.vn/sitemap/news-2023-12.xml đến https://tapchitaichinh.vn/sitemap/news-2014-1.xml,
    phù hợp với phạm vi thời gian nghiên cứu.

    Args:
        main_sitemap_url (str, optional): URL của sitemap chính của tapchitaichinh, mặc định là "https://tapchitaichinh.vn/sitemap.xml".

    Returns:
        list: Danh sách các URL sitemap được lọc theo năm và tháng trong phạm vi nghiên cứu.
    """
    # Thực hiện yêu cầu HTTP GET để lấy nội dung sitemap chính
    r = requests.get(main_sitemap_url)
    # Sử dụng BeautifulSoup để parse nội dung XML của sitemap
    soup = BeautifulSoup(r.content, "xml")  # Sử dụng parser XML để đảm bảo tính chính xác
    # Tìm tất cả các thẻ <sitemap> trong sitemap chính
    sitemap_tags = soup.find_all("sitemap")
    # Trích xuất URL từ các thẻ <loc> trong sitemap
    sitemap_urls = [tag.find("loc").text for tag in sitemap_tags]

    # Định nghĩa URL bắt đầu và kết thúc theo phạm vi nghiên cứu
    start_url = "https://tapchitaichinh.vn/sitemap/news-2023-12.xml"  # Sitemap mới nhất
    end_url = "https://tapchitaichinh.vn/sitemap/news-2023-10.xml"  # Sitemap đầu tiên
    # Kiểm tra xem các URL bắt đầu và kết thúc có tồn tại trong danh sách không
    if start_url in sitemap_urls and end_url in sitemap_urls:
        # Tìm chỉ số của URL bắt đầu và kết thúc trong danh sách
        start_index = sitemap_urls.index(start_url)
        end_index = sitemap_urls.index(end_url)
        # Giả sử danh sách được sắp xếp từ mới nhất đến cũ nhất
        # Nếu start_index < end_index, lấy slice từ start_index đến end_index + 1, ngược lại lấy slice đảo
        if start_index <= end_index:
            filtered = sitemap_urls[start_index:end_index + 1]  # Lấy từ 2025-2 đến 2014-1
        else:
            filtered = sitemap_urls[end_index:start_index + 1]  # Lấy theo thứ tự ngược lại nếu cần
        return filtered  # Trả về danh sách sitemap đã lọc
    else:
        # Nếu không tìm thấy URL bắt đầu hoặc kết thúc, trả về danh sách rỗng
        print(f"Không tìm thấy {start_url} hoặc {end_url} trong sitemap chính.")
        return []


def get_article_urls_from_sitemap(sitemap_url):
    """
    Lấy danh sách URL bài báo từ một sitemap cụ thể của tapchitaichinh, bao gồm cả ngày chỉnh sửa cuối (lastmod),
    để đảm bảo thu thập dữ liệu bài báo chính xác và có thời gian cập nhật.

    Args:
        sitemap_url (str): URL của file sitemap cần parse từ tapchitaichinh.

    Returns:
        list: Danh sách các dictionary chứa thông tin URL và ngày chỉnh sửa cuối của bài báo.
    """
    # Thực hiện yêu cầu HTTP GET để lấy nội dung sitemap
    r = requests.get(sitemap_url)
    # Sử dụng BeautifulSoup để parse nội dung XML của sitemap
    soup = BeautifulSoup(r.content, "xml")  # Sử dụng parser XML để đảm bảo tính chính xác
    # Tìm tất cả các thẻ <url> trong sitemap
    url_tags = soup.find_all("url")
    # Khởi tạo danh sách để lưu thông tin bài báo
    articles = []
    # Lặp qua từng thẻ URL để trích xuất thông tin
    for tag in url_tags:
        # Lấy URL từ thẻ <loc>
        loc = tag.find("loc").text
        # Lấy ngày chỉnh sửa cuối (lastmod) nếu có
        lastmod = tag.find("lastmod").text if tag.find("lastmod") else None
        # Thêm thông tin bài báo vào danh sách
        articles.append({"url": loc, "lastmod": lastmod})
    # Trả về danh sách các bài báo từ sitemap
    return articles

def crawl_articles_from_sitemaps():
    """
    Quét qua tất cả các sitemap đã lọc từ kinh te do thi để thu thập thông tin bài báo,
    lọc chỉ lấy các bài báo có ngày từ 01/01/2014 trở đi, phù hợp với phạm vi nghiên cứu.

    Returns:
        list: Danh sách các dictionary chứa thông tin URL và ngày của bài báo thỏa mãn điều kiện.
    """
    # Lấy danh sách URL sitemap từ hàm get_sitemap_urls
    sitemap_urls = get_sitemap_urls()
    # In số lượng sitemap tìm thấy để theo dõi tiến trình nghiên cứu
    print(f"Đã tìm thấy {len(sitemap_urls)} sitemap từ tapchitaichinh.")
    # Khởi tạo danh sách để lưu thông tin bài báo
    all_article_infos = []
    # Lặp qua từng sitemap để xử lý
    for sitemap_url in sitemap_urls:
        print(f"Đang xử lý sitemap: {sitemap_url}")
        # Lấy danh sách bài báo từ sitemap hiện tại
        articles = get_article_urls_from_sitemap(sitemap_url)
        # Xử lý từng bài báo trong sitemap
        for art in articles:
            # Chuyển đổi ngày lastmod thành đối tượng date nếu có
            if art["lastmod"]:
                try:
                    # Chuyển đổi chuỗi ISO 8601 thành đối tượng date, xử lý trường hợp kết thúc bằng 'Z'
                    art_date = datetime.datetime.fromisoformat(art["lastmod"].replace("Z", "+00:00")).date()
                except Exception as e:
                    # Nếu có lỗi trong quá trình chuyển đổi, gán date là None
                    art_date = None
            else:
                # Nếu không có lastmod, gán date là None
                art_date = None
            # Thêm ngày vào thông tin bài báo
            art["date"] = art_date
            # Kiểm tra và thêm bài báo vào danh sách nếu ngày >= 01/01/2014
            if art_date and art_date >= datetime.date(2014, 1, 1):
                all_article_infos.append(art)
    # In tổng số bài báo thu thập được để theo dõi
    print(f"Tổng số bài báo (theo sitemap) từ 01/01/2014: {len(all_article_infos)}")
    # Trả về danh sách bài báo thỏa mãn điều kiện nghiên cứu
    return all_article_infos

def process_article(url, sitemap_date, processed_urls, max_retries=3):
    """
    Xử lý một bài báo cụ thể từ kinh te do thi: trích xuất nội dung, kiểm tra từ khóa, và lưu nếu đáp ứng tiêu chí nghiên cứu.

    Args:
        url (str): URL của bài báo cần xử lý từ kinh te do thi.
        sitemap_date (datetime.date or None): Ngày chỉnh sửa cuối từ sitemap (lastmod).
        processed_urls (set): Tập hợp URL đã xử lý để tránh lặp lại trong nghiên cứu.
        max_retries (int, optional): Số lần thử lại khi gặp lỗi kết nối, mặc định là 3. Defaults to 3.

    Returns:
        dict or None: Thông tin bài báo nếu đáp ứng tiêu chí từ khóa, hoặc None nếu không thành công.
    """
    # Kiểm tra xem URL đã được xử lý chưa để tránh trùng lặp dữ liệu trong nghiên cứu
    if url in processed_urls:
        return None
    # Thêm URL vào tập hợp đã xử lý
    processed_urls.add(url)

    # Thử tải nội dung bài báo với cơ chế retry để xử lý lỗi kết nối, đảm bảo tính ổn định trong scraping
    for attempt in range(max_retries):
        try:
            # Thêm độ trễ 0.5 giây để tránh bị chặn bởi server của tapchitaichinh
            time.sleep(0.5)
            # Thực hiện yêu cầu GET với timeout 30 giây để tránh treo chương trình
            response = requests.get(url, timeout=30)
            # Lấy nội dung HTML của bài báo
            html = response.text
            break  # Thoát vòng lặp nếu thành công
        except requests.exceptions.ReadTimeout as e:
            # Xử lý lỗi timeout, in thông báo và thử lại
            print(f"Timeout khi truy cập {url}. Thử lại ({attempt+1}/{max_retries})...")
            if attempt == max_retries - 1:
                # Nếu hết số lần thử, trả về None để bỏ qua bài báo
                return None
        except Exception as e:
            # Xử lý các lỗi khác (ví dụ: không kết nối được), trả về None
            return None

    try:
        # Không thu thập thông tin tác giả nữa để đơn giản hóa và tập trung vào nội dung nghiên cứu
        # Sử dụng thư viện Newspaper để trích xuất nội dung bài báo từ HTML
        article = Article(url, language="vi")  # Thiết lập ngôn ngữ tiếng Việt
        article.set_html(html)  # Gán nội dung HTML đã tải
        article.parse()  # Phân tích và trích xuất nội dung, tiêu đề
        try:
            # Tạo tóm tắt bài báo bằng NLP để hỗ trợ phân tích nhanh
            article.nlp()
        except LookupError as e:
            # Nếu chưa tải gói punkt của NLTK, tải xuống để xử lý tóm tắt
            nltk.download("punkt")
            article.nlp()

        # Sử dụng ngày từ sitemap (lastmod) làm ngày xuất bản, đảm bảo tính nhất quán với dữ liệu nguồn từ doanhnhansaigon
        date_obj = sitemap_date if sitemap_date else "Unknown"
        # Kiểm tra xem ngày có nằm trong phạm vi nghiên cứu không
        if not is_within_date_range(date_obj):
            return None

        # Lấy nội dung và tóm tắt từ bài báo
        text = article.text  # Nội dung đầy đủ để phân tích từ khóa
        summary = article.summary  # Tóm tắt nội dung để nghiên cứu nhanh
        title = article.title if article.title else ""  # Tiêu đề, mặc định rỗng nếu không có

        # Kiểm tra từ khóa trong tiêu đề hoặc nội dung để xác định bài báo có liên quan đến nghiên cứu không
        matches = get_matching_keywords(title) or get_matching_keywords(text)
        if matches:
            # Nếu bài báo chứa đủ từ khóa, lưu vào file .txt để lưu trữ dữ liệu thô
            save_to_text_file(title, url, date_obj, text, summary)
            # In thông báo để theo dõi tiến trình, bao gồm từ khóa khớp
            print(f"Đã duyệt bài: \"{title}\" (Ngày: {date_obj}) | Từ khóa: {matches}")
            # Trả về thông tin bài báo để tổng hợp sau này
            return {
                "title": title,  # Tiêu đề bài báo
                "url": url,  # Đường dẫn bài báo từ VnEconomy
                "date": str(date_obj),  # Ngày xuất bản dưới dạng chuỗi
                "text": text,  # Nội dung đầy đủ
                "summary": summary,  # Tóm tắt
                "matching": matches  # Danh sách từ khóa khớp
            }
        else:
            # Nếu không có từ khóa phù hợp, trả về None để bỏ qua bài báo
            return None

    except Exception as e:
        # Xử lý lỗi trong quá trình trích xuất nội dung, trả về None và in lỗi để theo dõi
        print(f"Lỗi khi xử lý bài báo {url}: {e}")
        return None

def process_article_wrapper(article_info, processed_urls, idx):
    """
    Hàm bao bọc để xử lý bài báo, in tiến trình và xử lý các trường hợp ngoại lệ,
    đảm bảo tính theo dõi trong nghiên cứu quy mô lớn từ tapchitaichinh.

    Args:
        article_info (dict): Thông tin bài báo từ sitemap (URL và ngày).
        processed_urls (set): Tập hợp URL đã xử lý.
        idx (int): Chỉ số thứ tự của bài báo để theo dõi tiến trình nghiên cứu.

    Returns:
        dict or None: Kết quả xử lý bài báo hoặc thông tin URL và ngày nếu không đạt tiêu chí từ khóa.
    """
    # In thông báo để theo dõi tiến trình xử lý từng bài báo từ kinh te do thi
    print(f"Đang xử lý bài báo {idx}: {article_info['url']}")
    # Gọi hàm xử lý bài báo với thông tin ngày từ sitemap
    result = process_article(article_info["url"], article_info["date"], processed_urls)
    # Nếu bài báo được duyệt, in thông tin ngày và từ khóa
    if result and "date" in result:
        print(f"Bài báo {idx} có ngày {result['date']} và chứa từ khóa: {result.get('matching', [])}")
    # Nếu bài báo không đạt tiêu chí nhưng có ngày, trả về thông tin URL và ngày
    if result is None and article_info["date"]:
        result = {"url": article_info["url"], "date": str(article_info["date"])}
    # Trả về kết quả xử lý
    return result

def crawl_and_process_articles():
    """
    Thực hiện quét và xử lý tất cả bài báo từ các sitemap của tapchitaichinh,
    sử dụng xử lý song song để tăng hiệu suất và thu thập dữ liệu nghiên cứu.

    Returns:
        tuple: Danh sách bài báo được duyệt (approved_articles) và danh sách tất cả bài báo từ sitemap.
    """
    # Lấy thông tin bài báo từ các sitemap của
    article_infos = crawl_articles_from_sitemaps()
    # Khởi tạo tập hợp để lưu URL đã xử lý, tránh trùng lặp
    processed_urls = set()
    # Khởi tạo danh sách để lưu bài báo được duyệt
    approved_articles = []

    # Sử dụng ThreadPoolExecutor để xử lý song song, tối ưu hóa thời gian thực thi trên tapchitaichinh
    with ThreadPoolExecutor(max_workers=20) as executor:  # Sử dụng 20 worker để cân bằng hiệu suất và tải server
        # Tạo danh sách các tác vụ (futures) cho từng bài báo
        futures = {executor.submit(process_article_wrapper, info, processed_urls, idx): info
                   for idx, info in enumerate(article_infos, start=1)}  # Số thứ tự bắt đầu từ 1
        # Theo dõi và thu thập kết quả khi các tác vụ hoàn thành
        for future in as_completed(futures):
            # Lấy kết quả từ mỗi tác vụ
            result = future.result()
            # Nếu bài báo được duyệt (có tiêu đề), thêm vào danh sách approved_articles
            if result and "title" in result:
                approved_articles.append(result)
    # In số lượng bài báo được duyệt để theo dõi kết quả nghiên cứu
    print(f"Đã duyệt {len(approved_articles)} bài báo (đáp ứng tiêu chí từ khóa) từ sitemap của tapchitaichinh.")
    # Trả về danh sách bài báo được duyệt và danh sách tất cả bài báo từ sitemap
    return approved_articles, article_infos

#############################################
# PHẦN 3: TẠO BẢNG DATAFRAME THEO THÁNG
# Mục đích: Tổng hợp dữ liệu bài báo và từ khóa từ tapchitaichinh theo từng tháng để phân tích thống kê
#############################################

def get_month(date_obj):
    """
    Chuyển đổi đối tượng ngày thành chuỗi định dạng "Năm-MM" (ví dụ: "2021-03"),
    phù hợp với yêu cầu phân tích dữ liệu theo tháng trong nghiên cứu từ tapchitaichinh.

    Args:
        date_obj (datetime.date): Đối tượng ngày cần chuyển đổi.

    Returns:
        str: Chuỗi định dạng "YYYY-MM" đại diện cho tháng, dùng để nhóm dữ liệu.
    """
    # Sử dụng định dạng "YYYY-MM" với tháng có 2 chữ số (padding 0) để đảm bảo thống nhất
    return f"{date_obj.year}-{date_obj.month:02d}"

def create_month_summary(approved_articles, all_article_infos):
    """
    Tạo bảng tổng hợp dữ liệu bài báo từ tapchitaichinh theo tháng, bao gồm tổng số bài báo, số bài báo được duyệt,
    và tỷ lệ bài báo được duyệt so với tổng số, phục vụ phân tích xu hướng theo thời gian.

    Args:
        approved_articles (list): Danh sách các bài báo được duyệt theo tiêu chí từ khóa từ tapchitaichinh.
        all_article_infos (list): Danh sách tất cả bài báo thu thập từ sitemap của tapchitaichinh.

    Returns:
        pd.DataFrame: Bảng dữ liệu tổng hợp theo tháng với các cột: 'month', 'total_count', 'approved_count', 'approved_ratio'.
    """
    # Chuẩn bị dữ liệu cho tất cả bài báo từ sitemap
    all_data = []
    for info in all_article_infos:
        if info["date"]:  # Chỉ lấy bài báo có ngày hợp lệ
            all_data.append({"date": info["date"]})
    # Tạo DataFrame từ danh sách dữ liệu
    all_df = pd.DataFrame(all_data)
    if all_df.empty:  # Kiểm tra nếu không có dữ liệu
        return pd.DataFrame()
    # Chuyển đổi cột 'date' thành định dạng datetime để xử lý
    all_df['date'] = pd.to_datetime(all_df['date'])
    # Thêm cột 'month' theo định dạng "YYYY-MM"
    all_df['month'] = all_df['date'].dt.to_period("M").astype(str)

    # Chuẩn bị dữ liệu cho bài báo được duyệt
    approved_data = []
    for art in approved_articles:
        try:
            # Chuyển đổi ngày từ chuỗi "%Y-%m-%d" thành đối tượng datetime
            d = datetime.datetime.strptime(art["date"], "%Y-%m-%d")
            approved_data.append({"date": d})
        except Exception as e:
            # Bỏ qua bài báo nếu ngày không hợp lệ, ghi log lỗi nếu cần cho nghiên cứu
            print(f"Lỗi chuyển đổi ngày cho bài báo {art.get('url', 'Unknown')} từ tapchitaichinh: {e}")
            continue
    # Tạo DataFrame từ danh sách bài báo được duyệt
    approved_df = pd.DataFrame(approved_data)
    if not approved_df.empty:
        # Thêm cột 'month' cho bài báo được duyệt
        approved_df['month'] = approved_df['date'].dt.to_period("M").astype(str)

    # Tính tổng số bài báo theo tháng
    total_summary = all_df.groupby("month").size().reset_index(name="total_count")
    # Tính số bài báo được duyệt theo tháng, nếu không có thì tạo DataFrame rỗng với cột phù hợp
    approved_summary = approved_df.groupby("month").size().reset_index(name="approved_count") if not approved_df.empty else pd.DataFrame(columns=["month", "approved_count"])
    # Gộp dữ liệu tổng và dữ liệu duyệt, điền giá trị thiếu bằng 0
    summary = pd.merge(total_summary, approved_summary, on="month", how="left")
    # Đảm bảo cột approved_count là số nguyên, điền giá trị 0 nếu thiếu
    summary["approved_count"] = summary["approved_count"].fillna(0).astype(int)
    # Tính tỷ lệ bài báo được duyệt so với tổng số bài báo
    summary["approved_ratio"] = summary["approved_count"] / summary["total_count"]
    # Trả về bảng tổng hợp theo tháng
    return summary

def create_keyword_frequency_df(approved_articles, keyword_list):
    """
    Tạo bảng tần suất xuất hiện của các từ khóa từ kinh te do thi theo tháng, tính toán tỷ lệ xuất hiện dựa trên số bài báo trong tháng,
    phục vụ phân tích tần suất và xu hướng từ khóa trong nghiên cứu.

    Args:
        approved_articles (list): Danh sách các bài báo được duyệt theo tiêu chí từ khóa từ tapchitaichinh.
        keyword_list (list): Danh sách các từ khóa cần phân tích tần suất, dựa trên các chủ đề nghiên cứu.

    Returns:
        pd.DataFrame: Bảng dữ liệu tần suất từ khóa theo tháng với cột 'month' và các cột từ khóa.
    """
    # Chuẩn bị dữ liệu từ các bài báo được duyệt
    data = []
    for art in approved_articles:
        try:
            # Chuyển đổi ngày từ chuỗi "%Y-%m-%d" thành đối tượng datetime
            d = datetime.datetime.strptime(art["date"], "%Y-%m-%d")
            # Thêm thông tin ngày và danh sách từ khóa khớp vào danh sách dữ liệu
            data.append({"date": d, "matching": art.get("matching", [])})
        except Exception as e:
            # Bỏ qua bài báo nếu ngày không hợp lệ, ghi log lỗi nếu cần
            print(f"Lỗi chuyển đổi ngày cho bài báo {art.get('url', 'Unknown')} từ tapchitaichinh: {e}")
            continue
    # Tạo DataFrame từ danh sách dữ liệu
    df = pd.DataFrame(data)
    if df.empty:  # Kiểm tra nếu không có dữ liệu
        return pd.DataFrame()
    # Thêm cột 'month' theo định dạng "YYYY-MM"
    df['month'] = df['date'].dt.to_period("M").astype(str)
    # Tạo cột nhị phân (0 hoặc 1) cho từng từ khóa, 1 nếu từ khóa xuất hiện trong danh sách khớp
    for kw in keyword_list:
        df[kw] = df['matching'].apply(lambda lst: 1 if kw in lst else 0)
    # Tính tổng số lần xuất hiện của từng từ khóa theo tháng
    freq = df.groupby('month').agg({kw: 'sum' for kw in keyword_list})
    # Tính tổng số bài báo theo tháng
    total = df.groupby('month').size()
    # Tính tỷ lệ xuất hiện của từng từ khóa (số lần xuất hiện / tổng số bài báo trong tháng)
    freq = freq.divide(total, axis=0)
    # Đặt lại chỉ số và chuyển thành DataFrame với cột 'month'
    freq.reset_index(inplace=True)
    # Trả về bảng tần suất từ khóa theo tháng
    return freq

#############################################
# PHẦN CHÍNH (MAIN): THỰC THI CHƯƠNG TRÌNH
# Mục đích: Kết hợp các bước scraping, xử lý, và tổng hợp dữ liệu từ kinh te do thi để tạo ra kết quả nghiên cứu
#############################################
if __name__ == "__main__":
    # Bước 1: Thực hiện quét và xử lý bài báo từ các sitemap của doanhnhansaigon
    # Mục tiêu: Thu thập dữ liệu bài báo liên quan đến nghiên cứu về căng thẳng thương mại
    approved_articles, all_article_infos = crawl_and_process_articles()

    # Bước 2: Lưu danh sách các bài báo được duyệt vào file JSON để lưu trữ dữ liệu thô,
    # hỗ trợ tái tạo và kiểm tra trong nghiên cứu
    with open("approved_articles_tapchitaichinh.json", "w", encoding="utf-8") as f:
        # Sử dụng JSON để lưu trữ dữ liệu với định dạng dễ đọc, hỗ trợ tiếng Việt
        json.dump(approved_articles, f, ensure_ascii=False, indent=4)
    # In thông báo xác nhận đã lưu file
    print("\nĐã lưu kết quả vào file 'approved_articles_tapchitaichinh.json'.")

    # Bước 3: Tạo bảng tổng hợp số bài báo theo tháng (tổng số và số được duyệt)
    # Mục tiêu: Phân tích xu hướng bài báo theo thời gian theo tháng từ tapchitaichinh
    month_summary_df = create_month_summary(approved_articles, all_article_infos)
    # In bảng tổng hợp để theo dõi kết quả nghiên cứu
    print("\nBảng tổng hợp theo tháng (Approved / Total):")
    print(month_summary_df)

    # Bước 4: Tạo bảng tần suất từ khóa theo tháng để phân tích xu hướng từ khóa
    # Danh sách từ khóa được định nghĩa dựa trên các chủ đề nghiên cứu về thương mại và căng thẳng
    keyword_list = [ "tăng thuế", "áp thuế", "tiền tệ", "wto", "tổ chức thương mại thế giới",
        "ngân khố", "nhà máy", "thặng dư", "nợ", "thuế quan",
        "rào cản", "khoản vay", "thâm hụt", "nhân dân tệ",
        "trợ cấp", "đô la", "xuất khẩu", "sao chép bất hợp pháp", "hack", "Huawei", "mạng", "đại sứ quán", "Hồng Kông",
        "trộm cắp", "nhập khẩu", "viễn thông", "vi phạm bản quyền", "lãnh sự quán", "nhập cư", "visa", "thẻ xanh",
        "khí hậu", "năng lượng", "bán phá giá", "bản quyền", "bằng sáng chế", "giam giữ", "biểu tình", "tự trị", "bắt giữ", "bầu cử", "đàn áp", "độc tài", "biểu tình", "phong trào",
        "quyền sở hữu trí tuệ", "tấn công mạng", "bảo hộ thương mại", "tỷ giá hối đoái", "môi trường", "tôn giáo", "Tây Tạng", "Đạt Lai", "quân sự", "an ninh", "quốc phòng", "khủng bố",
        "Lầu Năm Góc", "rocket", "tàu sân bay", "tàu chiến", "đánh chìm", "Triều Tiên", "tuần tra", "máy bay phản lực", "khẩn cấp", "kho vũ khí", "tình báo", "tàu khu trục", "khiêu khích",
        "hạt nhân", "trinh sát", "vũ khí", "tàu", "giám sát", "hạm đội", "sự cố", "liên minh", "đối thủ", "Liên Hợp Quốc", "bảo mật", "gián điệp", "radar", "tàu ngầm",
        "chuyến bay", "NATO", "phi công", "hiệp ước", "Thái Bình Dương", "lãnh thổ", "vệ tinh", "hàng hải", "quân đội", "Đài Loan", "bom",
        "Đài Bắc", "bạo lực", "Biển Đông", "đảo", "an ninh thông tin", "ô nhiễm", "tài chính", "giao dịch", "chuỗi cung ứng",
        "bảo hộ thương mại", "chủ nghĩa bảo hộ", "máy bay", "đạn đạo", "xây dựng lực lượng", "lưu vong", "đầu đạn", "tên lửa", "hải quân", "triển khai",
        "dược phẩm", "nông nghiệp", "công nghệ", "đơn phương", "đại dịch", "covid", "coronavirus", "viêm phổi", "song phương", "bán dẫn",
        "nhân quyền", "thương chiến", "kiểm soát xuất khẩu", "địa chính trị", "hàng rào thuế quan", "hạn chế xuất khẩu", "cuộc chiến thương mại",
        "đe dọa", "chiến tranh", "thù địch", "gia tăng căng thẳng", "leo thang căng thẳng", "giảm căng thẳng", "vi phạm", "kiện tụng", "trừng phạt",
        "khủng bố", "phá hủy", "tổn hại", "cấm vận", "lệnh cấm", "lệnh hạn chế", "giả mạo", "cáo buộc", "thao túng", "leo thang",
        "âm mưu", "đàn áp", "lạm dụng", "kìm hãm", "phản đối", "va chạm", "ngăn chặn", "khiêu khích", "mâu thuẫn", "chia rẽ",
        "đối thủ cạnh tranh", "đình trệ", "xung đột", "cảnh báo", "trộm cắp", "làm suy yếu", "áp đặt", "tranh cãi", "đáp trả", "nổ ra",
        "hung hăng", "kẻ thù", "phòng thủ", "chiến đấu", "xấu đi", "tranh chấp", "leo thang", "trả đũa", "phân cực", "chống lại", "đối trọng", "danh sách đen",
        "tồi tệ hơn", "đối kháng", "đối đầu", "giám sát", "cản trở", "chỉ trích", "nhượng bộ", "chi phối", "xung đột", "thống trị"
    ]
    # Tạo bảng tần suất từ khóa theo tháng
    keyword_freq_df = create_keyword_frequency_df(approved_articles, keyword_list)
    # In bảng tần suất để theo dõi kết quả nghiên cứu
    print("\nBảng tần suất từ khóa theo tháng:")
    print(keyword_freq_df)

    #############################################
    # Bước 5: Lưu dữ liệu ra 3 file Excel tự động và tải về (nếu chạy trên Google Colab)
# Kết nối Google Drive
drive.mount('/content/drive')

# Định nghĩa đường dẫn thư mục trong Google Drive
drive_folder = "/content/drive/My Drive/BAIBAO_Reports/"

# Kiểm tra và tạo thư mục nếu chưa tồn tại
if not os.path.exists(drive_folder):
    os.makedirs(drive_folder)

# Lưu File 1: approved_articles
approved_articles_df = pd.DataFrame(approved_articles)
file_path1 = os.path.join(drive_folder, "approved_articles_tapchitaichinh.xlsx")
approved_articles_df.to_excel(file_path1, index=False)
print(f"\nĐã lưu thông tin các bài báo được duyệt vào: {file_path1}")

# Lưu File 2: month_summary
file_path2 = os.path.join(drive_folder, "month_summary_tapchitaichinh.xlsx")
month_summary_df.to_excel(file_path2, index=False)
print(f"Đã lưu bảng tổng hợp theo tháng vào: {file_path2}")

# Lưu File 3: keyword_frequency
file_path3 = os.path.join(drive_folder, "keyword_frequency_tapchitaichinh.xlsx")
keyword_freq_df.to_excel(file_path3, index=False)
print(f"Đã lưu bảng tần suất từ khóa theo tháng vào: {file_path3}")
```
