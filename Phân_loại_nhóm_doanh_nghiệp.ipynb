```python
from google.colab import files
import numpy as np
import pandas as pd
```

```python
import io
# Tải tệp lên Google Colab
uploaded = files.upload(r"data.xlsx")

# Lấy tên tệp từ dictionary 'uploaded' và sử dụng io.BytesIO để đọc tệp
for filename in uploaded.keys():
    data = pd.read_excel(io.BytesIO(uploaded[filename]))

# Kiểm tra dữ liệu
data
```

```python
# Bước 1: Tính trung bình tổng tài sản cho mỗi cổ phiếu trong toàn giai đoạn
avg_assets = data.groupby('CP')['TỔNG TÀI SẢN'].mean().reset_index()
avg_assets.columns = ['CP', 'Trung binh tong tai san']
avg_assets
```

```python
avg_assets['Phan nhom'] = pd.qcut(avg_assets['Trung binh tong tai san'], 2, labels=['1', '2'])
avg_assets
```

```python
# Quy mô nhỏ
data1=avg_assets[avg_assets['Phan nhom']=='1']
list1=data1['CP'].tolist()
list1
```

```python
# Quy mô lớn
data2=avg_assets[avg_assets['Phan nhom']=='2']
list2=data2['CP'].tolist()
list2
```

```python
import io
# Tải tệp lên Google Colab
uploaded = files.upload(r"data.xlsx")

# Lấy tên tệp từ dictionary 'uploaded' và sử dụng io.BytesIO để đọc tệp
for filename in uploaded.keys():
    df1 = pd.read_excel(io.BytesIO(uploaded[filename]))

# Kiểm tra dữ liệu
df1
```

```python
df1['bp'] = 1 / df1['Price_to_Book']
df1
```

```python
avg_assets = df1.groupby('CP')['bp'].mean().reset_index()
avg_assets.columns = ['CP', 'Trung binh bp']
avg_assets
```

```python
avg_assets['phannhom']=pd.qcut(avg_assets['Trung binh bp'],3,['SL','SM','SH'])
avg_assets
```

```python
sl=avg_assets[avg_assets['phannhom']=='SL'].CP.tolist()
sm=avg_assets[avg_assets['phannhom']=='SM'].CP.tolist()
sh=avg_assets[avg_assets['phannhom']=='SH'].CP.tolist()
```
